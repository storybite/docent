````python
import requests
from bs4 import BeautifulSoup
import time

def collect_relic_links():
    """
    국립중앙박물관 웹사이트에서 문화재 링크 정보를 수집하는 함수
    
    Returns:
        list: 페이지 번호와 문화재 링크 정보를 담은 딕셔너리 리스트
    """
    base_url = "https://www.museum.go.kr/MUSEUM/contents/M0504000000.do"
    results = []
    
    # 37페이지 순회 (startCount는 0부터 시작하여 12씩 증가)
    for page in range(37):
        start_count = page * 12
        page_url = f"{base_url}?startCount={start_count}&searchId=treasure&schM=list"
        
        try:
            # 웹페이지 요청
            response = requests.get(page_url)
            response.raise_for_status()  # 오류 발생 시 예외 발생
            
            # BeautifulSoup으로 HTML 파싱
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 카드 리스트 찾기
            cards = soup.select('li.card')
            
            # 각 카드에서 링크 추출
            for card in cards:
                link_tag = card.select_one('div.txt a')
                if link_tag and 'href' in link_tag.attrs:
                    query_string = link_tag['href']
                    results.append({
                        "page": page + 1,  # 페이지 번호는 1부터 시작
                        "link": query_string
                    })
            
            print(f"페이지 {page + 1}/37 처리 완료: {len(cards)}개 항목 발견")
            
            # 서버 부하 방지를 위한 1초 대기
            #time.sleep(1)
            
        except Exception as e:
            print(f"페이지 {page + 1} 처리 중 오류 발생: {e}")
    
    return results

# 함수 사용 예시
if __name__ == "__main__":
    links = collect_relic_links()
    print(f"총 {len(links)}개의 링크를 수집했습니다.")
    print("첫 5개 링크 샘플:")
    for i in range(min(5, len(links))):
        print(links[i])
````

````python
import os
import json
import time
import shutil
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def extract_relic_data():
    """
    relic_links.json 파일을 읽어 각 전시물의 정보를 추출하여 JSON 형태로 반환하는 함수
    """
    base_url = "https://www.museum.go.kr/MUSEUM/contents/M0504000000.do"
    relic_index_json = {}
    
    # relic_links.json 파일 읽기
    try:
        with open('relic_links.json', 'r', encoding='utf-8') as f:
            relic_links = json.load(f)
    except FileNotFoundError:
        print("relic_links.json 파일을 찾을 수 없습니다.")
        return {}
    
    # 각 링크에 접근하여 데이터 추출
    for idx, item in enumerate(relic_links):
        if idx > 1000:
            break
        link = item['link']
        # relicId 추출
        relic_id = link.split('relicId=')[1]
        
        # URL 생성
        url = urljoin(base_url, link)
        
        try:
            # 웹 페이지 요청
            response = requests.get(url)
            response.raise_for_status()  # 오류 발생 시 예외 발생
            
            # BeautifulSoup으로 HTML 파싱
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 이미지 URL 추출 (첫 번째 이미지)
            image_tag = soup.select_one('a.swiper-slide')
            if image_tag:
                bg_url = image_tag['style']
                img_url = bg_url.split("url('")[1].split("')")[0]
            else:
                img_url = ""
            
            # 라벨 정보 추출
            label_data = {}
            outview_tit = soup.select_one('.outveiw-tit')
            if outview_tit:
                label_data["명칭"] = outview_tit.text.strip()
            
            outview_list = soup.select('.outview-list li')
            for li in outview_list:
                key = li.select_one('strong').text.strip()
                value = li.select_one('p').text.strip()
                label_data[key] = value
            
            # 내용 추출
            content = soup.select_one('.view-info-cont p')
            content_text = content.text.strip() if content else ""
            
            # 저작권 이미지 URL 추출
            copyright_img = soup.select_one('.codeView01 img')
            copyright_img_url = copyright_img['src'] if copyright_img else ""
            
            # 데이터 구성
            relic_index_json[relic_id] = {
                "url": url,
                "img": img_url,
                "label": label_data,
                "content": content_text,
                "copyright_img": copyright_img_url
            }
            
            print(f"Extracted data for relic ID: {relic_id}")
            
            # 서버 부하 방지를 위한 대기
            #time.sleep(1)
            
        except Exception as e:
            print(f"Error extracting data for {url}: {e}")
    
    return relic_index_json

def check_copyright(relic_index_json):
    """
    공공누리 유형 1 (출처표시)가 아닌 전시물을 필터링하는 함수
    """
    kogl_type1_url = "https://www.kogl.or.kr/open/web/images/images_2014/codetype/new_img_opentype01.png"
    filtered_data = {}
    
    for relic_id, data in relic_index_json.items():
        if data.get("copyright_img") == kogl_type1_url:
            filtered_data[relic_id] = data
    
    removed_count = len(relic_index_json) - len(filtered_data)
    print(f"Removed {removed_count} items that don't have KOGL Type 1 license")
    
    return filtered_data

def build_database(relic_index_json):
    """
    데이터베이스 폴더 구조를 생성하고 데이터를 저장하는 함수
    """
    # database 폴더 생성 (이미 존재하면 삭제 후 재생성)
    if os.path.exists('database'):
        shutil.rmtree('database')
    os.makedirs('database')
    
    # relic_index.json 파일 생성
    with open('database/relic_index.json', 'w', encoding='utf-8') as f:
        json.dump(relic_index_json, f, ensure_ascii=False, indent=4)
    
    # 각 전시물별 폴더 생성 및 데이터 저장
    for relic_id, data in relic_index_json.items():
        # 전시물 폴더 생성
        relic_dir = os.path.join('database', relic_id)
        os.makedirs(relic_dir, exist_ok=True)
        
        # relic_data.json 파일 생성
        with open(os.path.join(relic_dir, 'relic_data.json'), 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        
        # 이미지 다운로드
        img_url = data.get('img', '')
        if img_url:
            # 상대 경로를 절대 경로로 변환
            if img_url.startswith('/'):
                img_url = 'https://www.museum.go.kr' + img_url
            
            try:
                img_response = requests.get(img_url, stream=True)
                img_response.raise_for_status()
                
                # 이미지 파일명 추출
                img_filename = os.path.basename(img_url)
                img_path = os.path.join(relic_dir, img_filename)
                
                # 이미지 저장
                with open(img_path, 'wb') as img_file:
                    for chunk in img_response.iter_content(chunk_size=8192):
                        img_file.write(chunk)
                
                print(f"Downloaded image for relic ID: {relic_id}")
                
            except Exception as e:
                print(f"Error downloading image for relic ID {relic_id}: {e}")
        
        # 서버 부하 방지를 위한 대기
        time.sleep(0.5)
    
    print(f"Database built successfully with {len(relic_index_json)} items")

def main():
    # 전시물 데이터 추출
    print("Extracting relic data...")
    relic_index_json = extract_relic_data()
    
    # 공공누리 유형 점검
    print("Checking copyright...")
    filtered_data = check_copyright(relic_index_json)
    
    # 데이터베이스 구축
    print("Building database...")
    build_database(filtered_data)
    
    print("Process completed!")

if __name__ == "__main__":
    main()
````

````python
import os
import json
import requests
from bs4 import BeautifulSoup
import shutil
import re
import time
from urllib.parse import urljoin

def extract_relic_data():
    """
    relic_links.json 파일을 읽어 각 유물의 정보를 추출하여 relic_index_json 변수에 저장
    """
    base_url = "https://www.museum.go.kr/MUSEUM/contents/M0504000000.do"
    
    # relic_links.json 파일 읽기
    with open('relic_links.json', 'r', encoding='utf-8') as f:
        relic_links = json.load(f)
    
    relic_index_json = {}
    
    for idx, item in enumerate(relic_links):
        link = item['link']
        # relicId 추출
        relic_id = re.search(r'relicId=(\d+)', link).group(1)
        
        # URL 생성
        url = f"{base_url}{link}"
        
        print(f"Processing relic ID: {relic_id}, URL: {url}")
        
        try:
            # 웹 페이지 요청
            response = requests.get(url)
            response.raise_for_status()
            
            # HTML 파싱
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 이미지 URL 추출
            image_links = soup.select('a.swiper-slide')
            img_url = None
            if image_links:
                # 첫 번째 이미지의 background-url 추출
                style = image_links[0].get('style', '')
                img_match = re.search(r"url\('([^']+)'\)", style)
                if img_match:
                    img_url = img_match.group(1)
                    # 이미지 경로에서 /700/ 부분을 제거하여 원본 경로 추출
                    img_url = re.sub(r'/700/', '/', img_url)
            
            # 라벨 정보 추출
            label_data = {}
            outview_tit = soup.select_one('.outveiw-tit')
            if outview_tit:
                label_data['명칭'] = outview_tit.text.strip()
            
            outview_list = soup.select('.outview-list li')
            for li in outview_list:
                strong = li.select_one('strong')
                p = li.select_one('p')
                if strong and p:
                    key = strong.text.strip()
                    value = p.text.strip()
                    label_data[key] = value
            
            # 내용 추출
            content = soup.select_one('.view-info-cont p')
            content_text = content.text.strip() if content else ""
            
            # 저작권 이미지 URL 추출
            copyright_img = soup.select_one('.codeView01 img')
            copyright_img_url = copyright_img.get('src') if copyright_img else None
            
            # 데이터 구성
            relic_data = {
                "url": url,
                "img": img_url,
                "label": label_data,
                "content": content_text,
                "copyright_img": copyright_img_url
            }
            
            relic_index_json[relic_id] = relic_data
            
            # 요청 간 간격 두기
            #time.sleep(1)
            
        except Exception as e:
            print(f"Error processing relic ID {relic_id}: {e}")

        # if idx > 5:
        #     break
    
    return relic_index_json

relic_index_json = extract_relic_data()    
````
