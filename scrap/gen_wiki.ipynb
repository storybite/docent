````python
<div id="bodyContent" class="vector-body ve-init-mw-desktopArticleTarget-targetContainer" aria-labelledby="firstHeading" data-mw-ve-target-container="">
					<div class="vector-body-before-content">
							<div class="mw-indicators">
		</div>

						<div id="siteSub" class="noprint">위키백과, 우리 모두의 백과사전.</div>
					</div>
					<div id="contentSub"><div id="mw-content-subtitle"></div></div>
					
					
					<div id="mw-content-text" class="mw-body-content"><div class="mw-content-ltr mw-parser-output" lang="ko" dir="ltr"><p><span class="nowrap"></span>
</p>
<div class="dablink hatnote"><span typeof="mw:File"><a href="/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%8F%99%EC%9D%8C%EC%9D%B4%EC%9D%98%EC%96%B4_%EB%AC%B8%EC%84%9C" title="위키백과:동음이의어 문서"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Disambig_grey.svg/23px-Disambig_grey.svg.png" decoding="async" width="23" height="18" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Disambig_grey.svg/35px-Disambig_grey.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Disambig_grey.svg/46px-Disambig_grey.svg.png 2x" data-file-width="260" data-file-height="200"></a></span> 이 문서는 대한민국의 <a href="/wiki/%EB%8C%80%EC%A4%91%EA%B0%80%EC%9A%94" class="mw-redirect" title="대중가요">대중가요</a>에 관한 것입니다. 음악 그룹에 대해서는 <a href="/wiki/K-POP_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="K-POP (음악 그룹)">K-POP (음악 그룹)</a> 문서를 참고하십시오.</div>
<style data-mw-deduplicate="TemplateStyles:r36480591">.mw-parser-output .hlist dl,.mw-parser-output .hlist ol,.mw-parser-output .hlist ul{margin:0;padding:0}.mw-parser-output .hlist dd,.mw-parser-output .hlist dt,.mw-parser-output .hlist li{margin:0;display:inline}.mw-parser-output .hlist.inline,.mw-parser-output .hlist.inline dl,.mw-parser-output .hlist.inline ol,.mw-parser-output .hlist.inline ul,.mw-parser-output .hlist dl dl,.mw-parser-output .hlist dl ol,.mw-parser-output .hlist dl ul,.mw-parser-output .hlist ol dl,.mw-parser-output .hlist ol ol,.mw-parser-output .hlist ol ul,.mw-parser-output .hlist ul dl,.mw-parser-output .hlist ul ol,.mw-parser-output .hlist ul ul{display:inline}.mw-parser-output .hlist .mw-empty-li{display:none}.mw-parser-output .hlist dt::after{content:": "}.mw-parser-output .hlist dd::after,.mw-parser-output .hlist li::after{content:" · ";font-weight:bold}.mw-parser-output .hlist dd:last-child::after,.mw-parser-output .hlist dt:last-child::after,.mw-parser-output .hlist li:last-child::after{content:none}.mw-parser-output .hlist dd dd:first-child::before,.mw-parser-output .hlist dd dt:first-child::before,.mw-parser-output .hlist dd li:first-child::before,.mw-parser-output .hlist dt dd:first-child::before,.mw-parser-output .hlist dt dt:first-child::before,.mw-parser-output .hlist dt li:first-child::before,.mw-parser-output .hlist li dd:first-child::before,.mw-parser-output .hlist li dt:first-child::before,.mw-parser-output .hlist li li:first-child::before{content:" (";font-weight:normal}.mw-parser-output .hlist dd dd:last-child::after,.mw-parser-output .hlist dd dt:last-child::after,.mw-parser-output .hlist dd li:last-child::after,.mw-parser-output .hlist dt dd:last-child::after,.mw-parser-output .hlist dt dt:last-child::after,.mw-parser-output .hlist dt li:last-child::after,.mw-parser-output .hlist li dd:last-child::after,.mw-parser-output .hlist li dt:last-child::after,.mw-parser-output .hlist li li:last-child::after{content:")";font-weight:normal}.mw-parser-output .hlist ol{counter-reset:listitem}.mw-parser-output .hlist ol>li{counter-increment:listitem}.mw-parser-output .hlist ol>li::before{content:" "counter(listitem)"\a0 "}.mw-parser-output .hlist dd ol>li:first-child::before,.mw-parser-output .hlist dt ol>li:first-child::before,.mw-parser-output .hlist li ol>li:first-child::before{content:" ("counter(listitem)"\a0 "}.mw-parser-output .hlist-pipe dd:after,.mw-parser-output .hlist-pipe li:after{content:" | ";font-weight:normal}.mw-parser-output .hlist-hyphen dd:after,.mw-parser-output .hlist-hyphen li:after{content:" - ";font-weight:normal}.mw-parser-output .hlist-comma dd:after,.mw-parser-output .hlist-comma li:after{content:", ";font-weight:normal}.mw-parser-output .hlist-slash dd:after,.mw-parser-output .hlist-slash li:after{content:" / ";font-weight:normal}</style><style data-mw-deduplicate="TemplateStyles:r34311307">.mw-parser-output .infobox-subbox{padding:0;border:none;margin:-3px;width:auto;min-width:100%;font-size:100%;clear:none;float:none;background-color:transparent}.mw-parser-output .infobox-3cols-child{margin:auto}.mw-parser-output .infobox .navbar{font-size:100%}body.skin-minerva .mw-parser-output .infobox-header,body.skin-minerva .mw-parser-output .infobox-subheader,body.skin-minerva .mw-parser-output .infobox-above,body.skin-minerva .mw-parser-output .infobox-title,body.skin-minerva .mw-parser-output .infobox-image,body.skin-minerva .mw-parser-output .infobox-full-data,body.skin-minerva .mw-parser-output .infobox-below{text-align:center}</style><table class="infobox vevent haudio" style="text-align:left"><tbody><tr><th colspan="2" class="infobox-above summary" style="background: #CCCCFF;"><div style="position:relative; width:100%; height:40px; display:table;"><div style="position:absolute; top:0; right:0;"><span class="mw-default-size" typeof="mw:File"><span><img src="//upload.wikimedia.org/wikipedia/commons/thumb/6/60/Picto_infobox_music.png/40px-Picto_infobox_music.png" decoding="async" width="33" height="34" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/6/60/Picto_infobox_music.png 1.5x" data-file-width="44" data-file-height="45"></span></span></div>
<span style="position:relative; width:100%; display:table-cell; vertical-align:middle; color:black;">K-POP<br>케이팝</span></div></th></tr><tr><th scope="row" class="infobox-label" style="width:33%;">장르 뿌리</th><td class="infobox-data hlist"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><div class="hlist"><ul><li><a href="/wiki/%ED%95%9C%EA%B5%AD%EC%9D%8C%EC%95%85" title="한국음악">한국음악</a></li><li><a href="/wiki/%EB%B0%9C%EB%9D%BC%EB%93%9C_(%EB%8C%80%EC%A4%91_%EC%9D%8C%EC%95%85)" title="발라드 (대중 음악)">발라드</a></li><li><a href="/wiki/%EB%A6%AC%EB%93%AC_%EC%95%A4%EB%93%9C_%EB%B8%94%EB%A3%A8%EC%8A%A4" class="mw-redirect" title="리듬 앤드 블루스">리듬 앤드 블루스</a></li><li><a href="/wiki/%EC%9E%AC%EC%A6%88" title="재즈">재즈</a></li><li><a href="/wiki/%EB%AF%BC%EC%86%8D_%EC%9D%8C%EC%95%85" class="mw-redirect" title="민속 음악">민속 음악</a></li><li><a href="/wiki/%EC%84%9C%EC%96%91_%EA%B3%A0%EC%A0%84_%EC%9D%8C%EC%95%85" class="mw-redirect" title="서양 고전 음악">서양 고전 음악</a></li><li><a href="/wiki/%EB%A1%9C%ED%81%B0%EB%A1%A4" title="로큰롤">로큰롤</a></li><li><a href="/wiki/%EB%8C%84%EC%8A%A4_%ED%8C%9D" title="댄스 팝">댄스 팝</a></li><li><a href="/wiki/%EC%9D%BC%EB%A0%89%ED%8A%B8%EB%A1%9C%ED%8C%9D" title="일렉트로팝">일렉트로팝</a></li><li><a href="/wiki/%EC%9D%BC%EB%A0%89%ED%8A%B8%EB%A1%9C%EB%8B%89_%EB%8C%84%EC%8A%A4_%EC%9D%8C%EC%95%85" class="mw-redirect" title="일렉트로닉 댄스 음악">일렉트로닉 댄스 음악</a></li><li><a href="/wiki/%ED%9E%99%ED%95%A9" title="힙합">힙합</a></li><li><a href="/wiki/%ED%8A%B8%EB%A1%9C%ED%8A%B8" title="트로트">트로트</a></li></ul></div></td></tr><tr><th scope="row" class="infobox-label" style="width:33%;">문화 뿌리</th><td class="infobox-data"><a href="/wiki/%ED%8C%9D_%EC%9D%8C%EC%95%85" title="팝 음악">미국 팝</a>, <a href="/wiki/%ED%9D%91%EC%9D%B8_%EC%9D%8C%EC%95%85" class="mw-redirect" title="흑인 음악">흑인 음악</a>(<a href="/wiki/%ED%9E%99%ED%95%A9" title="힙합">힙합</a>), 1990년대 <a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD" title="대한민국">대한민국</a>의 대중 음악</td></tr><tr><th scope="row" class="infobox-label" style="width:33%;">사용 악기</th><td class="infobox-data hlist"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><div class="hlist"><ul><li><a href="/wiki/%EB%B3%B4%EC%BB%AC" title="보컬">보컬</a></li><li><a href="/wiki/%EB%93%9C%EB%9F%BC_%EB%A8%B8%EC%8B%A0" title="드럼 머신">드럼 머신</a></li><li><a href="/w/index.php?title=%EB%93%9C%EB%9F%BC_%ED%8C%A8%EB%93%9C&amp;action=edit&amp;redlink=1" class="new" title="드럼 패드 (없는 문서)">드럼 패드</a></li><li><a href="/wiki/%EB%93%9C%EB%9F%BC" class="mw-redirect" title="드럼">드럼</a></li><li><a href="/wiki/%EB%B2%A0%EC%9D%B4%EC%8A%A4_%EA%B8%B0%ED%83%80" title="베이스 기타">베이스 기타</a></li><li><a href="/wiki/%EA%B1%B4%EB%B0%98_%EC%95%85%EA%B8%B0" title="건반 악기">건반 악기</a></li><li><a href="/wiki/%EC%83%98%ED%94%8C%EB%9F%AC" title="샘플러">샘플러</a></li><li><a href="/wiki/%EC%8B%9C%ED%80%80%EC%84%9C" title="시퀀서">시퀀서</a></li><li><a href="/wiki/%EC%8B%A0%EC%8B%9C%EC%82%AC%EC%9D%B4%EC%A0%80" title="신시사이저">신시사이저</a></li></ul></div></td></tr><tr><th colspan="2" class="infobox-header" style="background: #CCCCFF;"><span style="color:black;">관련 항목</span></th></tr><tr><td colspan="2" class="infobox-full-data"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><div class="hlist"><ul><li><a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%8C%EC%95%85" title="대한민국의 음악">대한민국의 음악</a></li><li><a href="/wiki/K-ROCK" class="mw-redirect" title="K-ROCK">K-ROCK</a></li></ul></div></td></tr></tbody></table>
<p><b>K-pop</b>(<span title="IPA 발음" class="IPA" lang="und-Latn-fonipa">/keɪ pɔp/</span>, <span style="font-size: smaller;"><a href="/wiki/%EC%98%81%EC%96%B4" title="영어">영어</a>: </span><span lang="en">Korean Popular Music</span>, 케이팝)은 <a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD" title="대한민국">대한민국</a>에서 대중적 인기를 얻는 유행 음악을 말한다. K-pop은 현재 <a href="/wiki/%ED%95%9C%EB%A5%98" title="한류">한류</a>로 인해 전세계에서 인기가 많아지고 있으며, 주로 <a href="/wiki/%EB%8C%80%EB%A7%8C" title="대만">대만</a>, <a href="/wiki/%EB%AF%B8%EA%B5%AD" title="미국">미국</a>, <a href="/wiki/%ED%99%8D%EC%BD%A9" title="홍콩">홍콩</a>, <a href="/wiki/%EC%9D%BC%EB%B3%B8" title="일본">일본</a>, <a href="/wiki/%ED%95%84%EB%A6%AC%ED%95%80" title="필리핀">필리핀</a>, <a href="/wiki/%ED%83%9C%EA%B5%AD" title="태국">태국</a>, <a href="/wiki/%EC%9D%B8%EB%8F%84%EB%84%A4%EC%8B%9C%EC%95%84" title="인도네시아">인도네시아</a> 등지에서 매우 인기가 높으며 <a href="/wiki/%EC%9D%B8%EB%8F%84" title="인도">인도</a> 등 <a href="/wiki/%EB%82%A8%EC%95%84%EC%8B%9C%EC%95%84" title="남아시아">남아시아</a> 국가들과 <a href="/wiki/%EC%9C%A0%EB%9F%BD" title="유럽">유럽</a>, <a href="/wiki/%EB%9D%BC%ED%8B%B4%EC%95%84%EB%A9%94%EB%A6%AC%EC%B9%B4" title="라틴아메리카">라틴아메리카</a>, <a href="/wiki/%EC%A4%91%EB%8F%99" title="중동">중동</a> 등에서도 인기가 높아지고 있는 팝 음악이다. 간단히 <b>가요</b>(歌謠)라고 부르기도 하며, 영미권의 대중음악을 <a href="/wiki/%ED%8C%9D_%EC%9D%8C%EC%95%85" title="팝 음악">팝</a>(pop)이라고 부르고, 영미권 이외의 국가의 대중가요도 국가 이니셜을 붙여 나타내는 데서 K-POP이라고 쓴다. 특히 대한민국 대중가요가 해외에서 인기를 얻게 된 이후에 K-pop이라는 명칭이 대한민국 이외 국가에서 대한민국 대중가요를 가리키는 이름으로 널리 사용된다.
</p><p>1992년에 데뷔한 <a href="/wiki/%EC%84%9C%ED%83%9C%EC%A7%80%EC%99%80_%EC%95%84%EC%9D%B4%EB%93%A4" title="서태지와 아이들">서태지와 아이들</a>로부터, H.O.T., 젝스키스, 신화, god 등 1세대 아이돌의 음악을 시작으로 일레트로닉 사운드와 힙합의 그 당시에는 다소 낯설면서 중독성 있는 노래들을 선보이며 K-pop의 문을 열었고 이후 <a href="/wiki/2000%EB%85%84%EB%8C%80" title="2000년대">2000년대</a> 중반 <a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%95%84%EC%9D%B4%EB%8F%8C_%EA%B7%B8%EB%A3%B9_%EB%AA%A9%EB%A1%9D" title="대한민국의 아이돌 그룹 목록">대한민국의 2세대 아이돌 보이그룹과 걸 그룹들</a>이 대한민국, 외국에서 큰 성공을 거둔 이후, K-pop이라는 단어를 이런 보이그룹과 걸그룹의 음악만으로 한정해서 정의하는 경향이 많아졌다. 과거에는 한국 음악 전반을 아우르는 느낌이었으나 최근에는 그냥 한국에서 만들어진 댄스음악, 혹은 아이돌 음악 등을 일컫는 의미로 통용되고 있는 듯하다. 반면, 솔로 가수의 경우 넓은 의미의 K-pop 아티스트 범주에 들어가기도 하나, 이를 인정하지 아니하는 경우도 존재한다. 예를 들면 <a href="/wiki/%EC%95%84%EC%9D%B4%EC%9C%A0" title="아이유">아이유</a>와 같은 솔로가수들에게 "솔로 가수가 어떻게 k-pop 가수인가" 등의 비판적 목소리가 존재한다.
</p>
<meta property="mw:PageProp/toc">
<div class="mw-heading mw-heading2"><h2 id="역사"><span id=".EC.97.AD.EC.82.AC"></span>역사</h2><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=1" title="부분 편집: 역사"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<figure class="mw-halign-right" typeof="mw:File/Thumb"><a href="https://commons.wikimedia.org/wiki/File:Search_volume_for_kpop.svg" class="mw-file-description"><img src="//upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Search_volume_for_kpop.svg/330px-Search_volume_for_kpop.svg.png" decoding="async" width="300" height="91" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Search_volume_for_kpop.svg/500px-Search_volume_for_kpop.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Search_volume_for_kpop.svg/600px-Search_volume_for_kpop.svg.png 2x" data-file-width="794" data-file-height="240"></a><figcaption>시간에 따른 구글 트렌드 케이팝 검색량 변화</figcaption></figure>
<p><br>
1990년대에는 대한민국 음반 시장이 가장 호황을 누린 시절로 <a href="/wiki/%EA%B9%80%EA%B1%B4%EB%AA%A8" title="김건모">김건모</a>의 &lt;<a href="/wiki/%EC%9E%98%EB%AA%BB%EB%90%9C_%EB%A7%8C%EB%82%A8_(%EC%9D%8C%EB%B0%98)" class="mw-redirect" title="잘못된 만남 (음반)">잘못된 만남</a>&gt;은 단일 음대 판매량만 250만 장 이상을 기록해 한국 <a href="/wiki/%EA%B8%B0%EB%84%A4%EC%8A%A4_%EC%84%B8%EA%B3%84_%EA%B8%B0%EB%A1%9D" title="기네스 세계 기록">기네스</a>에 등재되었다.<sup id="cite_ref-1" class="reference"><a href="#cite_note-1"><span class="cite-bracket">[</span>1<span class="cite-bracket">]</span></a></sup> 하지만 <a href="/wiki/%EC%84%B8%EB%B8%90%ED%8B%B4_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="세븐틴 (음악 그룹)">세븐틴</a>의 미니 11집 SEVENTEENTH HEAVEN 이 컴백4일만에 462만 9,479장 이상의 판매고를 기록하면서 이 기록은 깨졌다. 같은 시기 <a href="/wiki/%EC%84%9C%ED%83%9C%EC%A7%80%EC%99%80_%EC%95%84%EC%9D%B4%EB%93%A4" title="서태지와 아이들">서태지와 아이들</a>, <a href="/wiki/%EC%8B%A0%EC%8A%B9%ED%9B%88" title="신승훈">신승훈</a> 등이 100만 장 이상 판매고를 돌파했다. 또한 1990년 <a href="/wiki/%EA%B9%80%EC%99%84%EC%84%A0" title="김완선">김완선</a>은 여자 가수로서 최초로 100만 장을 팔았다.<sup id="cite_ref-2" class="reference"><a href="#cite_note-2" title=""><span class="cite-bracket">[</span>2<span class="cite-bracket">]</span></a></sup> 80년대 후반과 90년대 초반 한국 대중음악의 질적 도약을 이끌었고 동시에 대중성의 영토를 확장한 <a href="/wiki/%EC%9C%A0%EC%9E%AC%ED%95%98" title="유재하">유재하</a>, <a href="/wiki/%EC%9D%B4%EC%8A%B9%EC%B2%A0" title="이승철">이승철</a>, <a href="/wiki/%EC%9D%B4%EC%8A%B9%ED%99%98" title="이승환">이승환</a>, <a href="/wiki/%EC%8B%A0%ED%95%B4%EC%B2%A0" title="신해철">신해철</a>, <a href="/wiki/015B" title="015B">015B</a>, <a href="/wiki/%EA%B9%80%ED%98%84%EC%B2%A0_(%EA%B0%80%EC%88%98)" title="김현철 (가수)">김현철</a>, <a href="/wiki/%EC%9C%A4%EC%83%81" title="윤상">윤상</a>, <a href="/wiki/%EC%9C%A0%EC%98%81%EC%84%9D" title="유영석">유영석</a>, <a href="/wiki/%EC%9C%A4%EC%A2%85%EC%8B%A0" title="윤종신">윤종신</a>, 그리고 <a href="/wiki/%EB%B4%84%EC%97%AC%EB%A6%84%EA%B0%80%EC%9D%84%EA%B2%A8%EC%9A%B8" title="봄여름가을겨울">봄여름가을겨울</a> 등 당시 젊은 싱어송 라이터 집단의 영향은 끊어지지 않고 <a href="/wiki/%EC%9C%A0%ED%9D%AC%EC%97%B4" title="유희열">유희열</a>, <a href="/wiki/%EC%A0%84%EB%9E%8C%ED%9A%8C_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="전람회 (음악 그룹)">전람회</a>로 이어졌으며, 홍대 일대를 중심으로 한 언더그라운드에 음악적 모범을 제시하거나 함께 공존하며 현대 K-pop의 중요한 한 축을 형성하고 있다.
</p><p>지금의 보이그룹과 걸그룹이 이끄는 K-pop은 <a href="/wiki/%EC%A0%9D%EC%8A%A4%ED%82%A4%EC%8A%A4" title="젝스키스">젝스키스</a>와 <a href="/wiki/SM_Entertainment" class="mw-redirect" title="SM Entertainment">SM Entertainment</a>의 아이돌 그룹들인 <a href="/wiki/H.O.T" class="mw-redirect" title="H.O.T">H.O.T</a>와 <a href="/wiki/%EC%8B%A0%ED%99%94_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="신화 (음악 그룹)">신화</a>로 거슬러간다. 이 그룹들이 1990년대 서태지와 아이들의 여러 요소를 혼합하는 음악적 방향과 보이그룹이라는 성공의 모델을 모방하면서 이후 한국 대중 음악 산업은 현재의 K-pop 산업으로 변모하기 시작한다. 이들은 일렉트릭 댄스와 랩, 때로 R&amp;B를 혼합한 장르로 주로 10대와 20대 위주의 팬덤으로부터 인기를 끌었다. 서태지와 아이들에 뒤지지 않는 음악성과 대중성을 획득한 <a href="/wiki/%EB%93%80%EC%8A%A4_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="듀스 (음악 그룹)">듀스</a>는 이후 자생적인 한국 힙합 씬의 모범이 되었다. <a href="/wiki/%EB%B0%95%EC%A7%84%EC%98%81" title="박진영">박진영</a>의 <a href="/wiki/JYP_Entertainment" class="mw-redirect" title="JYP Entertainment">JYP Entertainment</a> 또한 이들의 노력에 뒤지지 않았고 서태지와 아이들의 유산은 이후 <a href="/wiki/%EC%96%91%ED%98%84%EC%84%9D" title="양현석">양현석</a>의 <a href="/wiki/YG_Entertainment" class="mw-redirect" title="YG Entertainment">YG Entertainment</a>를 중심으로 한 웰메이드 오버그라운드 아이돌 음악이라는 현대 K-pop의 한 축을 형성했고 음악성과 실력 면에서 중요한 부분을 차지하고 있다.
</p>
<div class="mw-heading mw-heading3"><h3 id="2000년대"><span id="2000.EB.85.84.EB.8C.80"></span>2000년대</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=2" title="부분 편집: 2000년대"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<figure class="mw-default-size" typeof="mw:File/Thumb"><a href="https://commons.wikimedia.org/wiki/File:DJ_DOC_@_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg" class="mw-file-description"><img src="//upload.wikimedia.org/wikipedia/commons/thumb/2/22/DJ_DOC_%40_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg/170px-DJ_DOC_%40_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg" decoding="async" width="170" height="255" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/2/22/DJ_DOC_%40_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg/255px-DJ_DOC_%40_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/2/22/DJ_DOC_%40_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg/340px-DJ_DOC_%40_Cyworld_Dream_Music_Festival_%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%93%9C%EB%A6%BC_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C_36.jpg 2x" data-file-width="1365" data-file-height="2048"></a><figcaption>1990년대 인기 힙합 트리오 중 하나인 네온-오렌지 반바지와 폴로 셔츠를 매치한 채 팔을 들어올린 채 무대에서 공연하는 DJ DOC.<sup id="cite_ref-3" class="reference"><a href="#cite_note-3"><span class="cite-bracket">[</span>3<span class="cite-bracket">]</span></a></sup></figcaption></figure>
<p>2000년대 들어서는 음반 시장이 급격히 작아지면서 음반 판매량 20만 장을 넘기도 힘들어졌다. 특히 불법 음원시장이 기승을 펴면서 음반시장이 급격히 축소되었고 정식 음원시장이 출시된 2004년 이전까지 가요계는 이로 인한 직격탄을 맞았다. 결국 2013년 엑소 이전까지 밀리언셀러는 2001년이 마지막이 되었고, 2003년부터는 디지털 음원시장이 음반시장의 규모를 추월하게 되었다. 2004년, 정식 온라인 음원 시장이 출시되면서 음반 시장은 완전히 쇠퇴기를 맞게 되면서, 가요 시장 역시 온라인 시장 위주로 재편되었다.<sup id="cite_ref-4" class="reference"><a href="#cite_note-4"><span class="cite-bracket">[</span>4<span class="cite-bracket">]</span></a></sup> 그러나 정확히 12년 후인 2013년, 그룹 엑소의 1집 앨범이 2001년 김건모 7집, 그룹 god 4집 이후 우리나라 가요계 12년 만에 100만장을 돌파하며 음반 시장에 다시금 활력을 불어넣었다. 이후 발매하는 앨범마다 100만장의 판매고를 올리며 퀸터플 밀리언셀러라는 사상 초유의 기록을 세웠다. 현재 엑소의 누적 음반 판매량은 1000만장이 넘어가며, 이는 21세기 데뷔한 가수 중 최초의 기록이다. 이후 방탄소년단도 큰 인기를 얻으며 단일 음반 최다 판매량 기록을 24년 만에 갈아치웠고, 엑소에 이어 두 번째로 누적 판매량 1000만장을 돌파하였다. 현재 방탄소년단의 통산 음반 판매량은 1500만장을 넘어섰다. <a href="/wiki/%EC%9B%90%EB%8D%94%EA%B1%B8%EC%8A%A4" title="원더걸스">원더걸스</a>는 'Nobody'로 미국 진출을 선언 한 뒤, 한국인으로서는 최초로 <a href="/wiki/%EB%B9%8C%EB%B3%B4%EB%93%9C_%ED%95%AB_100" title="빌보드 핫 100">빌보드 핫 100</a>에 76위로 진입했다.<sup id="cite_ref-5" class="reference"><a href="#cite_note-5"><span class="cite-bracket">[</span>5<span class="cite-bracket">]</span></a></sup><style data-mw-deduplicate="TemplateStyles:r36482206">.mw-parser-output .fix{background-color:#fff9f9;color:DarkSlateGray;border:1px solid #ffdcdc}</style><span title="모호한 표현" class="fix">이 외에 2010년대 들어서는 많은 그룹들이 일본 진출을 선언했는데</span><sup class="정리_필요 noprint">[<a href="/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%AA%A8%ED%98%B8%ED%95%9C_%ED%91%9C%ED%98%84%EC%9D%84_%EC%82%BC%EA%B0%80_%EC%A3%BC%EC%84%B8%EC%9A%94" class="mw-redirect" title="위키백과:모호한 표현을 삼가 주세요">모호한 표현</a>]</sup>, <a href="/wiki/%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80" title="소녀시대">소녀시대</a>는 'Gee'로 일본 <a href="/wiki/%EC%98%A4%EB%A6%AC%EC%BD%98" title="오리콘">오리콘</a> 싱글 차트 일간 1위에 올랐다.<sup id="cite_ref-6" class="reference"><a href="#cite_note-6"><span class="cite-bracket">[</span>6<span class="cite-bracket">]</span></a></sup> 이 외에 <a href="/wiki/%EC%B9%B4%EB%9D%BC_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="카라 (음악 그룹)">카라</a>도 오리콘 차트에 오르며 한류 열풍을 일으켰다. 2012년 가수 <a href="/wiki/%EC%8B%B8%EC%9D%B4" title="싸이">싸이</a>는 '강남스타일'로 미국 빌보드의 싱글차트인 핫100차트에서 7주 연속 2위를 차지했다. 2015년 이후 서구권에서 K-POP 팬층을 중심으로 은근히 약진을 거듭하며 실적을 쌓아가고 있던 중, 현재 아이돌들이 미국을 비롯한 북남미, 유럽 등 해외 지역에서 가장 가시적인 성과를 내고 있다.
</p><p>아이돌 음악으로 분류하기는 애매한 경우가 있지만 현지 프로듀서 측의 러브콜을 받고 미국으로 진출을 기획하거나 실제로 진행했던 사례도 몇몇 있기는 했다. 보아나 세븐 같은 가수들이 대표적인 케이스이다
</p><p>다만, 현지 사정을 비롯해 여러 가지 이유로 유의미한 성과를 거두지 못하고 대부분 씁쓸히 한국으로 돌아온 경우가 대부분이었다. 그 뿐 아니라 대중들에게 가장 유명한 사례였던 원더걸스의 경우에는 소속사 차원에서 미국 진출을 추진했지만 자국에서의 어마어마한 성공과 달리 현지에서는 이렇다 할 반응을 얻지 못하고 귀국해야 했다.
</p><p>K-POP 그룹 중에서는 <a href="/wiki/%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80" title="소녀시대">소녀시대</a>의 Gee <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오가 최초로 K-POP 유튜브 뮤직비디오 조회수 1억 회를 돌파했다. 남자 솔로 가수는 <a href="/wiki/%EC%8B%B8%EC%9D%B4" title="싸이">싸이</a>를 이어서 <a href="/wiki/G-DRAGON" class="mw-redirect" title="G-DRAGON">G-DRAGON</a>의 삐딱하게 (Crooked) 유튜브 뮤직비디오와 니가 뭔데 (Who You?) 유튜브 뮤직비디오와 무제(無題) (Untitled, 2014) 유튜브 뮤직비디오, <a href="/wiki/%ED%83%9C%EC%96%91_(%EA%B0%80%EC%88%98)" title="태양 (가수)">태양</a>의 눈, 코, 입 유튜브 뮤직비디오와 링가 링가 (RINGA LINGA) 유튜브 뮤직비디오, <a href="/wiki/%EC%A0%9C%EC%9D%B4%ED%99%89" title="제이홉">제이홉</a>의 Daydream (백일몽) 유튜브 뮤직비디오가 각각 유튜브 조회수 1억 회를 달성했다. 여자 솔로가수는 <a href="/wiki/%ED%98%84%EC%95%84" title="현아">현아</a>의 Bubble Pop! <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오와 Ice Cream (Feat. Maboos) 유튜브 뮤직비디오와 잘나가서 그래 (Feat. 정일훈 Of BTOB) 유튜브 뮤직비디오, <a href="/wiki/%EC%95%84%EC%9D%B4%EC%9C%A0" title="아이유">아이유</a>의 팔레트 (Feat. G-DRAGON) <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오와 삐삐 유튜브 뮤직비디오, <a href="/wiki/%ED%83%9C%EC%97%B0" title="태연">태연</a>의 I (Feat. 버벌진트) 유튜브 뮤직비디오, <a href="/wiki/%EC%A0%9C%EB%8B%88_(1996%EB%85%84)" title="제니 (1996년)">제니</a>의 SOLO 유튜브 뮤직비디오, <a href="/wiki/%EC%84%A0%EB%AF%B8" title="선미">선미</a>의 가시나 <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오가 각각 <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 조회수 1억 회를 달성했다.
</p><p><a href="/wiki/K-POP" class="mw-redirect" title="K-POP">K-POP</a> 그룹 중에서 <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오 조회수 2억 회를 최초로 돌파한 것은 <a href="/wiki/%EB%B9%85%EB%B1%85_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="빅뱅 (음악 그룹)">BIGBANG</a>의 FANTASTIC BABY 유튜브 뮤직비디오였다. 그리고 <a href="/wiki/K-POP" class="mw-redirect" title="K-POP">K-POP</a> <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오 조회수 3억 회도 <a href="/wiki/%EB%B9%85%EB%B1%85_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="빅뱅 (음악 그룹)">BIGBANG</a>의 FANTASTIC BABY <a href="/wiki/%EC%9C%A0%ED%8A%9C%EB%B8%8C" title="유튜브">유튜브</a> 뮤직비디오가 최초로 달성하였다.
</p><p><a href="/wiki/%EA%B1%B8_%EA%B7%B8%EB%A3%B9" title="걸 그룹">걸 그룹</a> 중에는 <a href="/wiki/%ED%8A%B8%EC%99%80%EC%9D%B4%EC%8A%A4" title="트와이스">트와이스</a>와 <a href="/wiki/%EB%B8%94%EB%9E%99%ED%95%91%ED%81%AC" class="mw-redirect" title="블랙핑크">블랙핑크</a>의 사례가 대표적이다. <a href="/wiki/%ED%8A%B8%EC%99%80%EC%9D%B4%EC%8A%A4" title="트와이스">트와이스</a>는 <a href="/wiki/K-POP" class="mw-redirect" title="K-POP">K-POP</a> 아이돌 최초 데뷔곡 뮤직비디오 조회수 1억뷰 달성, <a href="/wiki/%EA%B1%B8_%EA%B7%B8%EB%A3%B9" title="걸 그룹">걸 그룹</a> 최초 뮤직비디오 조회수 4억뷰 달성, <a href="/wiki/K-POP" class="mw-redirect" title="K-POP">K-POP</a> 가수 최초 한-일 누적앨범 1000만 장 판매, 역대 여자 가수 <a href="/wiki/GMTV_(%ED%85%94%EB%A0%88%EB%B9%84%EC%A0%84_%EC%B1%84%EB%84%90)" title="GMTV (텔레비전 채널)">음악방송</a> 수상 횟수 1위를 달성하며 엄청난 모습을 보여주고 있다. <a href="/wiki/%EB%B8%94%EB%9E%99%ED%95%91%ED%81%AC" class="mw-redirect" title="블랙핑크">블랙핑크</a> 역시 걸 그룹 최초 뮤직비디오 조회수 10억뷰, 역대 뮤직비디오 누적 조회수 1위 등 여러 기록들을 세우며 '월드스타' 면모를 보여주고 있다.
</p>
<div class="mw-heading mw-heading3"><h3 id="2010년대"><span id="2010.EB.85.84.EB.8C.80"></span>2010년대</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=3" title="부분 편집: 2010년대"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<figure class="mw-default-size mw-halign-left" typeof="mw:File/Thumb"><a href="https://commons.wikimedia.org/wiki/File:Girls%27_Generation_(SNSD)_intro_at_SMTown_Live_NY.jpg" class="mw-file-description"><img src="//upload.wikimedia.org/wikipedia/commons/thumb/3/32/Girls%27_Generation_%28SNSD%29_intro_at_SMTown_Live_NY.jpg/220px-Girls%27_Generation_%28SNSD%29_intro_at_SMTown_Live_NY.jpg" decoding="async" width="220" height="165" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/3/32/Girls%27_Generation_%28SNSD%29_intro_at_SMTown_Live_NY.jpg/330px-Girls%27_Generation_%28SNSD%29_intro_at_SMTown_Live_NY.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/3/32/Girls%27_Generation_%28SNSD%29_intro_at_SMTown_Live_NY.jpg/440px-Girls%27_Generation_%28SNSD%29_intro_at_SMTown_Live_NY.jpg 2x" data-file-width="1213" data-file-height="910"></a><figcaption>2011년 SM타운 라이브 <a href="/wiki/%EB%89%B4%EC%9A%95" title="뉴욕">뉴욕</a>에서 소녀시대.</figcaption></figure>
<p>2012년 2월 29일 발매한 <a href="/wiki/%EB%B9%85%EB%B1%85_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="빅뱅 (음악 그룹)">BIGBANG</a>의 ALIVE 앨범이 한국어 앨범으로 대한민국 최초로 빌보드 메인 차트 진입에 성공했다. 이 노래는 빌보드 200 150위에 랭크되었다.<sup id="cite_ref-7" class="reference"><a href="#cite_note-7"><span class="cite-bracket">[</span>7<span class="cite-bracket">]</span></a></sup> BIGBANG이 2016년 12월 13일 발매한 MADE 앨범으로 ALIVE 앨범 이후 4년여 만에 빌보드 메인 차트, 빌보드 200에서 172위에 랭크되었다.<sup id="cite_ref-8" class="reference"><a href="#cite_note-8"><span class="cite-bracket">[</span>8<span class="cite-bracket">]</span></a></sup>
</p><p><a href="/wiki/%EC%8B%B8%EC%9D%B4" title="싸이">싸이</a>(PSY)의 강남스타일 곡, <a href="/wiki/BLACKPINK" title="BLACKPINK">BLACKPINK</a>, <a href="/wiki/%ED%8A%B8%EC%99%80%EC%9D%B4%EC%8A%A4" title="트와이스">트와이스</a>, <a href="/wiki/%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8" title="방탄소년단">방탄소년단</a>의 영향으로 <a href="/wiki/%EB%AF%B8%EA%B5%AD" title="미국">미국</a> 내에서도 <a href="/wiki/K-POP" class="mw-redirect" title="K-POP">K-POP</a>에 대한 인식이 많아지기는 했지만 2010년대 초반에 작게나마 유의미한 움직임이 포착된 적은 있었다. <a href="/wiki/%ED%98%84%EC%95%84" title="현아">현아</a>의 버블팝이 리그오브레전드 플레이 동영상 배경음악으로 흥하면서 롤 <a href="/wiki/%EA%B2%8C%EC%9D%B4%EB%A8%B8" title="게이머">게이머</a>들 사이에서 컬트적인 인기를 누리는 등, 현지에서 유의적인 반응을 얻는 모습이 종종 눈에 띄기는 하였다. <a href="/wiki/G-DRAGON" class="mw-redirect" title="G-DRAGON">G-DRAGON</a>의 패션이나 TAEYANG의 웨딩드레스 곡이 미국 현지 대학생들 사이에서 소소하게 히트를 치며 커버 열풍이 불기도 했다.
</p><p>2010년대 후반에는 K-pop 음악의 반주로 <a href="/w/index.php?title=%ED%8A%B8%EB%A1%9C%ED%94%BC%EC%BB%AC_%ED%95%98%EC%9A%B0%EC%8A%A4&amp;action=edit&amp;redlink=1" class="new" title="트로피컬 하우스 (없는 문서)">트로피컬 하우스</a> 장르로 분류되는 반주를 사용하는 시도를 하였다. <a href="/wiki/%EB%B8%8C%EB%A0%88%EC%9D%B4%EB%B8%8C%EA%B1%B8%EC%8A%A4" title="브레이브걸스">브레이브걸스</a>의 〈Rollin'〉(<a href="/wiki/2017%EB%85%84" title="2017년">2017년</a>), <a href="/wiki/%EC%88%80_(%EC%9D%8C%EC%95%85%EA%B0%80)" title="숀 (음악가)">숀</a>의 〈Way back home〉(<a href="/wiki/2018%EB%85%84" title="2018년">2018년</a>)이 그 예시이다.
</p>
<div class="mw-heading mw-heading3"><h3 id="2020년대"><span id="2020.EB.85.84.EB.8C.80"></span>2020년대</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=4" title="부분 편집: 2020년대"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<div class="mw-heading mw-heading4"><h4 id="음악_짜임새의_큰_변화"><span id=".EC.9D.8C.EC.95.85_.EC.A7.9C.EC.9E.84.EC.83.88.EC.9D.98_.ED.81.B0_.EB.B3.80.ED.99.94"></span>음악 짜임새의 큰 변화</h4><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=5" title="부분 편집: 음악 짜임새의 큰 변화"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<style data-mw-deduplicate="TemplateStyles:r38501082">.mw-parser-output .side-box{margin:4px 0;box-sizing:border-box;border:1px solid #aaa;font-size:88%;line-height:1.25em;background-color:var(--background-color-interactive-subtle,#f8f9fa);display:flow-root}.mw-parser-output .side-box-abovebelow,.mw-parser-output .side-box-text{padding:0.25em 0.9em}.mw-parser-output .side-box-image{padding:2px 0 2px 0.9em;text-align:center}.mw-parser-output .side-box-imageright{padding:2px 0.9em 2px 0;text-align:center}@media(min-width:500px){.mw-parser-output .side-box-flex{display:flex;align-items:center}.mw-parser-output .side-box-text{flex:1;min-width:0}}@media(min-width:720px){.mw-parser-output .side-box{width:238px}.mw-parser-output .side-box-right{clear:right;float:right;margin-left:1em}.mw-parser-output .side-box-left{margin-right:1em}}</style><div class="side-box metadata side-box-right plainlinks"><style data-mw-deduplicate="TemplateStyles:r36480595">.mw-parser-output .plainlist ol,.mw-parser-output .plainlist ul{line-height:inherit;list-style:none;margin:0;padding:0}.mw-parser-output .plainlist ol li,.mw-parser-output .plainlist ul li{margin-bottom:0}</style>
<div class="side-box-flex">
<div class="side-box-image"><span typeof="mw:File"><span><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/2/24/Wikinews-logo.svg/40px-Wikinews-logo.svg.png" decoding="async" width="40" height="22" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/2/24/Wikinews-logo.svg/60px-Wikinews-logo.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/2/24/Wikinews-logo.svg/80px-Wikinews-logo.svg.png 2x" data-file-width="759" data-file-height="415"></span></span></div>
<div class="side-box-text plainlist"><b><a href="/wiki/%EC%9C%84%ED%82%A4%EB%89%B4%EC%8A%A4" title="위키뉴스">위키뉴스</a></b>에 이 문서와<br>관련된 기사가 있습니다.
<div class="plainlinks" style="padding-left:1em;"><b><a href="https://ko.wikinews.org/wiki/BTS_%E3%80%88Dynamite%E3%80%89%EC%9D%98_%EB%B9%8C%EB%B3%B4%EB%93%9C_1%EC%9C%84,_%EC%A2%8B%EC%A7%80%EB%A7%8C%EC%9D%80_%EC%95%8A%EC%9D%80_%EC%9D%B4%EC%9C%A0" class="extiw" title="n:BTS 〈Dynamite〉의 빌보드 1위, 좋지만은 않은 이유">BTS 〈Dynamite〉의 빌보드 1위, 좋지만은 않은 이유</a></b></div></div></div>
</div>
<p>2020년대에 들어서 K-pop이 해외 차트에 지속적으로 등장하자, <a href="/wiki/(%EC%97%AC%EC%9E%90)%EC%95%84%EC%9D%B4%EB%93%A4" title="(여자)아이들">(여자)아이들</a>의 〈MY BAG〉(<a href="/wiki/2022%EB%85%84" title="2022년">2022년</a> <a href="/wiki/3%EC%9B%94" title="3월">3월</a>), <a href="/wiki/Aespa" title="Aespa">aespa</a>의 〈Dream Comes True〉(<a href="/wiki/2021%EB%85%84" title="2021년">2021년</a>)와 같이 <a href="/wiki/%EC%98%81%EC%96%B4" title="영어">영어</a>를 가사의 대부분으로 사용하거나, K-pop 장르라고 불리지만 한국어가 한 문장도 존재하지 않는 곡인 <a href="/wiki/%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8" title="방탄소년단">방탄소년단</a>의 〈Dynamite〉(<a href="/wiki/2020%EB%85%84" title="2020년">2020년</a> <a href="/wiki/9%EC%9B%94" title="9월">9월</a>), 〈Butter〉(<a href="/wiki/2021%EB%85%84" title="2021년">2021년</a> <a href="/wiki/5%EC%9B%94" title="5월">5월</a>), 〈Permission to Dance〉 등 새로운 시도가 등장하기 시작한다.<sup id="cite_ref-9" class="reference"><a href="#cite_note-9"><span class="cite-bracket">[</span>내용주 1<span class="cite-bracket">]</span></a></sup> 이는 K-pop이라는 장르에 <a href="/wiki/%ED%95%9C%EA%B5%AD%EC%96%B4" title="한국어">한국어</a> 없이도, 오직 <a href="/wiki/%EC%98%81%EC%96%B4" title="영어">영어</a>만으로도 K-pop 장르에 속하는 노래가 만들어 질 수 있다는 것을 보여주었다. 하지만 이와 반대로, 2010년대까지 K-pop 장르의 음악 중 영어 문장이 곡 가사의 반절 이상 들어간 노래들이 많았다는 지적을 받은 영향으로, 이것을 받아들여 가사 내 외국어를 최대한 사용하지 않으려는 움직임도 보이기 시작하였다. 대표적인 노래는 <a href="/wiki/%EB%A7%88%ED%81%AC%ED%88%BD" title="마크툽">마크툽</a>의 〈찰나가 영원이 될 때〉(<a href="/wiki/2021%EB%85%84" title="2021년">2021년</a> <a href="/wiki/9%EC%9B%94" title="9월">9월</a>), <a href="/wiki/%EC%A3%BC%EC%8B%9C%ED%81%AC" title="주시크">주시크</a>의 〈너를 생각해〉(<a href="/wiki/2021%EB%85%84" title="2021년">2021년</a> <a href="/wiki/9%EC%9B%94" title="9월">9월</a>) 등이 있다. 더불어 "<a href="/wiki/%EC%82%AC%EB%9E%91" title="사랑">사랑</a>"이라는 주제의 곡이 너무 많다는 지적도 있어, 새로운 주제의 노래도 서서히 나타나고 있다. 예를 들어, <a href="/wiki/%EC%9D%B4%EB%AC%B4%EC%A7%84" title="이무진">이무진</a>의 〈<a href="/wiki/%EC%8B%A0%ED%98%B8%EB%93%B1_(%EB%85%B8%EB%9E%98)" title="신호등 (노래)">신호등</a>〉(<a href="/wiki/2021%EB%85%84" title="2021년">2021년</a> <a href="/wiki/5%EC%9B%94" title="5월">5월</a>)의 가사는 <a href="/wiki/%EC%9A%B4%EC%A0%84%EB%A9%B4%ED%97%88" class="mw-redirect" title="운전면허">운전면허</a>를 갓 딴 초보운전자를 사회 초년생인 사람에 빗대어 작사되었고, 어찌할 바 모르는 젊은 사람들, 20대 청춘을 주제로 한 곡이다.<sup id="cite_ref-10" class="reference"><a href="#cite_note-10"><span class="cite-bracket">[</span>9<span class="cite-bracket">]</span></a></sup>
</p><p>그에 더불어, 한국의 전통 음악인 <a href="/wiki/%ED%95%9C%EA%B5%AD%EC%9D%8C%EC%95%85" title="한국음악">국악</a> 혹은 <a href="/wiki/%ED%8C%90%EC%86%8C%EB%A6%AC" title="판소리">판소리</a>, 더 넓은 범위에서는 <a href="/wiki/%ED%8A%B8%EB%A1%9C%ED%8A%B8" title="트로트">트로트</a> 장르 반주와 <a href="/wiki/%EC%A0%84%EC%9E%90%EC%9D%8C%EC%95%85" class="mw-redirect" title="전자음악">전자음악</a>, <a href="/wiki/%EB%A1%9D_%EC%9D%8C%EC%95%85" title="록 음악">록 음악</a> 장르가 혼합되어 '<a href="/w/index.php?title=%EC%A1%B0%EC%84%A0%ED%8C%9D&amp;action=edit&amp;redlink=1" class="new" title="조선팝 (없는 문서)">조선팝</a>'이라는 새 장르가 만들어지기도 했다. 가장 대표적인 음악은 <a href="/wiki/%EC%9D%B4%EB%82%A0%EC%B9%98_(%EB%B0%B4%EB%93%9C)" title="이날치 (밴드)">이날치</a>의 〈<a href="/wiki/%EB%B2%94_%EB%82%B4%EB%A0%A4%EC%98%A8%EB%8B%A4" title="범 내려온다">범 내려온다</a>〉(<a href="/wiki/2020%EB%85%84" title="2020년">2020년</a> <a href="/wiki/5%EC%9B%94" title="5월">5월</a>)가 있다. 〈<a href="/wiki/%EB%B2%94_%EB%82%B4%EB%A0%A4%EC%98%A8%EB%8B%A4" title="범 내려온다">범 내려온다</a>〉는 케이팝 장르에서 거의 쓰이지 않았던 장르인 <a href="/wiki/%EB%89%B4_%EC%9B%A8%EC%9D%B4%EB%B8%8C_(%EC%9D%8C%EC%95%85)" title="뉴 웨이브 (음악)">뉴 웨이브</a>와도 조합되어 있다.
</p><p><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36482206"><span title="이 정보가 어느 시점에 해당하는지 알 수 없습니다." class="fix">최근,</span><sup class="noprint" title="2025년 1월에 수정이 요청되었습니다.">[<a href="/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%AA%A8%ED%98%B8%ED%95%9C_%ED%91%9C%ED%98%84%EC%9D%84_%EC%82%BC%EA%B0%80_%EC%A3%BC%EC%84%B8%EC%9A%94" class="mw-redirect" title="위키백과:모호한 표현을 삼가 주세요">언제</a>?]</sup> 남자아이돌 그룹 <a href="/wiki/%EC%84%B8%EB%B8%90%ED%8B%B4" class="mw-disambig" title="세븐틴">세븐틴</a>은 초동 500만장을 돌파하며 새로운 역사를 개척해나가고 있다.
</p>
<div class="mw-heading mw-heading4"><h4 id="위기"><span id=".EC.9C.84.EA.B8.B0"></span>위기</h4><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=6" title="부분 편집: 위기"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>2020년 초중반,  <a href="/wiki/FIFTY_FIFTY" title="FIFTY FIFTY">FIFTY FIFTY</a>의 계약 분쟁<sup id="cite_ref-11" class="reference"><a href="#cite_note-11"><span class="cite-bracket">[</span>10<span class="cite-bracket">]</span></a></sup> 및 <a href="/wiki/%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8" title="방탄소년단">방탄소년단</a>의 <a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EB%B3%91%EC%97%AD_%EC%A0%9C%EB%8F%84" title="대한민국의 병역 제도">군 입대</a>, <a href="/wiki/%EB%AF%BC%ED%9D%AC%EC%A7%84" title="민희진">민희진</a>의 <a href="/wiki/%ED%95%98%EC%9D%B4%EB%B8%8C_(%EA%B8%B0%EC%97%85)" class="mw-redirect" title="하이브 (기업)">하이브</a>-민희진 간 경영상의 분쟁,<sup id="cite_ref-12" class="reference"><a href="#cite_note-12"><span class="cite-bracket">[</span>11<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-13" class="reference"><a href="#cite_note-13"><span class="cite-bracket">[</span>12<span class="cite-bracket">]</span></a></sup> <a href="/wiki/%EC%8A%88%EA%B0%80_(%EB%9E%98%ED%8D%BC)" title="슈가 (래퍼)">슈가</a>의 음주운전 사고<sup id="cite_ref-14" class="reference"><a href="#cite_note-14"><span class="cite-bracket">[</span>13<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-15" class="reference"><a href="#cite_note-15"><span class="cite-bracket">[</span>14<span class="cite-bracket">]</span></a></sup> 및 하이브의 안일한 대응 때문에 큰 타격을 받았다. 또한 <a href="/wiki/%ED%8C%9D_%EC%9D%8C%EC%95%85" title="팝 음악">미국 팝</a>의 부활과 <a href="/wiki/%EB%9D%BC%ED%8B%B4_%ED%8C%9D" title="라틴 팝">라틴 팝</a>이나 <a href="/wiki/J-pop" title="J-pop">J-pop</a> 등의 경쟁 상대의 등장으로 위기를 맞고 있다. 실제로 많은 케이팝 음악가들이 복귀했음에도, 유튜브의 조회수가 1일에 300만회를 넘기는 것이 힘들어졌다. 많은 논평가들은 케이팝의 아이돌 체제 혹은 음반 짧은 수명 등을 극복하지 못하면 몰락하게 될 것이라는 우려섞인 목소리를 내기도 하였다.<sup id="cite_ref-16" class="reference"><a href="#cite_note-16"><span class="cite-bracket">[</span>15<span class="cite-bracket">]</span></a></sup> 일시적인 현상인지 장기적인 침체인지는 시간을 두고 해석해 봐야 할듯하다.
</p>
<div class="mw-heading mw-heading4"><h4 id="2020년대_중반"><span id="2020.EB.85.84.EB.8C.80_.EC.A4.91.EB.B0.98"></span>2020년대 중반</h4><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=7" title="부분 편집: 2020년대 중반"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>2024년 10월, <a href="/wiki/%EB%A1%9C%EC%A0%9C_(%EA%B0%80%EC%88%98)" title="로제 (가수)">로제</a>의 APT(아파트)가 큰 인기를 끌고, 여러 유저들이 다양한 패러디 영상을 창작해냈다.<sup id="cite_ref-17" class="reference"><a href="#cite_note-17"><span class="cite-bracket">[</span>16<span class="cite-bracket">]</span></a></sup>
</p><p>한편 K-POP이 일종의 <a href="/wiki/%EB%AF%BC%EC%A4%91%EA%B0%80%EC%9A%94" title="민중가요">민중가요</a>로서 기능하는 모습도 포착되었다. 앞서 로제의 APT 역시 단순한 유머성 패러디 뿐만 아니라, 다양한 시위에서 해당 음악을 기반으로 한 패러디 음악이 등장하였다.<sup id="cite_ref-18" class="reference"><a href="#cite_note-18"><span class="cite-bracket">[</span>17<span class="cite-bracket">]</span></a></sup> <a href="/wiki/%EC%9C%A4%EC%84%9D%EC%97%B4_%EB%8C%80%ED%86%B5%EB%A0%B9_%ED%83%84%ED%95%B5_%EC%9A%B4%EB%8F%99" title="윤석열 대통령 탄핵 운동">윤석열 대통령 탄핵 운동</a>에서도 다양한 K-POP 팬덤의 응원봉이 등장하였고,<sup id="cite_ref-19" class="reference"><a href="#cite_note-19"><span class="cite-bracket">[</span>18<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-20" class="reference"><a href="#cite_note-20"><span class="cite-bracket">[</span>19<span class="cite-bracket">]</span></a></sup> 2000년대부터 만들어져왔던 K-POP의 음악들이 시위 현장에서 재생되었다. 대표적인 곡은 <a href="/wiki/%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80" title="소녀시대">소녀시대</a>의 《다시 만난 세계》가 있다.<sup id="cite_ref-21" class="reference"><a href="#cite_note-21"><span class="cite-bracket">[</span>20<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-22" class="reference"><a href="#cite_note-22"><span class="cite-bracket">[</span>21<span class="cite-bracket">]</span></a></sup>
</p>
<div class="mw-heading mw-heading2"><h2 id="특징"><span id=".ED.8A.B9.EC.A7.95"></span>특징</h2><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=8" title="부분 편집: 특징"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<div class="mw-heading mw-heading3"><h3 id="해외_언론의_인식_중_일부"><span id=".ED.95.B4.EC.99.B8_.EC.96.B8.EB.A1.A0.EC.9D.98_.EC.9D.B8.EC.8B.9D_.EC.A4.91_.EC.9D.BC.EB.B6.80"></span>해외 언론의 인식 중 일부</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=9" title="부분 편집: 해외 언론의 인식 중 일부"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>영국방송 <a href="/wiki/BBC" class="mw-redirect" title="BBC">BBC</a>는 <a href="/wiki/%EC%8A%88%ED%8D%BC%EC%A3%BC%EB%8B%88%EC%96%B4" title="슈퍼주니어">슈퍼주니어</a>와 <a href="/wiki/%EC%9B%90%EB%8D%94%EA%B1%B8%EC%8A%A4" title="원더걸스">원더걸스</a>를 “잘 프로듀싱 된 멋지고 예쁜 보이, <a href="/wiki/%EA%B1%B8_%EA%B7%B8%EB%A3%B9" title="걸 그룹">걸 그룹</a>으로써 눈을 사로잡는 안무와 캐치한 멜로디를 구사한다.”<sup id="cite_ref-23" class="reference"><a href="#cite_note-23"><span class="cite-bracket">[</span>22<span class="cite-bracket">]</span></a></sup>
라고 표현했다. 케이팝은 젊고 매력적인 ‘예쁜’ 보이, 걸 그룹을 상징한다.<sup id="cite_ref-24" class="reference"><a href="#cite_note-24"><span class="cite-bracket">[</span>23<span class="cite-bracket">]</span></a></sup>
</p>
<div class="mw-heading mw-heading3"><h3 id="국제화된_음악"><span id=".EA.B5.AD.EC.A0.9C.ED.99.94.EB.90.9C_.EC.9D.8C.EC.95.85"></span>국제화된 음악</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=10" title="부분 편집: 국제화된 음악"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>총체적으로 <a href="/wiki/%EB%8C%80%EC%A4%91%EC%9D%8C%EC%95%85" title="대중음악">팝</a>, <a href="/wiki/%EB%A1%9D_%EC%9D%8C%EC%95%85" title="록 음악">록</a>, <a href="/wiki/%ED%9E%99%ED%95%A9" title="힙합">힙합</a>, <a href="/wiki/%EC%95%8C%EC%95%A4%EB%B9%84" class="mw-redirect" title="알앤비">알앤비</a> 그리고 <a href="/wiki/%EC%A0%84%EC%9E%90%EC%9D%8C%EC%95%85" class="mw-redirect" title="전자음악">전자음악</a>의 결합이라고 할 수 있으며, 서구권의 사운드에 아시아인의 퍼포먼스를 결합시키는 특징을 지니기 때문에 보다 글로벌한 음악으로 불린다. 대한민국 가수들의 소위 칼군무와 어려운 안무 동작들은 K-POP을 세계적으로 이끄는데 큰 공헌을 했다. 현재 K-POP은 비단 아시아뿐만 아니라 세계 음악시장에서도 큰 영향을 끼치고 있다.<sup id="cite_ref-25" class="reference"><a href="#cite_note-25"><span class="cite-bracket">[</span>24<span class="cite-bracket">]</span></a></sup>
한국 가수들에게 곡을 주는 외국 프로듀서들도 많다. <a href="/wiki/%EC%9C%8C%EC%95%84%EC%9D%B4%EC%97%A0" class="mw-redirect" title="윌아이엠">윌아이엠</a>, 션 가렛 에서부터 미국에서 자란 <a href="/wiki/%ED%85%8C%EB%94%94_(%EA%B0%80%EC%88%98)" class="mw-redirect" title="테디 (가수)">테디</a> 등이 이에 속한다. K-POP 가수들과 콜라보레이션을 한 힙합가수도 많다. <a href="/wiki/%EC%97%90%EC%9D%B4%EC%BD%98_(%EA%B0%80%EC%88%98)" title="에이콘 (가수)">에이콘</a>, <a href="/wiki/%EC%B9%B8%EC%98%88_%EC%9B%A8%EC%8A%A4%ED%8A%B8" class="mw-redirect" title="칸예 웨스트">칸예 웨스트</a>, <a href="/wiki/%EB%A3%A8%EB%8B%A4%ED%81%AC%EB%A6%AC%EC%8A%A4" title="루다크리스">루다크리스</a>, <a href="/wiki/%EC%8A%A4%EB%88%95%EB%8F%85" class="mw-redirect" title="스눕독">스눕독</a>, 그리고  <a href="/wiki/%EC%9C%8C%EC%95%84%EC%9D%B4%EC%97%A0" class="mw-redirect" title="윌아이엠">윌아이엠</a> 등이 이에 속한다. 노르웨이와 스웨덴에서 곡을 받기도 한다. ‘<a href="/wiki/%EC%9B%94%EC%8A%A4%ED%8A%B8%EB%A6%AC%ED%8A%B8%EC%A0%80%EB%84%90" class="mw-redirect" title="월스트리트저널">월스트리트저널</a>’에서 한 제작자는 “한국은 미국의 비트에 스칸디나비아 반도의 곡 스타일을 원한다“고 말했다.
</p><p>많은 경우 한국의 <a href="/wiki/%EC%82%AC%EB%8C%80%EC%A3%BC%EC%9D%98" title="사대주의">사대주의</a> 사상에 의해, 다른 장르에서는 볼 수 없는 특징으로 영어 가사가 큰 비중으로 들어가 있다.<sup id="cite_ref-26" class="reference"><a href="#cite_note-26"><span class="cite-bracket">[</span>25<span class="cite-bracket">]</span></a></sup> 때로는 가사 전체가 영어 가사로 되어 있을 정도로 그 비중은 높다. 이것은 미국 시장에 큰 어필을 하는 데 성공하였다. 또한 일본 시장을 향해 일본어 앨범을 먼저 발매하는 형태를 보이며, 종종 'Korean version' 음반을 재발매하는 방식으로 운영되어<sup id="cite_ref-27" class="reference"><a href="#cite_note-27"><span class="cite-bracket">[</span>26<span class="cite-bracket">]</span></a></sup> '국제화된 음악' 이라고 불린다.
</p>
<div class="mw-heading mw-heading3"><h3 id="디지털_앨범"><span id=".EB.94.94.EC.A7.80.ED.84.B8_.EC.95.A8.EB.B2.94"></span>디지털 앨범</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=11" title="부분 편집: 디지털 앨범"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>1세대 아이돌들의 음원의 경우 음반사에서만 CD를 구할 수 있었다면, 2세대라 불리는 아이돌들이 등장하면서  K-POP의 규모가 커짐에 따라 앨범 제작 및 공급처는 급격하게 증가했으며, 음반사뿐 아니라 서점 혹은 마트에서도 구할 수 있게 접근성이 증가하였다. 앨범의 구성 역시 변화되었는데 과거는 포장된 케이스에 CD만 들어있었다면, 요즘의 경우 CD를 포함한 포토북, 자켓앨범, 그리고 여러 종의 포토카드 등 CD뿐 아니라 팬들의 시각적 흥미와 구매력을 이끌어 내는 다양한 구성 품들이 들어있다. 현재는 CD를 구매하는 것뿐만 아니라 스트리밍 사이트에서 음원만 구매하기도 하며 실존하는 앨범을 내는 것이 아닌 디지털 싱글 즉, 앨범 없이 음원만 출시하기도 한다.
</p>
<div class="mw-heading mw-heading3"><h3 id="뮤직비디오"><span id=".EB.AE.A4.EC.A7.81.EB.B9.84.EB.94.94.EC.98.A4"></span>뮤직비디오</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=12" title="부분 편집: 뮤직비디오"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>K-POP의 성공요소 중 하나는 뮤직비디오라 할 수 있다. 잘 만들어진 뮤직비디오 또한 팬들의 시선을 사로잡는 비장의 무기 중 하나이다. 곡과 맞는 컨셉, 소품의 활용, 개개인의 특징을 살린 의상들과 다양한 특수효과의 조화로 영상미를 한껏 끌어올려 보는 팬들로 하여금 다시보기를 하게 만든다. 외국 가수의 뮤직비디오와 비교해 봐도 K-POP 가수들의 뮤직비디오는 그들만의 특성이 있으며, 외국 팬들 또한 거부감 없이 받아들인다. 그 이유는 대한민국 가수들의 뮤직비디오는 춤과 노래만 담아내는 것이 아닌 하나의 영화로 짜임새를 갖추어 만들어지기 때문이라고 할 수 있다. 실제로 다양한 가수들이 뮤직비디오의 스토리를 이어지게끔 만드는 경우도 있으며 팬들의 입장에선 다음 뮤직비디오를 기대하며 기다리게끔 한다.
</p>
<div class="mw-heading mw-heading3"><h3 id="군무(그룹댄스)"><span id=".EA.B5.B0.EB.AC.B4.28.EA.B7.B8.EB.A3.B9.EB.8C.84.EC.8A.A4.29"></span>군무(그룹댄스)</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=13" title="부분 편집: 군무(그룹댄스)"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>군무(群舞, group dance)는 케이팝에서 중요한 요소 중 하나다. 군무는 여러 명이 무대를 꾸밀 때 가수들이 자리를 바꾸는데에서 시작했다. <a href="/wiki/%EC%84%9C%ED%83%9C%EC%A7%80%EC%99%80_%EC%95%84%EC%9D%B4%EB%93%A4" title="서태지와 아이들">서태지와 아이들</a>의 무대부터 자리바꿈이 공공연해졌고, 이는 훗날 케이팝 안무에까지 영향을 끼쳤다.
</p><p>케이팝 안무는 포인트 안무라는 특징을 지니기도 하는데, 주로 훅 부분에서 쉬운 안무를 각인시키는 행위를 말한다. 이 포인트 안무는 주로 가사를 따라가기도 한다. 포인트 안무는 곡의 히트 유무를 결정하기도 한다.
</p><p>안무의 호흡을 잘 맞추고, 마치 한 사람인 것처럼 안무를 선보이는 아이돌을 군무돌, 또는 칼군무돌이라고 칭한다.
</p><p>안무에 맞춰 곡이 쓰이기도 하며, 팬들이 쉽게 따라할 수 있음의 유무에 중점을 둔다. LA의 한 안무가인 Ellen Kim은 “케이팝은 팬들과 함께 하기를 원한다. 그래서 안무가 쉽게 나오는 경향이 있다”라고 이야기한다.
</p><p>또 그룹댄스의 영향은 이후 <a href="/wiki/%ED%8B%B1%ED%86%A1" title="틱톡">틱톡</a> 등의 시장에서 큰 인기를 끌게 되었고, 하나의 콘텐츠로서 기능하게 되었다. 이러한 콘텐츠는 "챌린지" 라는 이름으로 불리며, 수많은 사람들이 자신의 방식으로 K-POP의 춤을 추는 것을 영상으로 촬영하여 공유하는 문화이다.<sup id="cite_ref-28" class="reference"><a href="#cite_note-28"><span class="cite-bracket">[</span>27<span class="cite-bracket">]</span></a></sup>
</p>
<div class="mw-heading mw-heading3"><h3 id="패션"><span id=".ED.8C.A8.EC.85.98"></span>패션</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=14" title="부분 편집: 패션"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>케이팝은 패션에도 큰 영향을 준다. 케이팝 가수의 스타일은 특히 범아시아 지역에서 젊은이들의 눈을 사로잡고 있다. 따라서 의류업계는 가수로 인해 유명해진 옷을 마케팅한다. <a href="/wiki/%EB%B9%85%EB%B1%85_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="빅뱅 (음악 그룹)">빅뱅</a>의 <a href="/wiki/%EC%A7%80%EB%93%9C%EB%9E%98%EA%B3%A4" title="지드래곤">지드래곤</a>은 해외의 많은 패션업계의 러브콜을 받고 있으며, <a href="/wiki/%ED%88%AC%EC%95%A0%EB%8B%88%EC%9B%90" class="mw-redirect" title="투애니원">투애니원</a>의 <a href="/wiki/CL_(%EA%B0%80%EC%88%98)" title="CL (가수)">씨엘</a> 또한 디자이너 <a href="/wiki/%EC%A0%9C%EB%A0%88%EB%AF%B8_%EC%8A%A4%EC%BA%87" class="mw-redirect" title="제레미 스캇">제레미 스캇</a>과 함께 작업하기도 했다. 이외에도 <a href="/wiki/%EB%B8%94%EB%9E%99%ED%95%91%ED%81%AC" class="mw-redirect" title="블랙핑크">블랙핑크</a>의 <a href="/wiki/%EC%A0%9C%EB%8B%88_(1996%EB%85%84)" title="제니 (1996년)">제니</a>는 <a href="/wiki/%EC%83%A4%EB%84%AC" title="샤넬">샤넬</a>의 대표적인 엠버서더로 패션쇼에도 서는 등의 활동도 진행 중이다. <a href="/wiki/%EB%B8%94%EB%9E%99%ED%95%91%ED%81%AC" class="mw-redirect" title="블랙핑크">블랙핑크</a>의 <a href="/wiki/%EC%A7%80%EC%88%98_(%EA%B0%80%EC%88%98)" title="지수 (가수)">지수</a>는 버버리와 디올의 엠버서더로, 얼마 전 디올로부터 <a href="/wiki/YG_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="YG 엔터테인먼트">YG</a>에서 쫓겨나면 디올로 오라는 장난스러운 권유를 받기도 하였다. 이외에도 다양한 케이팝 스타들이 패션계에 큰 영향을 미치고 있다.
</p><p>2012년 1월에는 한국 가수들이 일본에서 패션쇼를 열었고, 무려 3만 3천명의 인원이 패션쇼팝 가수해다행 팝가수패션이 날씨에 맞지 않아 문제가 되기도 했다 (레깅스 패션). <a href="/wiki/%EC%A1%B0%EC%84%A0%EB%AF%BC%EC%A3%BC%EC%A3%BC%EC%9D%98%EC%9D%B8%EB%AF%BC%EA%B3%B5%ED%99%94%EA%B5%AD" title="조선민주주의인민공화국">조선민주주의인민공화국</a> 또한 폐쇄적인 규제에도 불구하고 <a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD" title="대한민국">대한민국</a>의 패션이 암암리에 유행하고 있다. 하이힐, 탑, 패션 악세사리 등 조선민주주의인민공화국 내에서 자유롭지 않은 패션을 케이팝으로 인해 유행시키고 있다.
</p><p>또한 인기있는 케이팝 스타의 패션은 그들을 좋아하는 팬덤사이에서 똑같은 물건의 정보를 알아내어 따라 구매하는 '손민수' 문화를 만들어내기도 하였다. 가수가 소유하고 있는 패션 아이템을 많은 팬들이 따라 사다보니 특히 유행이 되고 화제가 된 제품들은 순식간에 품절이 되어버리기도 한다. 또한 구매하고 나면 본인의 SNS에 해시태그를 붙여 본인의 구매를 인증한다.
</p>
<div class="mw-heading mw-heading3"><h3 id="팬덤"><span id=".ED.8C.AC.EB.8D.A4"></span>팬덤</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=15" title="부분 편집: 팬덤"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>K-POP 아이돌의 경우 대부분의 아이돌들이 공식 팬클럽을 런칭하며 고정적 팬층을 확보하며 활동을 한다. 대부분의 아이돌은 데뷔 후 짧게는 1년 미만, 길게는 2년 이상을 팬클럽 없이 활동을 하다가 일정 수준이상의 고정적 팬 층이 확보가 되면 팬클럽을 창단해 다음 활동을 이어나간다. 대부분 1년 주기의 기수제이며 가입비도 1만원 ~ 3만원으로 다양한 편이다. 가입한 팬들은 소속감을 가지며 팬덤 활동을 하고, 팬클럽의 가입한 팬들과 안한 팬들에게 차등 혜택을 주며 가입한 팬들로 하여금 “가입하길 잘했다.”라는 생각을 하게한다. 이밖에도 콘서트때 화환을 보내거나, 예능 혹은 드라마 촬영 시 밥 차 혹은 간식차를 보내며 아티스트를 위한 선물을 보내며 가수를 위한 서포트 활동을 하기도하며, 역으로 아이돌들이 공개방송이나 행사 때  자신들의 팬덤을 위해 간식차를 준비하는 등 역 조공 이벤트를 하기도 한다. 이밖에도 공식 색, 공식 응원구호, 공식 응원봉 등 가수와 팬들만의 정체성을 세기며 단합하기도 한다.
</p>
<div class="mw-heading mw-heading3"><h3 id="유행에_따르는_장르_변화"><span id=".EC.9C.A0.ED.96.89.EC.97.90_.EB.94.B0.EB.A5.B4.EB.8A.94_.EC.9E.A5.EB.A5.B4_.EB.B3.80.ED.99.94"></span>유행에 따르는 장르 변화</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=16" title="부분 편집: 유행에 따르는 장르 변화"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>K-POP 산업의 중심은 <a href="/wiki/%EC%95%84%EC%9D%B4%EB%8F%8C" title="아이돌">아이돌</a>로 이루어져 있지만, 음악은 유행에 따라 다양한 장르를 팝 음악처럼 재해석하는 시도를 하고 있다.<sup id="cite_ref-29" class="reference"><a href="#cite_note-29"><span class="cite-bracket">[</span>28<span class="cite-bracket">]</span></a></sup> 2015년~2016년부터 <a href="/w/index.php?title=%ED%8A%B8%EB%A1%9C%ED%94%BC%EC%BB%AC_%ED%95%98%EC%9A%B0%EC%8A%A4&amp;action=edit&amp;redlink=1" class="new" title="트로피컬 하우스 (없는 문서)">트로피컬 하우스</a> 장르의 음악이 인기를 끌게 되자, 실제로 17~18년경부터 K-POP 음반 시장에 트로피컬 하우스 장르의 음악들이 다수 발매되었다. <a href="/wiki/%EB%B8%8C%EB%A0%88%EC%9D%B4%EB%B8%8C%EA%B1%B8%EC%8A%A4" title="브레이브걸스">브레이브걸스</a>의 《<a href="/wiki/Rollin%27_(%EB%B8%8C%EB%B8%8C%EA%B1%B8%EC%9D%98_EP)" title="Rollin' (브브걸의 EP)">Rollin'</a>》(2017), <a href="/wiki/AKMU" title="AKMU">AKMU</a>의 《DINOSAUR》(2017), <a href="/wiki/%EC%9C%84%EB%84%88_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="위너 (음악 그룹)">WINNER</a>의 《ISLAND》(2017), <a href="/wiki/%EC%88%80_(%EC%9D%8C%EC%95%85%EA%B0%80)" title="숀 (음악가)">숀</a>의 《Way Back Home》(2018) 등이 트로피칼 하우스 장르의 음악에 해당한다.
</p><p>2020년대 중반에는 <a href="/wiki/%EC%9D%BC%EB%B3%B8%EC%9D%98_%EC%95%A0%EB%8B%88%EB%A9%94%EC%9D%B4%EC%85%98" title="일본의 애니메이션">일본의 애니메이션</a>을 모티브로 한 분위기의 K-POP 곡들이 다수 발표되었다. 이 영향은 세계의 트렌드가 '<a href="/wiki/%EB%AF%B8%EB%8B%88%EB%A9%80%EB%A6%AC%EC%A6%98" title="미니멀리즘">미니멀리즘</a>'에 가까워지고 있는 도중, 일본의 애니메이션의 히트와 J-pop의 <a href="/wiki/%ED%8B%B1%ED%86%A1" title="틱톡">틱톡</a>을 통한 인기몰이가 포착되었고 한국에서도 <a href="/wiki/%EC%9A%94%EC%95%84%EC%86%8C%EB%B9%84" title="요아소비">요아소비</a>나 <a href="/w/index.php?title=Imase&amp;action=edit&amp;redlink=1" class="new" title="Imase (없는 문서)">imase</a> 등의 일본 음악가들이 큰 인기를 누리게 되었다.<sup id="cite_ref-30" class="reference"><a href="#cite_note-30"><span class="cite-bracket">[</span>29<span class="cite-bracket">]</span></a></sup> 특히 imase의 《NIGHT DANCER》(2022)는 틱톡에서 12억 조회수를 기록하였고,<sup id="cite_ref-31" class="reference"><a href="#cite_note-31"><span class="cite-bracket">[</span>30<span class="cite-bracket">]</span></a></sup> 과거 한국에서 일본 시장을 노릴 때 사용했던 방식으로 'Korean version' 앨범을 발매했다.<sup id="cite_ref-32" class="reference"><a href="#cite_note-32"><span class="cite-bracket">[</span>31<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-33" class="reference"><a href="#cite_note-33"><span class="cite-bracket">[</span>32<span class="cite-bracket">]</span></a></sup> 그에 따라 <a href="/wiki/QWER" title="QWER">QWER</a>, <a href="/wiki/%ED%94%8C%EB%A0%88%EC%9D%B4%EB%B8%8C" title="플레이브">플레이브</a> 등 일본의 <a href="/wiki/%EC%A0%9C%EC%9D%B4%ED%8C%9D" class="mw-redirect" title="제이팝">제이팝</a> 및 <a href="/wiki/%EC%A0%9C%EC%9D%B4%EB%9D%BD" class="mw-redirect" title="제이락">제이락</a>을 모방하여 운영하는 그룹이 형성되기도 하였다. 큰 인기를 누렸던 <a href="/wiki/(%EC%97%AC%EC%9E%90)%EC%95%84%EC%9D%B4%EB%93%A4" title="(여자)아이들">(여자)아이들</a> 역시 《나는 아픈 건 딱 질색이니까》(2024)에서 제이팝의 코드를 가지고 왔다. <a href="/wiki/%EB%9D%BC%EC%9D%B4%EC%A6%88_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="라이즈 (음악 그룹)">라이즈</a> 역시 《Love 119》(2024)에서 J-POP의 요소들을 뮤직비디오에 집어넣기도 하고, <a href="/wiki/%EC%8B%A0%EC%B9%B4%EC%9D%B4_%EB%A7%88%EC%BD%94%ED%86%A0" title="신카이 마코토">신카이 마코토</a>의 <a href="/wiki/%EB%84%88%EC%9D%98_%EC%9D%B4%EB%A6%84%EC%9D%80." title="너의 이름은.">너의 이름은.</a>을 모티브로 한 장면을 넣는 등 다양한 곳에서 일본의 요소들을 삽입하였다.<sup id="cite_ref-34" class="reference"><a href="#cite_note-34"><span class="cite-bracket">[</span>33<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-:1_35-0" class="reference"><a href="#cite_note-:1-35"><span class="cite-bracket">[</span>34<span class="cite-bracket">]</span></a></sup> <a href="/wiki/IVE" title="IVE">IVE</a> 역시 《Accendio》(2024)에서 <a href="/wiki/%EB%A7%88%EB%B2%95%EC%86%8C%EB%85%80" title="마법소녀">마법소녀</a>를 모티브로 한 뮤직비디오를 선보이기도 하였으며, <a href="/wiki/%EC%95%84%EC%9D%BC%EB%A6%BF" class="mw-redirect" title="아일릿">아일릿</a>의 《Magnetic》(2024) 역시 90년대 일본의 마법소녀 애니메이션에서 나온 요소들을 차용하기도 하였다.<sup id="cite_ref-:1_35-1" class="reference"><a href="#cite_note-:1-35"><span class="cite-bracket">[</span>34<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-36" class="reference"><a href="#cite_note-36"><span class="cite-bracket">[</span>35<span class="cite-bracket">]</span></a></sup>
</p>
<div class="mw-heading mw-heading2"><h2 id="단점과_비판"><span id=".EB.8B.A8.EC.A0.90.EA.B3.BC_.EB.B9.84.ED.8C.90"></span>단점과 비판</h2><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=17" title="부분 편집: 단점과 비판"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<div class="mw-heading mw-heading3"><h3 id="선정성"><span id=".EC.84.A0.EC.A0.95.EC.84.B1"></span>선정성</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=18" title="부분 편집: 선정성"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>2010년대 중반 경부터 심해진 K-POP의 선정성은 종종 논란거리가 되었다. 많은 아이돌 그룹들은 과도하게 노출된 의상과 선정적인 안무를 선보였고, 이는 특히 청소년들에게 부정적인 영향을 미칠 수 있다는 우려와 비판이 있었다. 당시 설문조사에서 선정성으로 마케팅을 진행했던 대표적인 음반들로 <a href="/wiki/%ED%8A%B8%EB%9F%AC%EB%B8%94_%EB%A9%94%EC%9D%B4%EC%BB%A4_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="트러블 메이커 (음악 그룹)">트러블 메이커</a>의 《내일은 없어》와 <a href="/wiki/%EC%8A%A4%ED%85%94%EB%9D%BC_(%EC%9D%8C%EC%95%85_%EA%B7%B8%EB%A3%B9)" title="스텔라 (음악 그룹)">스텔라</a>의 《마리오네트》의 음악을 유료로 다운로드했다는 비율은 고작 6%에 그쳤다. 특히 71%의 조사 응답자는 해당 걸그룹의 이미지나 인기에 오히려 부정적인 인식을 갖고 있다고 답변했다.<sup id="cite_ref-37" class="reference"><a href="#cite_note-37"><span class="cite-bracket">[</span>36<span class="cite-bracket">]</span></a></sup>
</p><p>시간이 지나 K-POP의 선정성은 줄어들기 시작하고, 비교적 더 보수적인 국가들에게서도 인기를 끌게 되었다. 가장 대표적으로 중동에서<sup id="cite_ref-38" class="reference"><a href="#cite_note-38"><span class="cite-bracket">[</span>37<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-39" class="reference"><a href="#cite_note-39"><span class="cite-bracket">[</span>38<span class="cite-bracket">]</span></a></sup>
</p><p>2020년대 초에는 <a href="/wiki/%EC%8A%A4%ED%8F%AC%EC%B8%A0%EC%A1%B0%EC%84%A0" title="스포츠조선">스포츠조선</a>을 중심으로 K-POP 아이돌의 선정성 문제가 제기되었다. <a href="/wiki/(%EC%97%AC%EC%9E%90)%EC%95%84%EC%9D%B4%EB%93%A4" title="(여자)아이들">(여자)아이들</a>의 《Wife》는 가사의 묘사에서 지나친 선정성이 나타나 <a href="/wiki/KBS" class="mw-redirect" title="KBS">KBS</a>에서 부적격 판정을 받았다.<sup id="cite_ref-40" class="reference"><a href="#cite_note-40"><span class="cite-bracket">[</span>39<span class="cite-bracket">]</span></a></sup> 하지만 오히려 이런 모습이 일종의 큰 서사의 일종이라는 의견도 존재했다.<sup id="cite_ref-41" class="reference"><a href="#cite_note-41"><span class="cite-bracket">[</span>40<span class="cite-bracket">]</span></a></sup> 또 동일 그룹의 《Nxde》 공연도 선정성 논란에 휩싸였다. 해당 음악의 가사는 선정적인 것을 비판하고 있지만 그 내용와 반대로 가장 선정적인 의상을 한 것이 좋지 않다는 입장과, 해당 곡이 오히려 여성의 신체를 '야한 것'으로만 보려는 부정적인 시선과 편견을 비판하는 곡이라는 의견이 대립하게 되었다.<sup id="cite_ref-42" class="reference"><a href="#cite_note-42"><span class="cite-bracket">[</span>41<span class="cite-bracket">]</span></a></sup><sup id="cite_ref-43" class="reference"><a href="#cite_note-43"><span class="cite-bracket">[</span>42<span class="cite-bracket">]</span></a></sup>
</p><p>2024년 10월, 말레이시아 보건부에서는 로제의 APT가 유혹적인 맥락이 있는 가사가 들어있다고 비판하기도 하였다. 하지만 말레이시아 현지의 여론은 동의하기 어렵다는 반응을 보이기도 했다.<sup id="cite_ref-44" class="reference"><a href="#cite_note-44"><span class="cite-bracket">[</span>43<span class="cite-bracket">]</span></a></sup>
</p>
<div class="mw-heading mw-heading3"><h3 id="경쟁과_아티스트의_짧은_수명"><span id=".EA.B2.BD.EC.9F.81.EA.B3.BC_.EC.95.84.ED.8B.B0.EC.8A.A4.ED.8A.B8.EC.9D.98_.EC.A7.A7.EC.9D.80_.EC.88.98.EB.AA.85"></span>경쟁과 아티스트의 짧은 수명</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=19" title="부분 편집: 경쟁과 아티스트의 짧은 수명"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>케이팝의 특징 중 하나는 시장이 빠르게 변하고 경쟁도 있는 점이다. 한국에서는 1년에 60팀이 넘는 아이돌그룹이 양산된다. 하지만 대부분의 그룹들은 일부 극소수의 그룹을 제외하면 멤버가 탈퇴하는 경우가 많고 계약기간이 만료되거나 기간을 채우지 못하고 해체되는 경우가 대다수이다.<sup id="cite_ref-45" class="reference"><a href="#cite_note-45"><span class="cite-bracket">[</span>44<span class="cite-bracket">]</span></a></sup>
그리고 극소수만 성공하거나 살아남기도 한다.
</p>
<div class="mw-heading mw-heading3"><h3 id="히트곡의_짧은_수명"><span id=".ED.9E.88.ED.8A.B8.EA.B3.A1.EC.9D.98_.EC.A7.A7.EC.9D.80_.EC.88.98.EB.AA.85"></span>히트곡의 짧은 수명</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=20" title="부분 편집: 히트곡의 짧은 수명"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p><a href="/wiki/%EC%BD%94%EB%A6%AC%EC%95%84_%ED%83%80%EC%9E%84%EC%A6%88" class="mw-redirect" title="코리아 타임즈">코리아 타임즈</a>에 따르면 케이팝은 쉽게 소비되는, 즉 수명이 짧은 히트곡을 주로 생산해 낸다고 한다. <link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36482206"><span title="이 정보가 어느 시점에 해당하는지 알 수 없습니다." class="fix">현재</span><sup class="noprint" title="2025년 1월에 수정이 요청되었습니다.">[<a href="/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EB%AA%A8%ED%98%B8%ED%95%9C_%ED%91%9C%ED%98%84%EC%9D%84_%EC%82%BC%EA%B0%80_%EC%A3%BC%EC%84%B8%EC%9A%94" class="mw-redirect" title="위키백과:모호한 표현을 삼가 주세요">언제</a>?]</sup> <a href="/wiki/%EC%9D%B4%EC%A7%80%EB%A6%AC%EC%8A%A4%EB%8B%9D" class="mw-redirect" title="이지리스닝">이지리스닝</a>이 유행하면서 현상은 더 심화됐다.
</p><p>대부분의 케이팝 노래는 차트에서 잠시 인기를 얻고는 금방 사라지며 몇 주간 차트에 진입해 있는 노래는 몇 되지 않는다. 케이팝 노래는 주로 쉬운 멜로디에 쉽게 따라할 수 있는 춤으로 구성된다 — <a href="/wiki/%EC%8A%88%ED%8D%BC%EC%A3%BC%EB%8B%88%EC%96%B4" title="슈퍼주니어">슈퍼주니어</a>의 ‘쏘리쏘리’, <a href="/wiki/%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80" title="소녀시대">소녀시대</a>의 ‘지’, <a href="/wiki/%EB%B8%8C%EB%9D%BC%EC%9A%B4%EC%95%84%EC%9D%B4%EB%93%9C%EA%B1%B8%EC%8A%A4" class="mw-redirect" title="브라운아이드걸스">브라운아이드걸스</a>의 ‘아브라카다브라다. 가수 <a href="/wiki/%EC%9D%B8%EC%88%9C%EC%9D%B4" title="인순이">인순이</a>는 “예전에 우리가 하던 노래는 아직까지 불리고 있다. 하지만 요즘 음악은 3개월이 채 가지 않는다. 음악시장의 책임감이 사라지고 있다”는 비판을 제기하기도 했다.
</p><p>빨리 소비되고 사라진다는 주장에 대해서는 좀 더 객관적이고 장기적인 검증이 필요하다. 비판하는 쪽에서 예로 든 곡들은 대부분 오래 시간이 지났지만 대중과 K-pop 팬들에 의해 사랑 받고 있다. 히트곡이 차트에서 한 달에서 두 달 정도 소비되고 신곡으로 대체되는 상황은 다른 나라도 마찬가지다. 오히려 창작자들의 활발한 창작을 긍정적으로 볼 수도 있다.
</p>
<div class="mw-heading mw-heading3"><h3 id="훅송_논란"><span id=".ED.9B.85.EC.86.A1_.EB.85.BC.EB.9E.80"></span><a href="/wiki/%ED%9B%85%EC%86%A1" class="mw-redirect" title="훅송">훅송</a> 논란</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=21" title="부분 편집: 훅송 논란"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>반복되는 구절로 인한 <a href="/wiki/%ED%9B%85%EC%86%A1" class="mw-redirect" title="훅송">훅송</a>의 범람을 야기한다는 논란은 구미를 비롯한 대한민국 전국 방방곡곡의 다양한 장르의 다른 대중 음악들도 다를 게 없어 K-pop 붐의 초기 이후 크게 다루어지지 않고 있다.
</p>
<div class="mw-heading mw-heading3"><h3 id="주제_편향"><span id=".EC.A3.BC.EC.A0.9C_.ED.8E.B8.ED.96.A5"></span>주제 편향</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=22" title="부분 편집: 주제 편향"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>K-pop은 <a href="/wiki/%EC%82%AC%EB%9E%91" title="사랑">사랑</a>이 대부분을 차지한다. 그러나 대중들은 주제가 너무나 획일적이라고 비판한다.
</p>
<div class="mw-heading mw-heading3"><h3 id="과도한_외국어_사용"><span id=".EA.B3.BC.EB.8F.84.ED.95.9C_.EC.99.B8.EA.B5.AD.EC.96.B4_.EC.82.AC.EC.9A.A9"></span>과도한 외국어 사용</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=23" title="부분 편집: 과도한 외국어 사용"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>K-pop의 노래 중에는 한 두 문장만 외국어 가사가 쓰인 것이 아니라, 한국어가 거의 보이지 않는 노래들이 다수 존재한다. 이는 다른 국가나 다른 언어권의 음악에 비해 매우 이례적으로 비쳐진다. 대부분의 사람들은 이러한 모습을 좋아하지 않아 조금만 외국어를 줄였으면 좋겠다는 반응을 보인다.
</p>
<div class="mw-heading mw-heading3"><h3 id="외국_발매_우선"><span id=".EC.99.B8.EA.B5.AD_.EB.B0.9C.EB.A7.A4_.EC.9A.B0.EC.84.A0"></span>외국 발매 우선</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=24" title="부분 편집: 외국 발매 우선"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<p>케이팝 노래 중 히트곡을 만들기 위해 해외 앨범, 혹은 외국어판 음반으로 먼저 발매한 후 시간이 지나 한국에 발매를 하는 양상을 가지고 있기도 하다. 일본의 음반 시장을 노리기 위해 가사를 <a href="/wiki/%EC%9D%BC%EB%B3%B8%EC%96%B4" title="일본어">일본어</a>로 번역하고 개사한 판이 일본에 먼저 발매된 뒤, 그 다음으로 한국 시장에 발매가 되는 형식이다. 허나 이 문제는 2020년까지도 지속되고 있는 문제이다.
</p>
<div class="mw-heading mw-heading3"><h3 id="표절_논란"><span id=".ED.91.9C.EC.A0.88_.EB.85.BC.EB.9E.80"></span>표절 논란</h3><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=25" title="부분 편집: 표절 논란"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<style data-mw-deduplicate="TemplateStyles:r34311305">.mw-parser-output .hatnote{}.mw-parser-output div.hatnote{padding-left:1.6em;margin-bottom:0.5em}.mw-parser-output .hatnote i{font-style:normal}.mw-parser-output .hatnote+link+.hatnote{margin-top:-0.5em}</style><div role="note" class="hatnote navigation-not-searchable"><span typeof="mw:File"><span><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/5/52/Icons8_flat_search.svg/18px-Icons8_flat_search.svg.png" decoding="async" width="18" height="18" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/5/52/Icons8_flat_search.svg/27px-Icons8_flat_search.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/5/52/Icons8_flat_search.svg/36px-Icons8_flat_search.svg.png 2x" data-file-width="512" data-file-height="512"></span></span>&nbsp;이 부분의 본문은 <a href="/wiki/%ED%91%9C%EC%A0%88_%EB%85%BC%EB%9E%80%EC%97%90_%ED%9C%A9%EC%8B%B8%EC%9D%B8_K-pop_%EC%9D%8C%EC%95%85" title="표절 논란에 휩싸인 K-pop 음악">표절 논란에 휩싸인 K-pop 음악</a>입니다.</div>
<p>과거부터 K-POP의 여러 노래들이 해외 싱어송라이터들의 음악을 베껴온 것이라는 주장이 지속되었으나, 별다른 지지를 받지 못하다가 2022년 7월 경 <a href="/wiki/%EC%9D%B4%EB%AC%B4%EC%A7%84" title="이무진">이무진</a>의 <a href="/wiki/%EC%8B%A0%ED%98%B8%EB%93%B1_(%EB%85%B8%EB%9E%98)" title="신호등 (노래)">신호등 (노래)</a>가 이 사건의 도화선이 되었다. 논란이 거세지자, 이에 이무진의 소속사 빅플래닛메이드엔터에서는 공식 입장문을 제시하였다.<sup id="cite_ref-46" class="reference"><a href="#cite_note-46"><span class="cite-bracket">[</span>45<span class="cite-bracket">]</span></a></sup>
</p>
<div class="mw-heading mw-heading2"><h2 id="같이_보기"><span id=".EA.B0.99.EC.9D.B4_.EB.B3.B4.EA.B8.B0"></span>같이 보기</h2><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=26" title="부분 편집: 같이 보기"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r38501082"><div class="side-box metadata side-box-right plainlinks"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480595">
<div class="side-box-flex">
<div class="side-box-image"><span typeof="mw:File"><span><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Commons-logo.svg/40px-Commons-logo.svg.png" decoding="async" width="30" height="40" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Commons-logo.svg/60px-Commons-logo.svg.png 1.5x" data-file-width="1024" data-file-height="1376"></span></span></div>
<div class="side-box-text plainlist"><b><a href="/wiki/%EC%9C%84%ED%82%A4%EB%AF%B8%EB%94%94%EC%96%B4_%EA%B3%B5%EC%9A%A9" title="위키미디어 공용">위키미디어 공용</a></b>에 관련된<br>미디어 분류가 있습니다.<div style="padding-left:1em;"><b><a class="external text" href="https://commons.wikimedia.org/wiki/Category:K-pop?uselang=ko">K-pop</a></b></div></div></div>
</div>
<ul><li><a href="/wiki/%ED%95%9C%EB%A5%98_(%EB%AC%B8%ED%99%94)" class="mw-redirect" title="한류 (문화)">한류 (문화)</a></li>
<li><a href="/wiki/K-ROCK" class="mw-redirect" title="K-ROCK">K-ROCK</a></li>
<li><a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%95%84%EC%9D%B4%EB%8F%8C_%EA%B7%B8%EB%A3%B9_%EB%AA%A9%EB%A1%9D" title="대한민국의 아이돌 그룹 목록">대한민국의 아이돌 그룹 목록</a></li>
<li><a href="/wiki/K-pop_%EC%9D%8C%EC%95%85%EA%B0%80_%EB%AA%A9%EB%A1%9D" title="K-pop 음악가 목록">K-pop 음악가 목록</a></li>
<li><a href="/wiki/%EC%95%84%EC%9D%B4%EB%8F%8C" title="아이돌">아이돌</a></li>
<li><a href="/wiki/%EC%BD%94%EB%A6%AC%EC%96%B8_%EB%AE%A4%EC%A7%81_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C" class="mw-redirect" title="코리언 뮤직 페스티벌">코리언 뮤직 페스티벌</a></li></ul>
<div class="mw-heading mw-heading2"><h2 id="각주"><span id=".EA.B0.81.EC.A3.BC"></span>각주</h2><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=K-pop&amp;action=edit&amp;section=27" title="부분 편집: 각주"><span>편집</span></a><span class="mw-editsection-bracket">]</span></span></div>
<dl><dt>내용주</dt></dl>
<style data-mw-deduplicate="TemplateStyles:r35556958">.mw-parser-output .reflist{font-size:90%;margin-bottom:0.5em;list-style-type:decimal}.mw-parser-output .reflist .references{font-size:100%;margin-bottom:0;list-style-type:inherit}.mw-parser-output .reflist-columns-2{column-width:30em}.mw-parser-output .reflist-columns-3{column-width:25em}.mw-parser-output .reflist-columns{margin-top:0.3em}.mw-parser-output .reflist-columns ol{margin-top:0}.mw-parser-output .reflist-columns li{page-break-inside:avoid;break-inside:avoid-column}.mw-parser-output .reflist-upper-alpha{list-style-type:upper-alpha}.mw-parser-output .reflist-upper-roman{list-style-type:upper-roman}.mw-parser-output .reflist-lower-alpha{list-style-type:lower-alpha}.mw-parser-output .reflist-lower-greek{list-style-type:lower-greek}.mw-parser-output .reflist-lower-roman{list-style-type:lower-roman}</style><div class="reflist">
<div class="mw-references-wrap"><ol class="references">
<li id="cite_note-9"><span class="mw-cite-backlink"><a href="#cite_ref-9" aria-label="이동" title="이동">↑</a></span> <span class="reference-text">단, 위 방탄소년단의 세 곡은 해외 DJ가 작곡한 반주에 방탄소년단 그룹의 보컬을 입혀 발매한 노래이기 때문에 K-pop이 아닌 <a href="/wiki/%EB%94%94%EC%8A%A4%EC%BD%94" title="디스코">디스코</a> 장르 혹은 일반 <a href="/wiki/%ED%8C%9D_%EC%9D%8C%EC%95%85" title="팝 음악">팝</a> 장르라고 분류되는 경우도 있다.</span>
</li>
</ol></div></div>
<dl><dt>각주</dt></dl>
<link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r35556958"><div class="reflist">
<div class="mw-references-wrap mw-references-columns"><ol class="references">
<li id="cite_note-1"><span class="mw-cite-backlink"><a href="#cite_ref-1" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news">윤여수 (2011년 1월 25일). <a rel="nofollow" class="external text" href="https://web.archive.org/web/20131203004115/http://news.donga.com/3//20110125/34350544/1">“김건모 ‘잘못된 만남’ 첫 200만장 돌파”</a>. 동아닷컴. 2013년 12월 3일에 <a rel="nofollow" class="external text" href="http://news.donga.com/3//20110125/34350544/1">원본 문서</a>에서 보존된 문서<span class="reference-accessdate">. 2011년 3월 13일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=%EA%B9%80%EA%B1%B4%EB%AA%A8+%E2%80%98%EC%9E%98%EB%AA%BB%EB%90%9C+%EB%A7%8C%EB%82%A8%E2%80%99+%EC%B2%AB+200%EB%A7%8C%EC%9E%A5+%EB%8F%8C%ED%8C%8C&amp;rft.date=2011-01-25&amp;rft.au=%EC%9C%A4%EC%97%AC%EC%88%98&amp;rft_id=http%3A%2F%2Fnews.donga.com%2F3%2F%2F20110125%2F34350544%2F1&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-2"><span class="mw-cite-backlink"><a href="#cite_ref-2" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news">김보경 (2011년 2월 11일). <a rel="nofollow" class="external text" href="http://www.newsen.com/news_view.php?uid=201102111107071001">“‘댄싱 퀸’ 김완선 6년 공백 깨고 올 3월 컴백”</a>. 뉴스엔<span class="reference-accessdate">. 2011년 3월 13일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=%E2%80%98%EB%8C%84%EC%8B%B1+%ED%80%B8%E2%80%99+%EA%B9%80%EC%99%84%EC%84%A0+6%EB%85%84+%EA%B3%B5%EB%B0%B1+%EA%B9%A8%EA%B3%A0+%EC%98%AC+3%EC%9B%94+%EC%BB%B4%EB%B0%B1&amp;rft.date=2011-02-11&amp;rft.au=%EA%B9%80%EB%B3%B4%EA%B2%BD&amp;rft_id=http%3A%2F%2Fwww.newsen.com%2Fnews_view.php%3Fuid%3D201102111107071001&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-3"><span class="mw-cite-backlink"><a href="#cite_ref-3" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="http://rki.kbs.co.kr/english/program/program_artist_detail.htm?No=10198">“DJ DOC”</a>. KBS World<span class="reference-accessdate">. 2012년 12월 10일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=DJ+DOC&amp;rft.pub=KBS+World&amp;rft_id=http%3A%2F%2Frki.kbs.co.kr%2Fenglish%2Fprogram%2Fprogram_artist_detail.htm%3FNo%3D10198&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-4"><span class="mw-cite-backlink"><a href="#cite_ref-4" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news"><a rel="nofollow" class="external text" href="https://news.naver.com/main/read.nhn?mode=LSD&amp;mid=sec&amp;sid1=106&amp;oid=022&amp;aid=0000137915">“2005 가요계 결산…온라인 음악시장에서 <i>희망</i> 찾다”</a>. 세계일보. 2005년 12월 25일<span class="reference-accessdate">. 2011년 3월 13일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=2005+%EA%B0%80%EC%9A%94%EA%B3%84+%EA%B2%B0%EC%82%B0%E2%80%A6%EC%98%A8%EB%9D%BC%EC%9D%B8+%EC%9D%8C%EC%95%85%EC%8B%9C%EC%9E%A5%EC%97%90%EC%84%9C+%ED%9D%AC%EB%A7%9D+%EC%B0%BE%EB%8B%A4&amp;rft.date=2005-12-25&amp;rft_id=https%3A%2F%2Fnews.naver.com%2Fmain%2Fread.nhn%3Fmode%3DLSD%26mid%3Dsec%26sid1%3D106%26oid%3D022%26aid%3D0000137915&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-5"><span class="mw-cite-backlink"><a href="#cite_ref-5" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external text" href="http://www.accesshollywood.com/south-korean-pop-sensation-wonder-girls-hits-the-states_article_25353">South Korean Pop Sensation Wonder Girls Hits The States | Access Hollywood – Celebrity News, Photos &amp; Videos</a>. Access Hollywood (2009-11-10). Retrieved on 2011-02-20.</span>
</li>
<li id="cite_note-6"><span class="mw-cite-backlink"><a href="#cite_ref-6" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news">최준용 (2010년 10월 27일). <a rel="nofollow" class="external text" href="http://www.asiae.co.kr/news/view.htm?idxno=2010102707543571707">“소녀시대, 日 오리콘 1위 달성··세가지 중요한 의미는”</a>. 아시아경제<span class="reference-accessdate">. 2011년 2월 25일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80%2C+%E6%97%A5+%EC%98%A4%EB%A6%AC%EC%BD%98+1%EC%9C%84+%EB%8B%AC%EC%84%B1%C2%B7%C2%B7%EC%84%B8%EA%B0%80%EC%A7%80+%EC%A4%91%EC%9A%94%ED%95%9C+%EC%9D%98%EB%AF%B8%EB%8A%94&amp;rft.date=2010-10-27&amp;rft.au=%EC%B5%9C%EC%A4%80%EC%9A%A9&amp;rft_id=http%3A%2F%2Fwww.asiae.co.kr%2Fnews%2Fview.htm%3Fidxno%3D2010102707543571707&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-7"><span class="mw-cite-backlink"><a href="#cite_ref-7" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external autonumber" href="https://www.allkpop.com/article/2012/03/big-bangs-alive-charts-on-the-billboard-200">[1]</a></span>
</li>
<li id="cite_note-8"><span class="mw-cite-backlink"><a href="#cite_ref-8" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external autonumber" href="https://www.billboard.com/articles/columns/k-town/7625656/bigbang-made-billboard-200-charts">[2]</a></span>
</li>
<li id="cite_note-10"><span class="mw-cite-backlink"><a href="#cite_ref-10" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">김, 도균 (2021년 5월 14일). <a rel="nofollow" class="external text" href="https://www.vop.co.kr/A00001569855.html">“(인터뷰①) ‘싱어게인’ 이후 첫 신곡 이무진 “‘신호등’에 20대 청춘들 고민 담았죠””</a>. 《민중의소리》<span class="reference-accessdate">. 2022년 4월 3일에 확인함</span>. <q>이제 막 성인이 된 청춘의 심정, 사회라는 울타리 안에서 인간관계, 법, 융통성, 돈 등 복잡한 사회적 개념과 법칙을 겪으며 혼란스러워하는 20대의 고민을 담았다</q></cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=unknown&amp;rft.jtitle=%EB%AF%BC%EC%A4%91%EC%9D%98%EC%86%8C%EB%A6%AC&amp;rft.atitle=%28%EC%9D%B8%ED%84%B0%EB%B7%B0%E2%91%A0%29+%E2%80%98%EC%8B%B1%EC%96%B4%EA%B2%8C%EC%9D%B8%E2%80%99+%EC%9D%B4%ED%9B%84+%EC%B2%AB+%EC%8B%A0%EA%B3%A1+%EC%9D%B4%EB%AC%B4%EC%A7%84+%E2%80%9C%E2%80%98%EC%8B%A0%ED%98%B8%EB%93%B1%E2%80%99%EC%97%90+20%EB%8C%80+%EC%B2%AD%EC%B6%98%EB%93%A4+%EA%B3%A0%EB%AF%BC+%EB%8B%B4%EC%95%98%EC%A3%A0%E2%80%9D&amp;rft.date=2021-05-14&amp;rft.aulast=%EA%B9%80&amp;rft.aufirst=%EB%8F%84%EA%B7%A0&amp;rft_id=https%3A%2F%2Fwww.vop.co.kr%2FA00001569855.html&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-11"><span class="mw-cite-backlink"><a href="#cite_ref-11" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.mhns.co.kr/news/articleView.html?idxno=614689">“[사건정리] 피프티피프티, 분쟁의 시작부터 '2기' 데뷔까지”</a>. 2024년 7월 11일<span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%5B%EC%82%AC%EA%B1%B4%EC%A0%95%EB%A6%AC%5D+%ED%94%BC%ED%94%84%ED%8B%B0%ED%94%BC%ED%94%84%ED%8B%B0%2C+%EB%B6%84%EC%9F%81%EC%9D%98+%EC%8B%9C%EC%9E%91%EB%B6%80%ED%84%B0+%272%EA%B8%B0%27+%EB%8D%B0%EB%B7%94%EA%B9%8C%EC%A7%80&amp;rft.date=2024-07-11&amp;rft_id=https%3A%2F%2Fwww.mhns.co.kr%2Fnews%2FarticleView.html%3Fidxno%3D614689&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-12"><span class="mw-cite-backlink"><a href="#cite_ref-12" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.hankookilbo.com/News/Read/A2024090313190001589">“<span style="padding-left:0.2em;">"</span>답답하다"는 뉴진스·팬들은 하이브 고발...민희진 해임 후폭풍”</a>. 2024년 9월 3일<span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%22%EB%8B%B5%EB%8B%B5%ED%95%98%EB%8B%A4%22%EB%8A%94+%EB%89%B4%EC%A7%84%EC%8A%A4%C2%B7%ED%8C%AC%EB%93%A4%EC%9D%80+%ED%95%98%EC%9D%B4%EB%B8%8C+%EA%B3%A0%EB%B0%9C...%EB%AF%BC%ED%9D%AC%EC%A7%84+%ED%95%B4%EC%9E%84+%ED%9B%84%ED%8F%AD%ED%92%8D&amp;rft.date=2024-09-03&amp;rft_id=https%3A%2F%2Fwww.hankookilbo.com%2FNews%2FRead%2FA2024090313190001589&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-13"><span class="mw-cite-backlink"><a href="#cite_ref-13" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.mk.co.kr/news/musics/11109215">““2026년 11월까지 민희진 임기 보장하라” 뉴진스 팬들, 하이브에 공개서한”</a>. 2024년 9월 4일<span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%E2%80%9C2026%EB%85%84+11%EC%9B%94%EA%B9%8C%EC%A7%80+%EB%AF%BC%ED%9D%AC%EC%A7%84+%EC%9E%84%EA%B8%B0+%EB%B3%B4%EC%9E%A5%ED%95%98%EB%9D%BC%E2%80%9D+%EB%89%B4%EC%A7%84%EC%8A%A4+%ED%8C%AC%EB%93%A4%2C+%ED%95%98%EC%9D%B4%EB%B8%8C%EC%97%90+%EA%B3%B5%EA%B0%9C%EC%84%9C%ED%95%9C&amp;rft.date=2024-09-04&amp;rft_id=https%3A%2F%2Fwww.mk.co.kr%2Fnews%2Fmusics%2F11109215&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-14"><span class="mw-cite-backlink"><a href="#cite_ref-14" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">김채운 (2024년 8월 7일). <a rel="nofollow" class="external text" href="https://www.hani.co.kr/arti/society/society_general/1152624.html">“BTS 슈가 음주운전…전동스쿠터 몰다 넘어진 채 발견”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=BTS+%EC%8A%88%EA%B0%80+%EC%9D%8C%EC%A3%BC%EC%9A%B4%EC%A0%84%E2%80%A6%EC%A0%84%EB%8F%99%EC%8A%A4%EC%BF%A0%ED%84%B0+%EB%AA%B0%EB%8B%A4+%EB%84%98%EC%96%B4%EC%A7%84+%EC%B1%84+%EB%B0%9C%EA%B2%AC&amp;rft.date=2024-08-07&amp;rft.au=%EA%B9%80%EC%B1%84%EC%9A%B4&amp;rft_id=https%3A%2F%2Fwww.hani.co.kr%2Farti%2Fsociety%2Fsociety_general%2F1152624.html&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-15"><span class="mw-cite-backlink"><a href="#cite_ref-15" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://v.daum.net/v/20240830143835948">“<span style="padding-left:0.2em;">'</span>전동 스쿠터 음주 운전' BTS 슈가, 불구속 송치”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%27%EC%A0%84%EB%8F%99+%EC%8A%A4%EC%BF%A0%ED%84%B0+%EC%9D%8C%EC%A3%BC+%EC%9A%B4%EC%A0%84%27+BTS+%EC%8A%88%EA%B0%80%2C+%EB%B6%88%EA%B5%AC%EC%86%8D+%EC%86%A1%EC%B9%98&amp;rft_id=https%3A%2F%2Fv.daum.net%2Fv%2F20240830143835948&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-16"><span class="mw-cite-backlink"><a href="#cite_ref-16" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.donga.com/news/Opinion/article/all/20230228/118117375/1">“K팝 ‘1인 체제’ 극복 못하면 J팝 몰락 따라간다[수요논점]”</a>. 2023년 3월 1일<span class="reference-accessdate">. 2024년 8월 27일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=K%ED%8C%9D+%E2%80%981%EC%9D%B8+%EC%B2%B4%EC%A0%9C%E2%80%99+%EA%B7%B9%EB%B3%B5+%EB%AA%BB%ED%95%98%EB%A9%B4+J%ED%8C%9D+%EB%AA%B0%EB%9D%BD+%EB%94%B0%EB%9D%BC%EA%B0%84%EB%8B%A4%5B%EC%88%98%EC%9A%94%EB%85%BC%EC%A0%90%5D&amp;rft.date=2023-03-01&amp;rft_id=https%3A%2F%2Fwww.donga.com%2Fnews%2FOpinion%2Farticle%2Fall%2F20230228%2F118117375%2F1&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-17"><span class="mw-cite-backlink"><a href="#cite_ref-17" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news">김연수 (2024년 11월 11일). <a rel="nofollow" class="external text" href="https://www.yna.co.kr/view/AKR20241111078200505">“[SNS세상] 쏟아지는 '아파트' 패러디…인기 1위는 딥페이크 | 연합뉴스”</a>. 《연합뉴스》<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=%EC%97%B0%ED%95%A9%EB%89%B4%EC%8A%A4&amp;rft.atitle=%5BSNS%EC%84%B8%EC%83%81%5D+%EC%8F%9F%EC%95%84%EC%A7%80%EB%8A%94+%27%EC%95%84%ED%8C%8C%ED%8A%B8%27+%ED%8C%A8%EB%9F%AC%EB%94%94%E2%80%A6%EC%9D%B8%EA%B8%B0+1%EC%9C%84%EB%8A%94+%EB%94%A5%ED%8E%98%EC%9D%B4%ED%81%AC+%7C+%EC%97%B0%ED%95%A9%EB%89%B4%EC%8A%A4&amp;rft.date=2024-11-11&amp;rft.au=%EA%B9%80%EC%97%B0%EC%88%98&amp;rft_id=https%3A%2F%2Fwww.yna.co.kr%2Fview%2FAKR20241111078200505&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-18"><span class="mw-cite-backlink"><a href="#cite_ref-18" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">부위원장, 박정훈 민주노총 공공운수 노조 (2024년 12월 2일). <a rel="nofollow" class="external text" href="https://www.khan.co.kr/article/202412022049005">“[직설]로제 ‘아파트’, 노동자 ‘총파업’”</a><span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%5B%EC%A7%81%EC%84%A4%5D%EB%A1%9C%EC%A0%9C+%E2%80%98%EC%95%84%ED%8C%8C%ED%8A%B8%E2%80%99%2C+%EB%85%B8%EB%8F%99%EC%9E%90+%E2%80%98%EC%B4%9D%ED%8C%8C%EC%97%85%E2%80%99&amp;rft.date=2024-12-02&amp;rft.aulast=%EB%B6%80%EC%9C%84%EC%9B%90%EC%9E%A5&amp;rft.aufirst=%EB%B0%95%EC%A0%95%ED%9B%88+%EB%AF%BC%EC%A3%BC%EB%85%B8%EC%B4%9D+%EA%B3%B5%EA%B3%B5%EC%9A%B4%EC%88%98+%EB%85%B8%EC%A1%B0&amp;rft_id=https%3A%2F%2Fwww.khan.co.kr%2Farticle%2F202412022049005&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-19"><span class="mw-cite-backlink"><a href="#cite_ref-19" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.mk.co.kr/news/world/11194805">“외신도 주목한 ‘K-팝 응원봉’ 시위…“콘서트 같다, 새 시위 문화””</a>. 2024년 12월 15일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%99%B8%EC%8B%A0%EB%8F%84+%EC%A3%BC%EB%AA%A9%ED%95%9C+%E2%80%98K-%ED%8C%9D+%EC%9D%91%EC%9B%90%EB%B4%89%E2%80%99+%EC%8B%9C%EC%9C%84%E2%80%A6%E2%80%9C%EC%BD%98%EC%84%9C%ED%8A%B8+%EA%B0%99%EB%8B%A4%2C+%EC%83%88+%EC%8B%9C%EC%9C%84+%EB%AC%B8%ED%99%94%E2%80%9D&amp;rft.date=2024-12-15&amp;rft_id=https%3A%2F%2Fwww.mk.co.kr%2Fnews%2Fworld%2F11194805&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-20"><span class="mw-cite-backlink"><a href="#cite_ref-20" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">장예지 (2024년 12월 15일). <a rel="nofollow" class="external text" href="https://www.hani.co.kr/arti/international/international_general/1173131.html">“외신 “K팝 콘서트장 같은 탄핵집회…민주주의 위해 온 세대 함께해””</a><span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%99%B8%EC%8B%A0+%E2%80%9CK%ED%8C%9D+%EC%BD%98%EC%84%9C%ED%8A%B8%EC%9E%A5+%EA%B0%99%EC%9D%80+%ED%83%84%ED%95%B5%EC%A7%91%ED%9A%8C%E2%80%A6%EB%AF%BC%EC%A3%BC%EC%A3%BC%EC%9D%98+%EC%9C%84%ED%95%B4+%EC%98%A8+%EC%84%B8%EB%8C%80+%ED%95%A8%EA%BB%98%ED%95%B4%E2%80%9D&amp;rft.date=2024-12-15&amp;rft.au=%EC%9E%A5%EC%98%88%EC%A7%80&amp;rft_id=https%3A%2F%2Fwww.hani.co.kr%2Farti%2Finternational%2Finternational_general%2F1173131.html&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-21"><span class="mw-cite-backlink"><a href="#cite_ref-21" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.ohmynews.com/NWS_Web/View/at_pg.aspx?CNTN_CD=A0003089439">“소녀시대의 명곡은 어떻게 광장의 노래가 되었나”</a>. 2024년 12월 17일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80%EC%9D%98+%EB%AA%85%EA%B3%A1%EC%9D%80+%EC%96%B4%EB%96%BB%EA%B2%8C+%EA%B4%91%EC%9E%A5%EC%9D%98+%EB%85%B8%EB%9E%98%EA%B0%80+%EB%90%98%EC%97%88%EB%82%98&amp;rft.date=2024-12-17&amp;rft_id=https%3A%2F%2Fwww.ohmynews.com%2FNWS_Web%2FView%2Fat_pg.aspx%3FCNTN_CD%3DA0003089439&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-22"><span class="mw-cite-backlink"><a href="#cite_ref-22" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.sportsseoul.com/news/read/1483779">“윤석열 대통령 탄핵 가결과 동시에 국회 앞에서 울려 퍼진 소녀시대 ‘다시 만난 세계’ [SS초점]”</a>. 2024년 12월 14일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%9C%A4%EC%84%9D%EC%97%B4+%EB%8C%80%ED%86%B5%EB%A0%B9+%ED%83%84%ED%95%B5+%EA%B0%80%EA%B2%B0%EA%B3%BC+%EB%8F%99%EC%8B%9C%EC%97%90+%EA%B5%AD%ED%9A%8C+%EC%95%9E%EC%97%90%EC%84%9C+%EC%9A%B8%EB%A0%A4+%ED%8D%BC%EC%A7%84+%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80+%E2%80%98%EB%8B%A4%EC%8B%9C+%EB%A7%8C%EB%82%9C+%EC%84%B8%EA%B3%84%E2%80%99+%5BSS%EC%B4%88%EC%A0%90%5D&amp;rft.date=2024-12-14&amp;rft_id=https%3A%2F%2Fwww.sportsseoul.com%2Fnews%2Fread%2F1483779&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-23"><span class="mw-cite-backlink"><a href="#cite_ref-23" aria-label="이동" title="이동">↑</a></span> <span class="reference-text">[Williamson, Lucy (2011-06-14). "The dark side of South Korean pop music". BBC. Retrieved 2012-09-17.]</span>
</li>
<li id="cite_note-24"><span class="mw-cite-backlink"><a href="#cite_ref-24" aria-label="이동" title="이동">↑</a></span> <span class="reference-text">["'Uncool' Korean star Psy goes viral with 'Gangnam Style'". Ibnlive.in.com. 2012-08-27. Retrieved 2012-09-17.]</span>
</li>
<li id="cite_note-25"><span class="mw-cite-backlink"><a href="#cite_ref-25" aria-label="이동" title="이동">↑</a></span> <span class="reference-text">[John, Seabrook. "Factory Girls- Cultural technology and the making of K-pop.". The New Yorker. Retrieved 23 April 2013.]</span>
</li>
<li id="cite_note-26"><span class="mw-cite-backlink"><a href="#cite_ref-26" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation journal">Jin, Dal Yong; Ryoo, Woongjae (2014년 3월 15일). <a rel="nofollow" class="external text" href="http://www.tandfonline.com/doi/abs/10.1080/03007766.2012.731721">“Critical Interpretation of Hybrid K-Pop: The Global-Local Paradigm of English Mixing in Lyrics”</a>. 《Popular Music and Society》 (영어) <b>37</b> (2): 113–131. <a href="/wiki/%EB%94%94%EC%A7%80%ED%84%B8_%EA%B0%9D%EC%B2%B4_%EC%8B%9D%EB%B3%84%EC%9E%90" title="디지털 객체 식별자">doi</a>:<a rel="nofollow" class="external text" href="https://dx.doi.org/10.1080%2F03007766.2012.731721">10.1080/03007766.2012.731721</a>. <a href="/wiki/%EA%B5%AD%EC%A0%9C_%ED%91%9C%EC%A4%80_%EC%9D%BC%EB%A0%A8_%EB%B2%88%ED%98%B8" class="mw-redirect" title="국제 표준 일련 번호">ISSN</a>&nbsp;<a rel="nofollow" class="external text" href="//www.worldcat.org/issn/0300-7766">0300-7766</a>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=Popular+Music+and+Society&amp;rft.atitle=Critical+Interpretation+of+Hybrid+K-Pop%3A+The+Global-Local+Paradigm+of+English+Mixing+in+Lyrics&amp;rft.volume=37&amp;rft.issue=2&amp;rft.pages=113-131&amp;rft.date=2014-03-15&amp;rft_id=info%3Adoi%2F10.1080%2F03007766.2012.731721&amp;rft.issn=0300-7766&amp;rft.aulast=Jin&amp;rft.aufirst=Dal+Yong&amp;rft.au=Ryoo%2C+Woongjae&amp;rft_id=http%3A%2F%2Fwww.tandfonline.com%2Fdoi%2Fabs%2F10.1080%2F03007766.2012.731721&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-27"><span class="mw-cite-backlink"><a href="#cite_ref-27" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">Park, Ju-min (2019년 5월 7일). <a rel="nofollow" class="external text" href="https://web.archive.org/web/20190606051141/https://www.japantimes.co.jp/culture/2019/05/07/entertainment-news/k-pop-stardom-lures-young-japanese-hopefuls-south-korea-despite-diplomatic-chill/">“K-Pop stardom lures young Japanese hopefuls to South Korea despite diplomatic chill”</a>. 《<a href="/wiki/The_Japan_Times" class="mw-redirect" title="The Japan Times">The Japan Times</a>》 (미국 영어). 2019년 6월 6일에 <a rel="nofollow" class="external text" href="https://www.japantimes.co.jp/culture/2019/05/07/entertainment-news/k-pop-stardom-lures-young-japanese-hopefuls-south-korea-despite-diplomatic-chill/">원본 문서</a>에서 보존된 문서<span class="reference-accessdate">. 2019년 6월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=unknown&amp;rft.jtitle=The+Japan+Times&amp;rft.atitle=K-Pop+stardom+lures+young+Japanese+hopefuls+to+South+Korea+despite+diplomatic+chill&amp;rft.date=2019-05-07&amp;rft.aulast=Park&amp;rft.aufirst=Ju-min&amp;rft_id=https%3A%2F%2Fwww.japantimes.co.jp%2Fculture%2F2019%2F05%2F07%2Fentertainment-news%2Fk-pop-stardom-lures-young-japanese-hopefuls-south-korea-despite-diplomatic-chill%2F&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-28"><span class="mw-cite-backlink"><a href="#cite_ref-28" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news">이태수 (2024년 5월 12일). <a rel="nofollow" class="external text" href="https://www.yna.co.kr/view/AKR20240511034200005">“틱톡 등 숏폼이 바꾼 K팝…"듣는 음악에서 참여하는 음악으로<span style="padding-right:0.2em;">"</span>”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=%ED%8B%B1%ED%86%A1+%EB%93%B1+%EC%88%8F%ED%8F%BC%EC%9D%B4+%EB%B0%94%EA%BE%BC+K%ED%8C%9D%E2%80%A6%22%EB%93%A3%EB%8A%94+%EC%9D%8C%EC%95%85%EC%97%90%EC%84%9C+%EC%B0%B8%EC%97%AC%ED%95%98%EB%8A%94+%EC%9D%8C%EC%95%85%EC%9C%BC%EB%A1%9C%22&amp;rft.date=2024-05-12&amp;rft.au=%EC%9D%B4%ED%83%9C%EC%88%98&amp;rft_id=https%3A%2F%2Fwww.yna.co.kr%2Fview%2FAKR20240511034200005&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-29"><span class="mw-cite-backlink"><a href="#cite_ref-29" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">리오, 박 (2023년 8월 24일). <a rel="nofollow" class="external text" href="https://cjnews.cj.net/%EA%B8%80%EB%A1%9C%EB%B2%8C-%EB%AC%B4%EB%8C%80-%EC%9A%B0%EB%9A%9D-%EC%84%A0-k-pop-%EC%84%B1%EA%B3%B5-%EB%B9%84%EA%B2%B0%EC%9D%80/">“글로벌 무대 우뚝 선 K-POP, 성공 비결은?”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EA%B8%80%EB%A1%9C%EB%B2%8C+%EB%AC%B4%EB%8C%80+%EC%9A%B0%EB%9A%9D+%EC%84%A0+K-POP%2C+%EC%84%B1%EA%B3%B5+%EB%B9%84%EA%B2%B0%EC%9D%80%3F&amp;rft.date=2023-08-24&amp;rft.aulast=%EB%A6%AC%EC%98%A4&amp;rft.aufirst=%EB%B0%95&amp;rft_id=https%3A%2F%2Fcjnews.cj.net%2F%25EA%25B8%2580%25EB%25A1%259C%25EB%25B2%258C-%25EB%25AC%25B4%25EB%258C%2580-%25EC%259A%25B0%25EB%259A%259D-%25EC%2584%25A0-k-pop-%25EC%2584%25B1%25EA%25B3%25B5-%25EB%25B9%2584%25EA%25B2%25B0%25EC%259D%2580%2F&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-30"><span class="mw-cite-backlink"><a href="#cite_ref-30" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.korea.kr/news/policyNewsView.do?newsId=148922968">“‘제이팝(J-POP)’의 귀환…왜?”</a>. 2023년 11월 24일<span class="reference-accessdate">. 2024년 6월 17일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%E2%80%98%EC%A0%9C%EC%9D%B4%ED%8C%9D%28J-POP%29%E2%80%99%EC%9D%98+%EA%B7%80%ED%99%98%E2%80%A6%EC%99%9C%3F&amp;rft.date=2023-11-24&amp;rft_id=https%3A%2F%2Fwww.korea.kr%2Fnews%2FpolicyNewsView.do%3FnewsId%3D148922968&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-31"><span class="mw-cite-backlink"><a href="#cite_ref-31" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">연예, iMBC (2023년 2월 24일). <a rel="nofollow" class="external text" href="https://imnews.imbc.com/news/2023/enter/article/6458746_36161.html">“12억 조회수 日'괴물신예' 이마세 'NIGHT DANCER', 차트 점령”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=12%EC%96%B5+%EC%A1%B0%ED%9A%8C%EC%88%98+%E6%97%A5%27%EA%B4%B4%EB%AC%BC%EC%8B%A0%EC%98%88%27+%EC%9D%B4%EB%A7%88%EC%84%B8+%27NIGHT+DANCER%27%2C+%EC%B0%A8%ED%8A%B8+%EC%A0%90%EB%A0%B9&amp;rft.date=2023-02-24&amp;rft.aulast=%EC%97%B0%EC%98%88&amp;rft.aufirst=iMBC&amp;rft_id=https%3A%2F%2Fimnews.imbc.com%2Fnews%2F2023%2Fenter%2Farticle%2F6458746_36161.html&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-32"><span class="mw-cite-backlink"><a href="#cite_ref-32" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.jayupress.com/news/articleView.html?idxno=17905">“<span style="padding-left:0.2em;">'</span>J팝계의 신성' 이마세 "이젠 한국어로도 노래해요<span style="padding-right:0.2em;">"</span>”</a>. 2023년 5월 30일<span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%27J%ED%8C%9D%EA%B3%84%EC%9D%98+%EC%8B%A0%EC%84%B1%27+%EC%9D%B4%EB%A7%88%EC%84%B8+%22%EC%9D%B4%EC%A0%A0+%ED%95%9C%EA%B5%AD%EC%96%B4%EB%A1%9C%EB%8F%84+%EB%85%B8%EB%9E%98%ED%95%B4%EC%9A%94%22&amp;rft.date=2023-05-30&amp;rft_id=https%3A%2F%2Fwww.jayupress.com%2Fnews%2FarticleView.html%3Fidxno%3D17905&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-33"><span class="mw-cite-backlink"><a href="#cite_ref-33" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">이데일리 (2024년 4월 24일). <a rel="nofollow" class="external text" href="https://m.edaily.co.kr/News/Read?newsId=01354646638859400&amp;mediaCodeNo=258">“이마세, 첫 정규 앨범 韓 특별판 발매…빅나티와 협업”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%9D%B4%EB%A7%88%EC%84%B8%2C+%EC%B2%AB+%EC%A0%95%EA%B7%9C+%EC%95%A8%EB%B2%94+%E9%9F%93+%ED%8A%B9%EB%B3%84%ED%8C%90+%EB%B0%9C%EB%A7%A4%E2%80%A6%EB%B9%85%EB%82%98%ED%8B%B0%EC%99%80+%ED%98%91%EC%97%85&amp;rft.date=2024-04-24&amp;rft.au=%EC%9D%B4%EB%8D%B0%EC%9D%BC%EB%A6%AC&amp;rft_id=https%3A%2F%2Fm.edaily.co.kr%2FNews%2FRead%3FnewsId%3D01354646638859400%26mediaCodeNo%3D258&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-34"><span class="mw-cite-backlink"><a href="#cite_ref-34" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web">박정빈 (2024년 6월 28일). <a rel="nofollow" class="external text" href="https://brunch.co.kr/@kpopberdiblue/252">“QWER, 아일릿, 투어스… ‘오타쿠’ 트렌드가 뜬다”</a><span class="reference-accessdate">. 2024년 8월 27일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=QWER%2C+%EC%95%84%EC%9D%BC%EB%A6%BF%2C+%ED%88%AC%EC%96%B4%EC%8A%A4%E2%80%A6+%E2%80%98%EC%98%A4%ED%83%80%EC%BF%A0%E2%80%99+%ED%8A%B8%EB%A0%8C%EB%93%9C%EA%B0%80+%EB%9C%AC%EB%8B%A4&amp;rft.date=2024-06-28&amp;rft.au=%EB%B0%95%EC%A0%95%EB%B9%88&amp;rft_id=https%3A%2F%2Fbrunch.co.kr%2F%40kpopberdiblue%2F252&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-:1-35"><span class="mw-cite-backlink">↑ <sup><a href="#cite_ref-:1_35-0"><span class="cite-accessibility-label">이동: </span>가</a></sup> <sup><a href="#cite_ref-:1_35-1">나</a></sup></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://star.ohmynews.com/NWS_Web/OhmyStar/at_pg.aspx?CNTN_CD=A0003042158">“아일릿, 투어스… 케이팝에 '오타쿠' 트렌드가 뜬다”</a>. 2024년 6월 28일<span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%95%84%EC%9D%BC%EB%A6%BF%2C+%ED%88%AC%EC%96%B4%EC%8A%A4%E2%80%A6+%EC%BC%80%EC%9D%B4%ED%8C%9D%EC%97%90+%27%EC%98%A4%ED%83%80%EC%BF%A0%27+%ED%8A%B8%EB%A0%8C%EB%93%9C%EA%B0%80+%EB%9C%AC%EB%8B%A4&amp;rft.date=2024-06-28&amp;rft_id=https%3A%2F%2Fstar.ohmynews.com%2FNWS_Web%2FOhmyStar%2Fat_pg.aspx%3FCNTN_CD%3DA0003042158&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-36"><span class="mw-cite-backlink"><a href="#cite_ref-36" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation news">최주성 (2024년 5월 20일). <a rel="nofollow" class="external text" href="https://www.yna.co.kr/view/AKR20240519051400005">“파워레인저·마법소녀…뮤비 속 히어로로 변신한 걸그룹”</a><span class="reference-accessdate">. 2024년 9월 6일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.atitle=%ED%8C%8C%EC%9B%8C%EB%A0%88%EC%9D%B8%EC%A0%80%C2%B7%EB%A7%88%EB%B2%95%EC%86%8C%EB%85%80%E2%80%A6%EB%AE%A4%EB%B9%84+%EC%86%8D+%ED%9E%88%EC%96%B4%EB%A1%9C%EB%A1%9C+%EB%B3%80%EC%8B%A0%ED%95%9C+%EA%B1%B8%EA%B7%B8%EB%A3%B9&amp;rft.date=2024-05-20&amp;rft.au=%EC%B5%9C%EC%A3%BC%EC%84%B1&amp;rft_id=https%3A%2F%2Fwww.yna.co.kr%2Fview%2FAKR20240519051400005&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-37"><span class="mw-cite-backlink"><a href="#cite_ref-37" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation journal">장규수 / 김태룡. <a rel="nofollow" class="external text" href="https://www.kci.go.kr/kciportal/ci/sereArticleSearch/ciSereArtiView.kci?sereArticleSearchBean.artiId=ART001962786">“케이팝의 선정성과 흥행효과에 관한 연구: &lt;내일은 없어&gt;와 &lt;마리오네트&gt;를 중심으로”</a>. 《글로벌문화콘텐츠》.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal&amp;rft.genre=article&amp;rft.jtitle=%EA%B8%80%EB%A1%9C%EB%B2%8C%EB%AC%B8%ED%99%94%EC%BD%98%ED%85%90%EC%B8%A0&amp;rft.atitle=%EC%BC%80%EC%9D%B4%ED%8C%9D%EC%9D%98+%EC%84%A0%EC%A0%95%EC%84%B1%EA%B3%BC+%ED%9D%A5%ED%96%89%ED%9A%A8%EA%B3%BC%EC%97%90+%EA%B4%80%ED%95%9C+%EC%97%B0%EA%B5%AC%3A+%3C%EB%82%B4%EC%9D%BC%EC%9D%80+%EC%97%86%EC%96%B4%3E%EC%99%80+%3C%EB%A7%88%EB%A6%AC%EC%98%A4%EB%84%A4%ED%8A%B8%3E%EB%A5%BC+%EC%A4%91%EC%8B%AC%EC%9C%BC%EB%A1%9C&amp;rft.au=%EC%9E%A5%EA%B7%9C%EC%88%98+%2F+%EA%B9%80%ED%83%9C%EB%A3%A1&amp;rft_id=https%3A%2F%2Fwww.kci.go.kr%2Fkciportal%2Fci%2FsereArticleSearch%2FciSereArtiView.kci%3FsereArticleSearchBean.artiId%3DART001962786&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-38"><span class="mw-cite-backlink"><a href="#cite_ref-38" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://news.kbs.co.kr/news/pc/view/view.do?ncd=5569524">“사우디에서도 K-pop 콘서트…“한류 건전” 중동 다시 열기”</a><span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%EC%82%AC%EC%9A%B0%EB%94%94%EC%97%90%EC%84%9C%EB%8F%84+K-pop+%EC%BD%98%EC%84%9C%ED%8A%B8%E2%80%A6%E2%80%9C%ED%95%9C%EB%A5%98+%EA%B1%B4%EC%A0%84%E2%80%9D+%EC%A4%91%EB%8F%99+%EB%8B%A4%EC%8B%9C+%EC%97%B4%EA%B8%B0&amp;rft_id=https%3A%2F%2Fnews.kbs.co.kr%2Fnews%2Fpc%2Fview%2Fview.do%3Fncd%3D5569524&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-39"><span class="mw-cite-backlink"><a href="#cite_ref-39" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://news.kbs.co.kr/news/pc/view/view.do?ncd=5572537">“[특파원 리포트] 사우디가 ‘들썩’…중동은 K-pop을 어디까지 받아들일 수 있을까?”</a><span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%5B%ED%8A%B9%ED%8C%8C%EC%9B%90+%EB%A6%AC%ED%8F%AC%ED%8A%B8%5D+%EC%82%AC%EC%9A%B0%EB%94%94%EA%B0%80+%E2%80%98%EB%93%A4%EC%8D%A9%E2%80%99%E2%80%A6%EC%A4%91%EB%8F%99%EC%9D%80+K-pop%EC%9D%84+%EC%96%B4%EB%94%94%EA%B9%8C%EC%A7%80+%EB%B0%9B%EC%95%84%EB%93%A4%EC%9D%BC+%EC%88%98+%EC%9E%88%EC%9D%84%EA%B9%8C%3F&amp;rft_id=https%3A%2F%2Fnews.kbs.co.kr%2Fnews%2Fpc%2Fview%2Fview.do%3Fncd%3D5572537&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-40"><span class="mw-cite-backlink"><a href="#cite_ref-40" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://v.daum.net/v/20240124100429121">“(여자)아이들 '와이프', 선정성 논란→KBS 심의 부적격 판정 "지나친 묘사 때문" [SC이슈]”</a><span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%28%EC%97%AC%EC%9E%90%29%EC%95%84%EC%9D%B4%EB%93%A4+%27%EC%99%80%EC%9D%B4%ED%94%84%27%2C+%EC%84%A0%EC%A0%95%EC%84%B1+%EB%85%BC%EB%9E%80%E2%86%92KBS+%EC%8B%AC%EC%9D%98+%EB%B6%80%EC%A0%81%EA%B2%A9+%ED%8C%90%EC%A0%95+%22%EC%A7%80%EB%82%98%EC%B9%9C+%EB%AC%98%EC%82%AC+%EB%95%8C%EB%AC%B8%22+%5BSC%EC%9D%B4%EC%8A%88%5D&amp;rft_id=https%3A%2F%2Fv.daum.net%2Fv%2F20240124100429121&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-41"><span class="mw-cite-backlink"><a href="#cite_ref-41" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://star.ohmynews.com/NWS_Web/OhmyStar/at_pg.aspx?CNTN_CD=A0002996120">“<span style="padding-left:0.2em;">'</span>와이프' 선정성 논란... 어쩌면 (여자)아이들의 큰 그림?”</a>. 2024년 1월 24일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%27%EC%99%80%EC%9D%B4%ED%94%84%27+%EC%84%A0%EC%A0%95%EC%84%B1+%EB%85%BC%EB%9E%80...+%EC%96%B4%EC%A9%8C%EB%A9%B4+%28%EC%97%AC%EC%9E%90%29%EC%95%84%EC%9D%B4%EB%93%A4%EC%9D%98+%ED%81%B0+%EA%B7%B8%EB%A6%BC%3F&amp;rft.date=2024-01-24&amp;rft_id=https%3A%2F%2Fstar.ohmynews.com%2FNWS_Web%2FOhmyStar%2Fat_pg.aspx%3FCNTN_CD%3DA0002996120&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-42"><span class="mw-cite-backlink"><a href="#cite_ref-42" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.donga.com/news/Entertainment/article/all/20230105/117312431/1">““어디까지 내리나”…걸그룹, 로라이즈 패션 선정성 논란”</a>. 2023년 1월 5일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%E2%80%9C%EC%96%B4%EB%94%94%EA%B9%8C%EC%A7%80+%EB%82%B4%EB%A6%AC%EB%82%98%E2%80%9D%E2%80%A6%EA%B1%B8%EA%B7%B8%EB%A3%B9%2C+%EB%A1%9C%EB%9D%BC%EC%9D%B4%EC%A6%88+%ED%8C%A8%EC%85%98+%EC%84%A0%EC%A0%95%EC%84%B1+%EB%85%BC%EB%9E%80&amp;rft.date=2023-01-05&amp;rft_id=https%3A%2F%2Fwww.donga.com%2Fnews%2FEntertainment%2Farticle%2Fall%2F20230105%2F117312431%2F1&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-43"><span class="mw-cite-backlink"><a href="#cite_ref-43" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://sports.chosun.com/entertainment/2023-01-05/202301060100039990004579">“[SC이슈](여자)아이들 민니, 의상이 어떠했길래 선정성 논란? 팬티 윗…”</a>. 2023년 1월 5일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%5BSC%EC%9D%B4%EC%8A%88%5D%28%EC%97%AC%EC%9E%90%29%EC%95%84%EC%9D%B4%EB%93%A4+%EB%AF%BC%EB%8B%88%2C+%EC%9D%98%EC%83%81%EC%9D%B4+%EC%96%B4%EB%96%A0%ED%96%88%EA%B8%B8%EB%9E%98+%EC%84%A0%EC%A0%95%EC%84%B1+%EB%85%BC%EB%9E%80%3F+%ED%8C%AC%ED%8B%B0+%EC%9C%97%E2%80%A6&amp;rft.date=2023-01-05&amp;rft_id=https%3A%2F%2Fsports.chosun.com%2Fentertainment%2F2023-01-05%2F202301060100039990004579&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-44"><span class="mw-cite-backlink"><a href="#cite_ref-44" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><cite class="citation web"><a rel="nofollow" class="external text" href="https://www.kmib.co.kr/article/view.asp?arcid=0020667166">““유혹적 맥락 안돼” 말레이시아 보건부, 로제 ‘아파트’ 가사 지적”</a>. 2024년 10월 28일<span class="reference-accessdate">. 2025년 1월 15일에 확인함</span>.</cite><span title="ctx_ver=Z39.88-2004&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook&amp;rft.genre=unknown&amp;rft.btitle=%E2%80%9C%EC%9C%A0%ED%98%B9%EC%A0%81+%EB%A7%A5%EB%9D%BD+%EC%95%88%EB%8F%BC%E2%80%9D+%EB%A7%90%EB%A0%88%EC%9D%B4%EC%8B%9C%EC%95%84+%EB%B3%B4%EA%B1%B4%EB%B6%80%2C+%EB%A1%9C%EC%A0%9C+%E2%80%98%EC%95%84%ED%8C%8C%ED%8A%B8%E2%80%99+%EA%B0%80%EC%82%AC+%EC%A7%80%EC%A0%81&amp;rft.date=2024-10-28&amp;rft_id=https%3A%2F%2Fwww.kmib.co.kr%2Farticle%2Fview.asp%3Farcid%3D0020667166&amp;rfr_id=info%3Asid%2Fko.wikipedia.org%3AK-pop" class="Z3988"><span style="display:none;">&nbsp;</span></span></span>
</li>
<li id="cite_note-45"><span class="mw-cite-backlink"><a href="#cite_ref-45" aria-label="이동" title="이동">↑</a></span> <span class="reference-text">["How K-pop may have lowered Korean Music Standards". HelloKpop. 2011-12-01. Retrieved 2012-03-27.]</span>
</li>
<li id="cite_note-46"><span class="mw-cite-backlink"><a href="#cite_ref-46" aria-label="이동" title="이동">↑</a></span> <span class="reference-text"><a rel="nofollow" class="external free" href="https://n.news.naver.com/mnews/article/005/0001541230?sid=103">https://n.news.naver.com/mnews/article/005/0001541230?sid=103</a></span>
</li>
</ol></div></div>
<div class="navbox-styles"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><style data-mw-deduplicate="TemplateStyles:r38501015">.mw-parser-output .navbox{box-sizing:border-box;border:1px solid #a2a9b1;width:100%;clear:both;font-size:88%;text-align:center;padding:1px;margin:1em auto 0}.mw-parser-output .navbox .navbox{margin-top:0}.mw-parser-output .navbox+.navbox,.mw-parser-output .navbox+.navbox-styles+.navbox{margin-top:-1px}.mw-parser-output .navbox-inner,.mw-parser-output .navbox-subgroup{width:100%}.mw-parser-output .navbox-group,.mw-parser-output .navbox-title,.mw-parser-output .navbox-abovebelow{padding:0.25em 1em;line-height:1.5em;text-align:center}.mw-parser-output .navbox-group{white-space:nowrap;text-align:right}.mw-parser-output .navbox,.mw-parser-output .navbox-subgroup{background-color:#fdfdfd}.mw-parser-output .navbox-list{line-height:1.5em;border-color:#fdfdfd}.mw-parser-output .navbox-list-with-group{text-align:left;border-left-width:2px;border-left-style:solid}.mw-parser-output tr+tr>.navbox-abovebelow,.mw-parser-output tr+tr>.navbox-group,.mw-parser-output tr+tr>.navbox-image,.mw-parser-output tr+tr>.navbox-list{border-top:2px solid #fdfdfd}.mw-parser-output .navbox-title{background-color:#ccf}.mw-parser-output .navbox-abovebelow,.mw-parser-output .navbox-group,.mw-parser-output .navbox-subgroup .navbox-title{background-color:#ddf}.mw-parser-output .navbox-subgroup .navbox-group,.mw-parser-output .navbox-subgroup .navbox-abovebelow{background-color:#e6e6ff}.mw-parser-output .navbox-even{background-color:#f7f7f7}.mw-parser-output .navbox-odd{background-color:transparent}.mw-parser-output .navbox .hlist td dl,.mw-parser-output .navbox .hlist td ol,.mw-parser-output .navbox .hlist td ul,.mw-parser-output .navbox td.hlist dl,.mw-parser-output .navbox td.hlist ol,.mw-parser-output .navbox td.hlist ul{padding:0.125em 0}.mw-parser-output .navbox .navbar{display:block;font-size:100%}.mw-parser-output .navbox-title .navbar{float:left;text-align:left;margin-right:0.5em}body.skin--responsive .mw-parser-output .navbox-image img{max-width:none!important}@media print{body.ns-0 .mw-parser-output .navbox{display:none!important}}</style></div><div role="navigation" class="navbox" aria-labelledby="K-pop" style="padding:3px"><table class="nowraplinks hlist mw-collapsible mw-collapsed navbox-inner mw-made-collapsible" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><button type="button" class="mw-collapsible-toggle mw-collapsible-toggle-default mw-collapsible-toggle-collapsed" aria-expanded="false" tabindex="0"><span class="mw-collapsible-text">펼치기</span></button><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><style data-mw-deduplicate="TemplateStyles:r38585741">.mw-parser-output .navbar{display:inline;font-size:88%;font-weight:normal}.mw-parser-output .navbar-collapse{float:left;text-align:left}.mw-parser-output .navbar-boxtext{word-spacing:0}.mw-parser-output .navbar ul{display:inline-block;white-space:nowrap;line-height:inherit}.mw-parser-output .navbar-brackets::before{margin-right:-0.125em;content:"[ "}.mw-parser-output .navbar-brackets::after{margin-left:-0.125em;content:" ]"}.mw-parser-output .navbar li{word-spacing:-0.125em}.mw-parser-output .navbar a>span,.mw-parser-output .navbar a>abbr{text-decoration:inherit}.mw-parser-output .navbar-mini abbr{font-variant:small-caps;border-bottom:none;text-decoration:none;cursor:inherit}.mw-parser-output .navbar-ct-full{font-size:114%;margin:0 7em}.mw-parser-output .navbar-ct-mini{font-size:114%;margin:0 4em}html.skin-theme-clientpref-night .mw-parser-output .navbar li a abbr{color:var(--color-base)!important}@media(prefers-color-scheme:dark){html.skin-theme-clientpref-os .mw-parser-output .navbar li a abbr{color:var(--color-base)!important}}@media print{.mw-parser-output .navbar{display:none!important}}</style><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-보기"><a href="/wiki/%ED%8B%80:K-pop" title="틀:K-pop"><abbr title="이 틀을 보기" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">v</abbr></a></li><li class="nv-토론"><a href="/w/index.php?title=%ED%8B%80%ED%86%A0%EB%A1%A0:K-pop&amp;action=edit&amp;redlink=1" class="new" title="틀토론:K-pop (없는 문서)"><abbr title="이 틀에 관해 토론하기" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">t</abbr></a></li><li class="nv-편집"><a href="/wiki/%ED%8A%B9%EC%88%98:%EB%AC%B8%EC%84%9C%ED%8E%B8%EC%A7%91/%ED%8B%80:K-pop" title="특수:문서편집/틀:K-pop"><abbr title="이 틀을 편집하기" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">e</abbr></a></li></ul></div><div id="K-pop" style="font-size:114%;margin:0 4em"><a class="mw-selflink selflink">K-pop</a></div></th></tr><tr style="display: none;"><td class="navbox-abovebelow" colspan="2"><div>
<ul><li><a href="/wiki/%ED%95%9C%EB%A5%98_(%EB%AC%B8%ED%99%94)" class="mw-redirect" title="한류 (문화)">한류 (문화)</a></li>
<li><a href="/w/index.php?title=%ED%95%9C%EA%B5%AD%EC%9D%98_%EB%B0%9C%EB%9D%BC%EB%93%9C&amp;action=edit&amp;redlink=1" class="new" title="한국의 발라드 (없는 문서)">한국의 발라드</a></li>
<li><a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%ED%9E%99%ED%95%A9" title="대한민국의 힙합">대한민국의 힙합</a></li>
<li><a href="/wiki/K-ROCK" class="mw-redirect" title="K-ROCK">K-ROCK</a></li>
<li><a href="/wiki/%ED%8A%B8%EB%A1%9C%ED%8A%B8" title="트로트">트로트</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">용어</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EB%AC%B8%ED%99%94%EA%B8%B0%EC%88%A0" title="문화기술">문화기술</a></li>
<li><a href="/wiki/%EC%8C%80_%ED%99%94%ED%99%98" title="쌀 화환">쌀 화환</a></li>
<li><a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%95%84%EC%9D%B4%EB%8F%8C" title="대한민국의 아이돌">아이돌</a></li>
<li><a class="mw-selflink selflink">K-pop</a></li>
<li><a href="/wiki/%EC%82%AC%EC%83%9D%ED%8C%AC" title="사생팬">사생팬</a></li>
<li><a href="/wiki/%EB%85%B8%EC%98%88_%EA%B3%84%EC%95%BD" title="노예 계약">노예 계약</a></li>
<li><a href="/wiki/%ED%8A%B8%EB%9F%AD_%EC%8B%9C%EC%9C%84" title="트럭 시위">트럭 시위</a></li>
<li><a href="/wiki/%ED%99%80%EB%A1%9C%EA%B7%B8%EB%9E%A8_%EC%BD%98%EC%84%9C%ED%8A%B8" title="홀로그램 콘서트">홀로그램 콘서트</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">음악가</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/K-pop_%EC%9D%8C%EC%95%85%EA%B0%80_%EB%AA%A9%EB%A1%9D" title="K-pop 음악가 목록">개인 음악가</a></li>
<li><a href="/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%95%84%EC%9D%B4%EB%8F%8C_%EA%B7%B8%EB%A3%B9_%EB%AA%A9%EB%A1%9D" title="대한민국의 아이돌 그룹 목록">아이돌 그룹</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">행사</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/w/index.php?title=%EC%95%84%EC%8B%9C%EC%95%84_%EB%B0%96%EC%97%90%EC%84%9C_%EA%B0%9C%EC%B5%9C%EB%90%9C_K-pop_%EC%BD%98%EC%84%9C%ED%8A%B8_%EB%AA%A9%EB%A1%9D&amp;action=edit&amp;redlink=1" class="new" title="아시아 밖에서 개최된 K-pop 콘서트 목록 (없는 문서)">아시아 밖에서 개최된 K-pop 콘서트 목록</a></li>
<li><a href="/wiki/%EB%93%9C%EB%A6%BC%EC%BD%98%EC%84%9C%ED%8A%B8" title="드림콘서트">드림콘서트</a></li>
<li><a href="/wiki/KCON" title="KCON">KCON</a></li>
<li><a href="/wiki/K%ED%8C%9D_%EC%BB%A4%EB%B2%84%EB%8C%84%EC%8A%A4_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C" title="K팝 커버댄스 페스티벌">K팝 커버댄스 페스티벌</a></li>
<li><a href="/w/index.php?title=K-Pop_%EC%9B%94%EB%93%9C_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C&amp;action=edit&amp;redlink=1" class="new" title="K-Pop 월드 페스티벌 (없는 문서)">K-Pop 월드 페스티벌</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">웹사이트</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%98%AC%EC%BC%80%EC%9D%B4%ED%8C%9D" title="올케이팝">올케이팝</a></li>
<li><a href="/w/index.php?title=%EB%B9%8C%EB%B3%B4%EB%93%9C_K-Town&amp;action=edit&amp;redlink=1" class="new" title="빌보드 K-Town (없는 문서)">빌보드 K-Town</a>
<ul><li><a href="/w/index.php?title=%EB%B9%8C%EB%B3%B4%EB%93%9C_K-pop_%EC%97%B0%ED%91%9C&amp;action=edit&amp;redlink=1" class="new" title="빌보드 K-pop 연표 (없는 문서)">연표</a></li></ul></li>
<li><a href="/wiki/Eatyourkimchi" title="Eatyourkimchi">Eatyourkimchi</a></li>
<li><a href="/wiki/%EC%88%A8%ED%94%BC" title="숨피">숨피</a></li>
<li><a href="/w/index.php?title=%EC%BD%94%EB%A6%AC%EC%95%84%EB%B6%80&amp;action=edit&amp;redlink=1" class="new" title="코리아부 (없는 문서)">코리아부</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%"><a href="/w/index.php?title=%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%8C%EC%95%85_%ED%94%84%EB%A1%9C%EA%B7%B8%EB%9E%A8&amp;action=edit&amp;redlink=1" class="new" title="대한민국의 음악 프로그램 (없는 문서)">음악 쇼</a></th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/SBS_%EC%9D%B8%EA%B8%B0%EA%B0%80%EC%9A%94" title="SBS 인기가요">SBS 인기가요</a> <span style="font-size:85%">(<a href="/wiki/SBS" title="SBS">SBS</a>)</span></li>
<li><a href="/wiki/%EC%97%A0%EC%B9%B4%EC%9A%B4%ED%8A%B8%EB%8B%A4%EC%9A%B4" title="엠카운트다운">엠카운트다운</a> <span style="font-size:85%">(<a href="/wiki/%EC%97%A0%EB%84%B7" title="엠넷">Mnet</a>)</span></li>
<li><a href="/wiki/%EB%AE%A4%EC%A7%81%EB%B1%85%ED%81%AC" title="뮤직뱅크">Music Bank</a> <span style="font-size:85%">(<a href="/wiki/KBS_2TV" title="KBS 2TV">KBS 2TV</a>)</span></li>
<li><a href="/wiki/%EB%AE%A4%EC%A7%81_%EC%98%A8_%ED%83%91" title="뮤직 온 탑">뮤직 온 탑</a> <span style="font-size:85%">(<a href="/wiki/JTBC" title="JTBC">JTBC</a>)</span></li>
<li><a href="/wiki/Pops_in_Seoul" title="Pops in Seoul">Pops in Seoul</a> <span style="font-size:85%">(<a href="/wiki/%EC%95%84%EB%A6%AC%EB%9E%91_TV" class="mw-redirect" title="아리랑 TV">아리랑 TV</a>)</span></li>
<li><a href="/wiki/%EC%87%BC_%EC%B1%94%ED%94%BC%EC%96%B8" title="쇼 챔피언">쇼 챔피언</a> <span style="font-size:85%">(<a href="/wiki/MBC_%EB%AE%A4%EC%A7%81" title="MBC 뮤직">MBC 뮤직</a>)</span></li>
<li><a href="/wiki/%EC%87%BC!_%EC%9D%8C%EC%95%85%EC%A4%91%EC%8B%AC" title="쇼! 음악중심">쇼! 음악중심</a> <span style="font-size:85%">(<a href="/wiki/%EB%AC%B8%ED%99%94%EB%B0%A9%EC%86%A1" title="문화방송">MBC</a>)</span></li>
<li><a href="/wiki/%EB%8D%94_%EC%87%BC" title="더 쇼">The Show</a> <span style="font-size:85%">(<a href="/wiki/SBS_MTV" class="mw-redirect" title="SBS MTV">SBS MTV</a>)</span></li>
<li><a href="/wiki/Simply_K-Pop" title="Simply K-Pop">Simply K-Pop</a> <span style="font-size:85%">(<a href="/wiki/%EC%95%84%EB%A6%AC%EB%9E%91_TV" class="mw-redirect" title="아리랑 TV">아리랑 TV</a>)</span></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">음악 차트</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/w/index.php?title=%EB%B9%8C%EB%B3%B4%EB%93%9C_%EC%B0%A8%ED%8A%B8%EC%9D%98_K-pop_%EB%AA%A9%EB%A1%9D&amp;action=edit&amp;redlink=1" class="new" title="빌보드 차트의 K-pop 목록 (없는 문서)">빌보드 차트의 K-pop 목록</a>
<ul><li><a href="/w/index.php?title=%EB%B9%8C%EB%B3%B4%EB%93%9C_%EC%B0%A8%ED%8A%B8%EC%9D%98_K-pop_%EC%9D%8C%EB%B0%98_%EB%AA%A9%EB%A1%9D&amp;action=edit&amp;redlink=1" class="new" title="빌보드 차트의 K-pop 음반 목록 (없는 문서)">음반</a></li>
<li><a href="/w/index.php?title=%EB%B9%8C%EB%B3%B4%EB%93%9C_%EC%B0%A8%ED%8A%B8%EC%9D%98_K-pop_%EB%85%B8%EB%9E%98_%EB%AA%A9%EB%A1%9D&amp;action=edit&amp;redlink=1" class="new" title="빌보드 차트의 K-pop 노래 목록 (없는 문서)">노래</a></li>
<li><a href="/w/index.php?title=%EB%B9%8C%EB%B3%B4%EB%93%9C_%EC%97%B0%EB%A7%90_%EC%B0%A8%ED%8A%B8%EC%9D%98_K-pop_%EB%AA%A9%EB%A1%9D&amp;action=edit&amp;redlink=1" class="new" title="빌보드 연말 차트의 K-pop 목록 (없는 문서)">연말</a></li></ul></li>
<li><a href="/wiki/%EA%B0%80%EC%98%A8_%EC%B0%A8%ED%8A%B8" class="mw-redirect" title="가온 차트">가온 차트</a>
<ul><li><a href="/wiki/%EA%B0%80%EC%98%A8_%EC%95%A8%EB%B2%94_%EC%B0%A8%ED%8A%B8" class="mw-redirect" title="가온 앨범 차트">앨범</a></li>
<li><a href="/wiki/%EA%B0%80%EC%98%A8_%EB%94%94%EC%A7%80%ED%84%B8_%EC%B0%A8%ED%8A%B8" class="mw-redirect" title="가온 디지털 차트">싱글</a></li></ul></li>
<li><a href="/wiki/%EC%BD%94%EB%A6%AC%EC%95%84_K-Pop_%ED%95%AB_100" class="mw-redirect" title="코리아 K-Pop 핫 100">빌보드 K-pop 핫 100</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%EB%B6%84%EB%A5%98:%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%8C%EB%B0%98%EC%82%AC" title="분류:대한민국의 음반사">저명한 레이블<br>및 퍼블리셔</a></th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em"></div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%">대행사</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/B2M_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="B2M 엔터테인먼트">B2M 엔터테인먼트</a></li>
<li><a href="/wiki/%EB%B8%94%EB%9F%AC%EC%8D%B8_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="블러썸 엔터테인먼트">블러썸 엔터테인먼트</a></li>
<li><a href="/wiki/%EC%BD%94%EB%A6%AC%EB%8D%B8_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="코리델 엔터테인먼트">코리델 엔터테인먼트</a></li>
<li><a href="/wiki/%ED%81%90%EB%B8%8C_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="큐브 엔터테인먼트">큐브 엔터테인먼트</a></li>
<li><a href="/wiki/DSP_%EB%AF%B8%EB%94%94%EC%96%B4" title="DSP 미디어">DSP 미디어</a></li>
<li><a href="/wiki/FNC_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="FNC 엔터테인먼트">FNC 엔터테인먼트</a></li>
<li><a href="/wiki/%ED%8C%90%ED%83%80%EC%A7%80%EC%98%A4" title="판타지오">판타지오</a></li>
<li><a href="/wiki/%EB%93%9C%EB%A6%BC%EC%BA%90%EC%B3%90_%EC%BB%B4%ED%8D%BC%EB%8B%88" title="드림캐쳐 컴퍼니">드림캐쳐 컴퍼니</a></li>
<li><a href="/wiki/%EC%9D%B4%EB%A7%A4%EC%A7%84%EC%95%84%EC%8B%9C%EC%95%84" title="이매진아시아">이매진아시아</a>
<ul><li><a href="/wiki/%EC%9D%B4%EB%A7%A4%EC%A7%84%EC%95%84%EC%8B%9C%EC%95%84" title="이매진아시아">블루 스타 엔터테인먼트</a></li>
<li><a href="/wiki/%EC%9D%B4%EB%A7%A4%EC%A7%84%EC%95%84%EC%8B%9C%EC%95%84" title="이매진아시아">드림 티 엔터테인먼트</a></li>
<li><a href="/wiki/%EC%9D%B4%EB%A7%A4%EC%A7%84%EC%95%84%EC%8B%9C%EC%95%84" title="이매진아시아">YMC 엔터테인먼트</a></li></ul></li>
<li><a href="/wiki/%EC%A0%A4%EB%A6%AC%ED%94%BC%EC%89%AC_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="젤리피쉬 엔터테인먼트">젤리피쉬 엔터테인먼트</a></li>
<li><a href="/wiki/HYBE" class="mw-redirect" title="HYBE">HYBE</a>
<ul><li><a href="/wiki/%EB%B9%85%ED%9E%88%ED%8A%B8_%EB%AE%A4%EC%A7%81" title="빅히트 뮤직">빅히트 뮤직</a></li>
<li><a href="/wiki/%EC%8F%98%EC%8A%A4%EB%AE%A4%EC%A7%81" title="쏘스뮤직">쏘스뮤직</a></li>
<li><a href="/wiki/%ED%94%8C%EB%A0%88%EB%94%94%EC%8A%A4_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="플레디스 엔터테인먼트">플레디스 엔터테인먼트</a></li></ul></li>
<li><a href="/wiki/JYP_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="JYP 엔터테인먼트">JYP 엔터테인먼트</a></li>
<li><a href="/wiki/%EC%B9%B4%EC%B9%B4%EC%98%A4%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="카카오엔터테인먼트">카카오엔터테인먼트</a>
<ul><li><a href="/wiki/%ED%94%8C%EB%9E%9C%EC%97%90%EC%9D%B4_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="플랜에이 엔터테인먼트">플랜에이 엔터테인먼트</a></li>
<li><a href="/wiki/%EC%8A%A4%ED%83%80%EC%89%BD_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="스타쉽 엔터테인먼트">스타쉽 엔터테인먼트</a></li>
<li><a href="/wiki/EDAM_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" class="mw-redirect" title="EDAM 엔터테인먼트">EDAM 엔터테인먼트</a></li></ul></li>
<li><a href="/wiki/MBK_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="MBK 엔터테인먼트">MBK 엔터테인먼트</a></li>
<li><a href="/w/index.php?title=Music%26NEW&amp;action=edit&amp;redlink=1" class="new" title="Music&amp;NEW (없는 문서)">Music&amp;NEW</a></li>
<li><a href="/wiki/P_NATION" title="P NATION">P NATION</a></li>
<li><a href="/wiki/%ED%94%8C%EB%A0%88%EB%94%94%EC%8A%A4_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="플레디스 엔터테인먼트">플레디스 엔터테인먼트</a></li>
<li><a href="/wiki/SM_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="SM 엔터테인먼트">SM 엔터테인먼트</a>
<ul><li><a href="/wiki/%EC%9A%B8%EB%A6%BC%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="울림엔터테인먼트">울림엔터테인먼트</a></li>
<li><a href="/wiki/%EB%AF%B8%EC%8A%A4%ED%8B%B1%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" class="mw-redirect" title="미스틱엔터테인먼트">미스틱엔터테인먼트</a></li></ul></li>
<li><a href="/wiki/%EC%94%A8%EA%B7%B8%EB%84%90%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8%EA%B7%B8%EB%A3%B9" title="씨그널엔터테인먼트그룹">씨그널엔터테인먼트그룹</a>
<ul><li><a href="/wiki/NH_%EB%AF%B8%EB%94%94%EC%96%B4" class="mw-redirect" title="NH 미디어">NH 미디어</a></li></ul></li>
<li><a href="/wiki/%EC%8A%A4%ED%86%A4%EB%AE%A4%EC%A7%81_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="스톤뮤직 엔터테인먼트">스톤뮤직 엔터테인먼트</a>
<ul><li><a href="/wiki/%EC%97%90%EC%9D%B4%EC%98%A4%EC%97%A0%EC%A7%80" class="mw-redirect" title="에이오엠지">에이오엠지</a></li>
<li><a href="/wiki/%ED%95%98%EC%9D%B4%EB%9D%BC%EC%9D%B4%ED%8A%B8_%EB%A0%88%EC%BD%94%EC%A6%88" title="하이라이트 레코즈">하이라이트 레코즈</a></li>
<li><a href="/wiki/%EC%8A%A4%ED%86%A4%EB%AE%A4%EC%A7%81_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="스톤뮤직 엔터테인먼트">1877 엔터테인먼트</a></li>
<li><a href="/wiki/MMO_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" class="mw-redirect" title="MMO 엔터테인먼트">MMO 엔터테인먼트</a></li>
<li><a href="/wiki/%EB%AE%A4%EC%A7%81%EC%9B%8D%EC%8A%A4" title="뮤직웍스">뮤직웍스</a></li></ul></li>
<li><a href="/wiki/TS_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="TS 엔터테인먼트">TS 엔터테인먼트</a></li>
<li><a href="/wiki/YG_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="YG 엔터테인먼트">YG 엔터테인먼트</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%">퍼블리셔</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%A7%80%EB%8B%88%EB%AE%A4%EC%A7%81" title="지니뮤직">지니뮤직</a></li>
<li><a href="/wiki/%EB%AC%B8%ED%99%94%EB%B0%A9%EC%86%A1" title="문화방송">iMBC</a></li>
<li><a href="/wiki/%EC%9D%B8%ED%84%B0%ED%8C%8C%ED%81%AC" class="mw-redirect" title="인터파크">인터파크 뮤직</a></li>
<li><a href="/wiki/%EC%95%84%EC%9D%B4%EB%A6%AC%EB%B2%84" class="mw-redirect" title="아이리버">아이리버</a></li>
<li><a href="/wiki/%EC%B9%B4%EC%B9%B4%EC%98%A4%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="카카오엔터테인먼트">카카오엔터테인먼트</a></li>
<li><a href="/w/index.php?title=Music%26NEW&amp;action=edit&amp;redlink=1" class="new" title="Music&amp;NEW (없는 문서)">Music&amp;NEW</a></li>
<li><a href="/wiki/NHN%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" class="mw-redirect" title="NHN엔터테인먼트">NHN 벅스</a></li>
<li><a href="/wiki/%ED%8F%AC%EB%8B%88_%EC%BA%90%EB%85%84" title="포니 캐년">포니 캐년 코리아</a></li>
<li><a href="/wiki/SBS_%EC%BD%98%ED%85%90%EC%B8%A0_%ED%97%88%EB%B8%8C" class="mw-redirect" title="SBS 콘텐츠 허브">SBS 콘텐츠 허브</a></li>
<li><a href="/wiki/SK%EC%BB%A4%EB%AE%A4%EB%8B%88%EC%BC%80%EC%9D%B4%EC%85%98%EC%A6%88" class="mw-redirect" title="SK커뮤니케이션즈">SK커뮤니케이션즈</a></li>
<li><a href="/wiki/%EC%86%8C%EB%8B%88_%EB%AE%A4%EC%A7%81_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="소니 뮤직 엔터테인먼트">소니 뮤직 코리아</a></li>
<li><a href="/wiki/%EC%86%8C%EB%A6%AC%EB%B0%94%EB%8B%A4" title="소리바다">소리바다</a></li>
<li><a href="/wiki/%EC%8A%A4%ED%86%A4%EB%AE%A4%EC%A7%81_%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8" title="스톤뮤직 엔터테인먼트">스톤뮤직 엔터테인먼트</a></li>
<li><a href="/wiki/%EC%9C%A0%EB%8B%88%EB%B2%84%EC%84%A4_%EB%AE%A4%EC%A7%81_%EA%B7%B8%EB%A3%B9" title="유니버설 뮤직 그룹">유니버설 뮤직 코리아</a></li>
<li><a href="/wiki/%EC%9B%8C%EB%84%88_%EB%AE%A4%EC%A7%81_%EA%B7%B8%EB%A3%B9" title="워너 뮤직 그룹">워너 뮤직 코리아</a></li></ul>
</div></td></tr></tbody></table><div></div></td></tr><tr style="display: none;"><td colspan="2" class="navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em"></div><table class="nowraplinks mw-collapsible mw-collapsed navbox-subgroup mw-made-collapsible" style="border-spacing:0"><tbody><tr><th scope="col" class="navbox-title" colspan="2"><button type="button" class="mw-collapsible-toggle mw-collapsible-toggle-default mw-collapsible-toggle-collapsed" aria-expanded="false" tabindex="0"><span class="mw-collapsible-text">펼치기</span></button><div id="음악_어워즈_쇼" style="font-size:114%;margin:0 4em">음악 어워즈 쇼</div></th></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">주요 어워드</th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%95%84%EC%8B%9C%EC%95%84_%EC%95%84%ED%8B%B0%EC%8A%A4%ED%8A%B8_%EC%96%B4%EC%9B%8C%EC%A6%88" title="아시아 아티스트 어워즈">아시아 아티스트 어워즈</a></li>
<li><a href="/wiki/%EC%95%84%EC%8B%9C%EC%95%84%EC%86%A1_%ED%8E%98%EC%8A%A4%ED%8B%B0%EB%B2%8C" title="아시아송 페스티벌">아시아송 페스티벌</a></li>
<li><a href="/wiki/%EA%B0%80%EC%98%A8%EC%B0%A8%ED%8A%B8_%EB%AE%A4%EC%A7%81_%EC%96%B4%EC%9B%8C%EC%A6%88" class="mw-redirect" title="가온차트 뮤직 어워즈">가온차트 뮤직 어워즈</a></li>
<li><a href="/wiki/%EA%B3%A8%EB%93%A0_%EB%94%94%EC%8A%A4%ED%81%AC" class="mw-redirect" title="골든 디스크">골든 디스크</a></li>
<li><a href="/wiki/KBS_%EA%B0%80%EC%9A%94%EB%8C%80%EC%B6%95%EC%A0%9C" title="KBS 가요대축제">KBS 가요대축제</a></li>
<li><a href="/wiki/%ED%95%9C%EA%B5%AD%EB%8C%80%EC%A4%91%EC%9D%8C%EC%95%85%EC%83%81" title="한국대중음악상">한국대중음악상</a></li>
<li><a href="/wiki/MBC_%EA%B0%80%EC%9A%94%EB%8C%80%EC%A0%9C%EC%A0%84" title="MBC 가요대제전">MBC 가요대제전</a></li>
<li><a href="/wiki/Mnet_20%27s_Choice" title="Mnet 20's Choice">Mnet 20's Choice</a></li>
<li><a href="/wiki/%EC%97%A0%EB%84%B7_%EC%95%84%EC%8B%9C%EC%95%88_%EB%AE%A4%EC%A7%81_%EC%96%B4%EC%9B%8C%EB%93%9C" title="엠넷 아시안 뮤직 어워드">엠넷 아시안 뮤직 어워드</a></li>
<li><a href="/wiki/SBS_%EA%B0%80%EC%9A%94%EB%8C%80%EC%A0%84" title="SBS 가요대전">SBS 가요대전</a></li>
<li><a href="/wiki/%ED%95%98%EC%9D%B4%EC%9B%90_%EC%84%9C%EC%9A%B8%EA%B0%80%EC%9A%94%EB%8C%80%EC%83%81" title="하이원 서울가요대상">하이원 서울가요대상</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">디지털 음악</th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%8B%B8%EC%9D%B4%EC%9B%94%EB%93%9C_%EB%94%94%EC%A7%80%ED%84%B8_%EB%AE%A4%EC%A7%81_%EC%96%B4%EC%9B%8C%EB%93%9C" title="싸이월드 디지털 뮤직 어워드">싸이월드 디지털 뮤직 어워드</a></li>
<li><a href="/wiki/%EB%A9%9C%EB%A1%A0_%EB%AE%A4%EC%A7%81_%EC%96%B4%EC%9B%8C%EB%93%9C" title="멜론 뮤직 어워드">멜론 뮤직 어워드</a></li></ul>
</div></td></tr></tbody></table><div></div></td></tr></tbody></table></div>
<div class="navbox-styles"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r38501015"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"></div><div role="navigation" class="navbox" aria-labelledby="음악" style="padding:3px"><table class="nowraplinks mw-collapsible autocollapse navbox-inner mw-made-collapsible mw-collapsed" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="col" class="navbox-title" colspan="3"><button type="button" class="mw-collapsible-toggle mw-collapsible-toggle-default mw-collapsible-toggle-collapsed" aria-expanded="false" tabindex="0"><span class="mw-collapsible-text">펼치기</span></button><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r38585741"><div class="navbar plainlinks hlist navbar-mini"><ul><li class="nv-보기"><a href="/wiki/%ED%8B%80:%EC%9D%8C%EC%95%85_%EC%A3%BC%EC%A0%9C" title="틀:음악 주제"><abbr title="이 틀을 보기" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">v</abbr></a></li><li class="nv-토론"><a href="/w/index.php?title=%ED%8B%80%ED%86%A0%EB%A1%A0:%EC%9D%8C%EC%95%85_%EC%A3%BC%EC%A0%9C&amp;action=edit&amp;redlink=1" class="new" title="틀토론:음악 주제 (없는 문서)"><abbr title="이 틀에 관해 토론하기" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">t</abbr></a></li><li class="nv-편집"><a href="/wiki/%ED%8A%B9%EC%88%98:%EB%AC%B8%EC%84%9C%ED%8E%B8%EC%A7%91/%ED%8B%80:%EC%9D%8C%EC%95%85_%EC%A3%BC%EC%A0%9C" title="특수:문서편집/틀:음악 주제"><abbr title="이 틀을 편집하기" style=";;background:none transparent;border:none;box-shadow:none;padding:0;">e</abbr></a></li></ul></div><div id="음악" style="font-size:114%;margin:0 4em"><a href="/wiki/%EC%9D%8C%EC%95%85" title="음악">음악</a></div></th></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%EC%9D%8C%EC%95%85%EC%9D%98_%EC%97%AD%EC%82%AC" title="음악의 역사">음악사</a></th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%84%A0%EC%82%AC_%EC%9D%8C%EC%95%85" title="선사 음악">선사</a></li>
<li><a href="/wiki/%EC%84%9C%EC%96%91%EC%9D%98_%EA%B3%A0%EB%8C%80_%EC%9D%8C%EC%95%85" title="서양의 고대 음악">고대</a></li>
<li><a href="/wiki/%EB%AF%BC%EC%9A%94" title="민요">민요</a></li>
<li><a href="/wiki/%EC%A2%85%EA%B5%90_%EC%9D%8C%EC%95%85" title="종교 음악">종교 음악</a></li>
<li><a href="/wiki/%EC%84%B8%EC%86%8D%EA%B0%80%EA%B3%A1" title="세속가곡">세속가곡</a></li>
<li><a href="/wiki/%EA%B5%B0%EC%95%85" class="mw-redirect" title="군악">군악</a></li></ul>
</div><table class="nowraplinks navbox-subgroup" style="border-spacing:0"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%ED%81%B4%EB%9E%98%EC%8B%9D_%EC%9D%8C%EC%95%85" title="클래식 음악">클래식</a></th><td class="navbox-list-with-group navbox-list navbox-even" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%84%9C%EC%96%91_%EC%A4%91%EC%84%B8_%EC%9D%8C%EC%95%85" title="서양 중세 음악">중세</a></li>
<li><a href="/wiki/%EB%A5%B4%EB%84%A4%EC%83%81%EC%8A%A4_%EC%9D%8C%EC%95%85" title="르네상스 음악">르네상스</a></li>
<li><a href="/wiki/%EB%B0%94%EB%A1%9C%ED%81%AC_%EC%9D%8C%EC%95%85" title="바로크 음악">바로크</a></li>
<li><a href="/wiki/%EA%B3%A0%EC%A0%84%EC%A3%BC%EC%9D%98_%EC%9D%8C%EC%95%85" title="고전주의 음악">고전주의</a></li>
<li><a href="/wiki/%EB%82%AD%EB%A7%8C%EC%A3%BC%EC%9D%98_%EC%9D%8C%EC%95%85" title="낭만주의 음악">낭만주의</a></li>
<li><a href="/wiki/%EC%9D%B8%EC%83%81%EC%A3%BC%EC%9D%98_%EC%9D%8C%EC%95%85" title="인상주의 음악">인상주의</a></li>
<li><a href="/wiki/20%EC%84%B8%EA%B8%B0_%EC%9D%8C%EC%95%85" title="20세기 음악">20세기</a></li>
<li><a href="/wiki/%ED%98%84%EB%8C%80_%ED%81%B4%EB%9E%98%EC%8B%9D_%EC%9D%8C%EC%95%85" title="현대 클래식 음악">현대의 클래식</a></li></ul>
</div></td></tr><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%EB%8C%80%EC%A4%91%EC%9D%8C%EC%95%85" title="대중음악">대중음악</a>/<br><a href="/wiki/%EC%8B%A4%EC%9A%A9%EC%9D%8C%EC%95%85" title="실용음악">실용음악</a></th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EB%A1%9D_%EC%9D%8C%EC%95%85" title="록 음악">록</a>
<ul><li><a href="/wiki/%ED%97%A4%EB%B9%84_%EB%A9%94%ED%83%88" title="헤비 메탈">메탈</a></li>
<li><a href="/wiki/%ED%8E%91%ED%81%AC_%EB%A1%9D" title="펑크 록">펑크 록</a></li>
<li><a href="/wiki/%EC%96%BC%ED%84%B0%EB%84%88%ED%8B%B0%EB%B8%8C_%EB%A1%9D" title="얼터너티브 록">얼터너티브 록</a></li></ul></li>
<li><a href="/wiki/%EB%B8%94%EB%A3%A8%EC%8A%A4" title="블루스">블루스</a>
<ul><li><a href="/wiki/%EB%A6%AC%EB%93%AC_%EC%95%A4_%EB%B8%94%EB%A3%A8%EC%8A%A4" title="리듬 앤 블루스">R&amp;B</a></li></ul></li>
<li><a href="/wiki/%EC%82%AC%EC%9D%B4%ED%82%A4%EB%8D%B8%EB%A6%AD_%EC%9D%8C%EC%95%85" title="사이키델릭 음악">사이키델릭</a></li>
<li><a href="/wiki/%EC%86%94_%EC%9D%8C%EC%95%85" title="솔 음악">소울</a></li>
<li><a href="/wiki/%EC%9E%AC%EC%A6%88" title="재즈">재즈</a></li>
<li><a href="/wiki/%EC%A0%84%EC%9E%90_%EC%9D%8C%EC%95%85" title="전자 음악">전자 음악</a>
<ul><li><a href="/wiki/%EC%9D%BC%EB%A0%89%ED%8A%B8%EB%A1%9C%EB%8B%88%EC%B9%B4" title="일렉트로니카">일렉트로니카</a></li>
<li><a href="/wiki/%EC%9D%BC%EB%A0%89%ED%8A%B8%EB%A1%9C%EB%8B%89_%EB%8C%84%EC%8A%A4_%EB%AE%A4%EC%A7%81" title="일렉트로닉 댄스 뮤직">EDM</a></li></ul></li>
<li><a href="/wiki/%EC%BB%A8%ED%8A%B8%EB%A6%AC_%EC%9D%8C%EC%95%85" title="컨트리 음악">컨트리</a></li>
<li><a href="/wiki/%ED%8C%9D_%EC%9D%8C%EC%95%85" title="팝 음악">팝</a>
<ul><li><a href="/wiki/%EC%8B%A0%EC%8A%A4%ED%8C%9D" title="신스팝">신스팝</a></li>
<li><a href="/wiki/%EB%8C%84%EC%8A%A4_%ED%8C%9D" title="댄스 팝">댄스 팝</a></li>
<li><a class="mw-selflink selflink">K-pop</a></li></ul></li>
<li><a href="/wiki/%ED%94%84%EB%A1%9C%EA%B7%B8%EB%A0%88%EC%8B%9C%EB%B8%8C_%EC%9D%8C%EC%95%85" title="프로그레시브 음악">프로그레시브</a></li>
<li><a href="/wiki/%ED%9E%99%ED%95%A9_%EC%9D%8C%EC%95%85" title="힙합 음악">힙합</a></li></ul>
</div></td></tr></tbody></table><div>
</div></td><td class="noviewer navbox-image" rowspan="7" style="width:1px;padding:0 0 0 2px"><div><span typeof="mw:File"><a href="https://commons.wikimedia.org/wiki/File:GClef.svg" class="mw-file-description" title="G (treble) clef symbol"><img alt="G (treble) clef symbol" src="//upload.wikimedia.org/wikipedia/commons/thumb/f/ff/GClef.svg/45px-GClef.svg.png" decoding="async" width="45" height="123" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/f/ff/GClef.svg/68px-GClef.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/f/ff/GClef.svg/90px-GClef.svg.png 2x" data-file-width="15" data-file-height="41"></a></span></div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">서양 <a href="/wiki/%EC%9D%8C%EC%95%85_%EC%9D%B4%EB%A1%A0" title="음악 이론">음악 이론</a>/<br><a href="/wiki/%EC%9E%91%EA%B3%A1" title="작곡">작곡</a></th><td class="navbox-list-with-group navbox-list navbox-even hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%9D%8C%EC%95%85_%EC%9E%A5%EB%A5%B4" title="음악 장르">음악 장르</a></li>
<li><a href="/wiki/%EC%95%85%EC%A0%84" title="악전">악전</a></li>
<li><a href="/wiki/%EB%8C%80%EC%9C%84%EB%B2%95" title="대위법">대위법</a></li>
<li><a href="/wiki/%ED%99%94%EC%84%B1_(%EC%9D%8C%EC%95%85)" title="화성 (음악)">화성</a></li>
<li><a href="/wiki/%EC%95%85%EA%B3%A1%EC%9D%98_%ED%98%95%EC%8B%9D" title="악곡의 형식">악곡의 형식</a></li>
<li><a href="/wiki/%EA%B8%B0%EB%B3%B4%EB%B2%95" title="기보법">기보법</a>
<ul><li><a href="/wiki/%EB%82%98%ED%83%80%EB%83%84%EB%A7%90" title="나타냄말">나타냄말</a></li></ul></li>
<li><a href="/wiki/%EC%9E%91%EA%B3%A1%EA%B0%80" title="작곡가">작곡가</a>
<ul><li><a href="/wiki/%EC%A6%89%ED%9D%A5_%EC%97%B0%EC%A3%BC" title="즉흥 연주">즉흥 연주</a></li></ul></li>
<li><a href="/wiki/%EC%9E%91%EC%82%AC%EA%B0%80" title="작사가">작사가</a></li>
<li><a href="/wiki/%EC%86%A1%EB%9D%BC%EC%9D%B4%ED%84%B0" title="송라이터">송라이터</a></li>
<li><a href="/wiki/%EB%85%B8%EB%9E%AB%EB%A7%90" title="노랫말">노랫말</a></li>
<li><a href="/wiki/%EB%85%B8%EB%9E%98" title="노래">노래</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">관련 학문</th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%9D%8C%EC%95%85%ED%95%99" title="음악학">음악학</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85%EC%82%AC%ED%95%99" title="음악사학">음악사학</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85_%EA%B3%A0%EA%B3%A0%ED%95%99" title="음악 고고학">음악 고고학</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85_%EC%8B%AC%EB%A6%AC%ED%95%99" class="mw-redirect" title="음악 심리학">음악 심리학</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%"><div class="hlist"><ul><li><a href="/wiki/%EA%B0%80%EC%B0%BD" title="가창">가창</a></li><li><a href="/wiki/%EC%97%B0%EC%A3%BC" class="mw-disambig" title="연주">연주</a></li></ul></div></th><td class="navbox-list-with-group navbox-list navbox-even hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%95%85%EB%8B%A8" title="악단">악단</a></li>
<li><a href="/wiki/%EB%B0%B4%EB%93%9C_(%EC%9D%8C%EC%95%85)" title="밴드 (음악)">밴드</a></li>
<li><a href="/wiki/%EB%B9%85%EB%B0%B4%EB%93%9C" title="빅밴드">빅밴드</a></li>
<li><a href="/w/index.php?title=%ED%98%91%EC%A3%BC%EB%8B%A8&amp;action=edit&amp;redlink=1" class="new" title="협주단 (없는 문서)">협주단</a></li>
<li><a href="/wiki/%EC%A7%80%ED%9C%98" title="지휘">지휘</a></li>
<li><a href="/wiki/%ED%95%A9%EC%B0%BD" title="합창">합창</a></li>
<li><a href="/wiki/%EB%94%94%EC%8A%A4%ED%81%AC_%EC%9E%90%ED%82%A4" title="디스크 자키">디스크자키</a></li>
<li><a href="/wiki/%EA%B5%B0%EC%95%85%EB%8C%80" title="군악대">군악대</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85%EA%B0%80" title="음악가">음악가</a></li>
<li><a href="/wiki/%EA%B4%80%ED%98%84%EC%95%85%EB%8B%A8" title="관현악단">관현악단</a></li>
<li><a href="/wiki/%EB%B3%B4%EC%BB%AC" title="보컬">보컬</a>
<ul><li><a href="/wiki/%EB%A6%AC%EB%93%9C_%EB%B3%B4%EC%BB%AC" title="리드 보컬">리드 보컬</a></li>
<li><a href="/wiki/%EC%BD%94%EB%9F%AC%EC%8A%A4_(%EB%8C%80%EC%A4%91%EC%9D%8C%EC%95%85)" title="코러스 (대중음악)">코러스</a></li></ul></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%EB%85%B9%EC%9D%8C" class="mw-redirect" title="녹음">녹음</a></th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%8B%B1%EA%B8%80_(%EC%9D%8C%EC%95%85)" title="싱글 (음악)">싱글</a>
<ul><li><a href="/wiki/A%EB%A9%B4%EA%B3%BC_B%EB%A9%B4" title="A면과 B면">A면과 B면</a></li>
<li><a href="/wiki/%EB%A6%AC%EB%93%9C_%EC%8B%B1%EA%B8%80" title="리드 싱글">리드 싱글</a>(선행 싱글)</li></ul></li>
<li><a href="/wiki/%EC%9D%8C%EB%B0%98" title="음반">음반</a>
<ul><li><a href="/wiki/%EC%A0%95%EA%B7%9C_%EC%9D%8C%EB%B0%98" title="정규 음반">앨범</a>(정규)</li>
<li><a href="/wiki/%EC%9D%B5%EC%8A%A4%ED%85%90%EB%94%94%EB%93%9C_%ED%94%8C%EB%A0%88%EC%9D%B4" title="익스텐디드 플레이">EP</a>(미니)</li>
<li><a href="/wiki/%EC%BB%B4%ED%95%84%EB%A0%88%EC%9D%B4%EC%85%98_%EC%9D%8C%EB%B0%98" title="컴필레이션 음반">컴필레이션</a></li>
<li><a href="/wiki/%EB%9D%BC%EC%9D%B4%EB%B8%8C_%EC%9D%8C%EB%B0%98" title="라이브 음반">라이브</a></li>
<li><a href="/wiki/%EB%A6%AC%EB%AF%B9%EC%8A%A4_%EC%9D%8C%EB%B0%98" title="리믹스 음반">리믹스</a></li>
<li><a href="/wiki/%EB%A6%AC%ED%8C%A8%ED%82%A4%EC%A7%80" title="리패키지">리패키지</a></li></ul></li>
<li><a href="/wiki/%EC%B6%95%EC%9D%8C%EA%B8%B0_%EC%9D%8C%EB%B0%98" title="축음기 음반">축음기 음반</a>
<ul><li><a href="/wiki/%ED%91%9C%EC%A4%80%EC%8B%9C%EA%B0%84_%EC%9D%8C%EB%B0%98" title="표준시간 음반">SP</a></li>
<li><a href="/wiki/%EC%9E%A5%EC%8B%9C%EA%B0%84_%EC%9D%8C%EB%B0%98" title="장시간 음반">LP</a></li>
<li><a href="/wiki/%EC%9D%B5%EC%8A%A4%ED%85%90%EB%94%94%EB%93%9C_%ED%94%8C%EB%A0%88%EC%9D%B4" title="익스텐디드 플레이">EP</a></li></ul></li>
<li><a href="/wiki/%EB%A0%88%EC%BD%94%EB%93%9C_%EB%A0%88%EC%9D%B4%EB%B8%94" title="레코드 레이블">레이블</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85_%ED%94%84%EB%A1%9C%EB%93%80%EC%84%9C" title="음악 프로듀서">프로듀서</a></li>
<li><a href="/wiki/%EC%98%A4%EB%94%94%EC%98%A4_%EC%97%94%EC%A7%80%EB%8B%88%EC%96%B4" title="오디오 엔지니어">오디오 엔지니어</a></li>
<li><a href="/wiki/%EC%83%98%ED%94%8C%EB%A7%81_(%EC%9D%8C%EC%95%85)" title="샘플링 (음악)">샘플링</a></li>
<li><a href="/wiki/%EB%A6%AC%EB%AF%B9%EC%8A%A4" title="리믹스">리믹스</a></li>
<li><a href="/wiki/%EC%BB%A4%EB%B2%84_(%EC%9D%8C%EC%95%85)" title="커버 (음악)">커버</a></li>
<li><a href="/w/index.php?title=%EC%9D%8C%EC%95%85%EA%B3%BC_%EA%B8%B0%EC%88%A0&amp;action=edit&amp;redlink=1" class="new" title="음악과 기술 (없는 문서)">음악과 기술</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">지역별 음악</th><td class="navbox-list-with-group navbox-list navbox-even hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%95%84%EC%8B%9C%EC%95%84%EC%9D%98_%EC%9D%8C%EC%95%85" title="아시아의 음악">아시아</a>
<ul><li><a href="/wiki/%ED%95%9C%EA%B5%AD%EC%9D%8C%EC%95%85" title="한국음악">한국</a></li>
<li><a href="/wiki/%EC%A4%91%EA%B5%AD_%EC%9D%8C%EC%95%85" title="중국 음악">중국</a></li>
<li><a href="/wiki/%EC%9D%BC%EB%B3%B8_%EC%9D%8C%EC%95%85" title="일본 음악">일본</a></li>
<li><a href="/wiki/%EC%9D%B8%EB%8F%84%EC%9D%98_%EC%9D%8C%EC%95%85" title="인도의 음악">인도</a></li>
<li><a href="/wiki/%EC%95%84%EB%9E%8D_%EC%9D%8C%EC%95%85" title="아랍 음악">아랍</a></li></ul></li>
<li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9C%A0%EB%9F%BD%EC%9D%98_%EC%9D%8C%EC%95%85" title="분류:유럽의 음악">유럽</a></li>
<li><a href="/wiki/%EB%B6%84%EB%A5%98:%EB%B6%81%EC%95%84%EB%A9%94%EB%A6%AC%EC%B9%B4%EC%9D%98_%EC%9D%8C%EC%95%85" title="분류:북아메리카의 음악">북아메리카</a></li>
<li><a href="/wiki/%EB%9D%BC%ED%8B%B4_%EC%95%84%EB%A9%94%EB%A6%AC%EC%B9%B4%EC%9D%98_%EC%9D%8C%EC%95%85" class="mw-redirect" title="라틴 아메리카의 음악">라틴 아메리카</a></li>
<li><a href="/wiki/%EC%95%84%ED%94%84%EB%A6%AC%EC%B9%B4%EC%9D%98_%EC%9D%8C%EC%95%85" title="아프리카의 음악">아프리카</a></li>
<li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%98%A4%EC%84%B8%EC%95%84%EB%8B%88%EC%95%84%EC%9D%98_%EC%9D%8C%EC%95%85" title="분류:오세아니아의 음악">오세아니아</a></li></ul>
</div></td></tr><tr style="display: none;"><th scope="row" class="navbox-group" style="width:1%">관련 문서</th><td class="navbox-list-with-group navbox-list navbox-odd hlist" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><a href="/wiki/%EC%95%85%EA%B8%B0" title="악기">악기</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85%EC%A0%9C" title="음악제">음악제</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85_%EA%B5%90%EC%9C%A1" title="음악 교육">음악 교육</a></li>
<li><a href="/wiki/%EC%9D%8C%EC%95%85_%EC%B9%98%EB%A3%8C" title="음악 치료">음악 치료</a></li></ul>
</div></td></tr><tr style="display: none;"><td class="navbox-abovebelow" colspan="3"><div><b><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9D%8C%EC%95%85" title="분류:음악">분류</a></b>&nbsp;<span style="font-weight:bold;">·</span>  <b><a href="/wiki/%ED%8F%AC%ED%84%B8:%EC%9D%8C%EC%95%85" title="포털:음악">포털</a></b></div></td></tr></tbody></table></div>
<div class="navbox-styles"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r36480591"><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r38501015"></div><div role="navigation" class="navbox authority-control" aria-label="Navbox" style="padding:3px"><table class="nowraplinks hlist navbox-inner" style="border-spacing:0;background:transparent;color:inherit"><tbody><tr><th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC:%EC%A0%84%EA%B1%B0_%ED%86%B5%EC%A0%9C" title="위키백과:전거 통제">전거 통제</a>: 국가 <span class="mw-valign-text-top noprint" typeof="mw:File/Frameless"><a href="https://www.wikidata.org/wiki/Q213665#identifiers" title="위키데이터에서 편집하기"><img alt="위키데이터에서 편집하기" src="//upload.wikimedia.org/wikipedia/commons/thumb/8/8a/OOjs_UI_icon_edit-ltr-progressive.svg/10px-OOjs_UI_icon_edit-ltr-progressive.svg.png" decoding="async" width="10" height="10" class="mw-file-element" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/8/8a/OOjs_UI_icon_edit-ltr-progressive.svg/15px-OOjs_UI_icon_edit-ltr-progressive.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/8/8a/OOjs_UI_icon_edit-ltr-progressive.svg/20px-OOjs_UI_icon_edit-ltr-progressive.svg.png 2x" data-file-width="20" data-file-height="20"></a></span></th><td class="navbox-list-with-group navbox-list navbox-odd" style="width:100%;padding:0"><div style="padding:0 0.25em">
<ul><li><span class="uid"><a rel="nofollow" class="external text" href="http://uli.nli.org.il/F/?func=find-b&amp;local_base=NLX10&amp;find_code=UID&amp;request=987012575222405171">이스라엘</a></span></li>
<li><span class="uid"><abbr title="Popular music--Korea (South)"><a rel="nofollow" class="external text" href="https://id.loc.gov/authorities/sh2019103591">미국</a></abbr></span>
<ul><li><span class="uid"><abbr title="K-pop (subculture)"><a rel="nofollow" class="external text" href="https://id.loc.gov/authorities/sh2020006575">2</a></abbr></span></li></ul></li>
<li><span class="uid"><abbr title="K-pop"><a rel="nofollow" class="external text" href="https://aleph.nkp.cz/F/?func=find-c&amp;local_base=aut&amp;ccl_term=ica=ph1064778&amp;CON_LNG=ENG">체코</a></abbr></span></li></ul>
</div></td></tr></tbody></table></div>
<!-- 
NewPP limit report
Parsed by mw‐api‐ext.eqiad.main‐577d96bfdc‐6gttj
Cached time: 20250319081414
Cache expiry: 2592000
Reduced expiry: false
Complications: [vary‐revision‐sha1, show‐toc]
CPU time usage: 0.720 seconds
Real time usage: 0.955 seconds
Preprocessor visited node count: 3942/1000000
Post‐expand include size: 142389/2097152 bytes
Template argument size: 3919/2097152 bytes
Highest expansion depth: 20/100
Expensive parser function count: 5/500
Unstrip recursion depth: 0/20
Unstrip post‐expand size: 94960/5000000 bytes
Lua time usage: 0.366/10.000 seconds
Lua memory usage: 5404201/52428800 bytes
Number of Wikibase entities loaded: 1/400
-->
<!--
Transclusion expansion time report (%,ms,calls,template)
100.00%  796.475      1 -total
 37.03%  294.964      2 틀:각주
 21.18%  168.731      1 틀:위키데이터_속성_추적
 18.46%  147.041     28 틀:웹_인용
 12.02%   95.730      7 틀:뉴스_인용
 10.25%   81.671      5 틀:둘러보기_상자
 10.23%   81.447      1 틀:음악_장르_정보
  9.90%   78.825      1 틀:정보상자
  7.91%   62.971      1 틀:K-pop
  7.77%   61.917      2 틀:Sister
-->

<!-- Saved in parser cache with key kowiki:pcache:329164:|#|:idhash:canonical and timestamp 20250319081414 and revision id 38932809. Rendering was triggered because: api-parse
 -->
</div><!--esi <esi:include src="/esitest-fa8a495983347898/content" /> --><noscript><img src="https://login.wikimedia.org/wiki/Special:CentralAutoLogin/start?useformat=desktop&amp;type=1x1&amp;usesul3=0" alt="" width="1" height="1" style="border: none; position: absolute;"></noscript>
<div class="printfooter" data-nosnippet="">원본 주소 "<a dir="ltr" href="https://ko.wikipedia.org/w/index.php?title=K-pop&amp;oldid=38932809">https://ko.wikipedia.org/w/index.php?title=K-pop&amp;oldid=38932809</a>"</div></div>
					<div id="catlinks" class="catlinks" data-mw="interface"><div id="mw-normal-catlinks" class="mw-normal-catlinks"><a href="/wiki/%ED%8A%B9%EC%88%98:%EB%B6%84%EB%A5%98" title="특수:분류">분류</a>: <ul><li><a href="/wiki/%EB%B6%84%EB%A5%98:K-pop" title="분류:K-pop">K-pop</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EB%8C%80%EC%A4%91%EB%AC%B8%ED%99%94" title="분류:대한민국의 대중문화">대한민국의 대중문화</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%ED%8C%9D_%EC%9D%8C%EC%95%85_%EC%9E%A5%EB%A5%B4" title="분류:팝 음악 장르">팝 음악 장르</a></li></ul></div><div id="mw-hidden-catlinks" class="mw-hidden-catlinks mw-hidden-cats-hidden">숨은 분류: <ul><li><a href="/wiki/%EB%B6%84%EB%A5%98:%ED%95%B4%EA%B2%B0%EB%90%98%EC%A7%80_%EC%95%8A%EC%9D%80_%EC%86%8D%EC%84%B1%EC%9D%B4_%EC%9E%88%EB%8A%94_%EB%AC%B8%EC%84%9C" title="분류:해결되지 않은 속성이 있는 문서">해결되지 않은 속성이 있는 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:CS1_-_%EC%98%81%EC%96%B4_%EC%9D%B8%EC%9A%A9_(en)" title="분류:CS1 - 영어 인용 (en)">CS1 - 영어 인용 (en)</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:CS1_-_%EB%AF%B8%EA%B5%AD_%EC%98%81%EC%96%B4_%EC%9D%B8%EC%9A%A9_(en)" title="분류:CS1 - 미국 영어 인용 (en)">CS1 - 미국 영어 인용 (en)</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9C%84%ED%82%A4%EB%8D%B0%EC%9D%B4%ED%84%B0_%EC%86%8D%EC%84%B1_P373%EC%9D%84_%EC%82%AC%EC%9A%A9%ED%95%98%EB%8A%94_%EB%AC%B8%EC%84%9C" title="분류:위키데이터 속성 P373을 사용하는 문서">위키데이터 속성 P373을 사용하는 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9C%84%ED%82%A4%EB%8D%B0%EC%9D%B4%ED%84%B0_%EC%86%8D%EC%84%B1_P244%EB%A5%BC_%EC%82%AC%EC%9A%A9%ED%95%98%EB%8A%94_%EB%AC%B8%EC%84%9C" title="분류:위키데이터 속성 P244를 사용하는 문서">위키데이터 속성 P244를 사용하는 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9C%84%ED%82%A4%EB%8D%B0%EC%9D%B4%ED%84%B0_%EC%86%8D%EC%84%B1_P691%EC%9D%84_%EC%82%AC%EC%9A%A9%ED%95%98%EB%8A%94_%EB%AC%B8%EC%84%9C" title="분류:위키데이터 속성 P691을 사용하는 문서">위키데이터 속성 P691을 사용하는 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9C%84%ED%82%A4%EB%8D%B0%EC%9D%B4%ED%84%B0_%EC%86%8D%EC%84%B1_P7859%EB%A5%BC_%EC%82%AC%EC%9A%A9%ED%95%98%EB%8A%94_%EB%AC%B8%EC%84%9C" title="분류:위키데이터 속성 P7859를 사용하는 문서">위키데이터 속성 P7859를 사용하는 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%9C%84%ED%82%A4%EB%8D%B0%EC%9D%B4%ED%84%B0_%EC%86%8D%EC%84%B1_P8189%EB%A5%BC_%EC%82%AC%EC%9A%A9%ED%95%98%EB%8A%94_%EB%AC%B8%EC%84%9C" title="분류:위키데이터 속성 P8189를 사용하는 문서">위키데이터 속성 P8189를 사용하는 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EC%98%81%EC%96%B4_%ED%91%9C%EA%B8%B0%EB%A5%BC_%ED%8F%AC%ED%95%A8%ED%95%9C_%EB%AC%B8%EC%84%9C" title="분류:영어 표기를 포함한 문서">영어 표기를 포함한 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:%EB%AA%A8%ED%98%B8%ED%95%9C_%ED%91%9C%ED%98%84%EC%9D%B4_%EA%B5%AC%EC%B2%B4%EC%A0%81%EC%9C%BC%EB%A1%9C_%EC%A7%80%EC%A0%81%EB%90%9C_%EB%AC%B8%EC%84%9C" title="분류:모호한 표현이 구체적으로 지적된 문서">모호한 표현이 구체적으로 지적된 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:J9U_%EC%8B%9D%EB%B3%84%EC%9E%90%EB%A5%BC_%ED%8F%AC%ED%95%A8%ED%95%9C_%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC_%EB%AC%B8%EC%84%9C" title="분류:J9U 식별자를 포함한 위키백과 문서">J9U 식별자를 포함한 위키백과 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:LCCN_%EC%8B%9D%EB%B3%84%EC%9E%90%EB%A5%BC_%ED%8F%AC%ED%95%A8%ED%95%9C_%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC_%EB%AC%B8%EC%84%9C" title="분류:LCCN 식별자를 포함한 위키백과 문서">LCCN 식별자를 포함한 위키백과 문서</a></li><li><a href="/wiki/%EB%B6%84%EB%A5%98:NKC_%EC%8B%9D%EB%B3%84%EC%9E%90%EB%A5%BC_%ED%8F%AC%ED%95%A8%ED%95%9C_%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC_%EB%AC%B8%EC%84%9C" title="분류:NKC 식별자를 포함한 위키백과 문서">NKC 식별자를 포함한 위키백과 문서</a></li></ul></div></div>
				</div>
````

````python
import requests
from bs4 import BeautifulSoup
import re

url = "https://ko.wikipedia.org/wiki/K-pop"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

# 1) 본문 div 찾기
content_div = soup.find("div", {"id": "mw-content-text"})
if not content_div:
    print("본문을 찾지 못했습니다.")
    exit()

# 2) 각주 sup 태그 제거
for sup_ref in content_div.find_all("sup", class_="reference"):
    sup_ref.decompose()

# 3) navbox(둘러보기 상자)나 table 같은 것 제거
for navbox in content_div.find_all("table", class_="navbox"):
    navbox.decompose()

# 필요하면, 아래처럼 infobox나 사이드바(table.infobox)도 제거할 수 있음
# for infobox in content_div.find_all("table", class_="infobox"):
#     infobox.decompose()

# 4) get_text()로 가져오되, separator로 공백 or 줄바꿈 등 지정
raw_text = content_div.get_text(separator="\n", strip=True)

# 5) [편집], [모호한 표현] 같은 괄호 표기도 정규식으로 제거
#   - \[.*?\] 는 대괄호로 감싸인 텍스트를 매칭
cleaned_text = re.sub(r"\[.*?\]", "", raw_text)

# 6) 여러 줄바꿈/공백 정리 (여러 개의 연속 공백 -> 1개로 치환)
cleaned_text = re.sub(r"\n+", "\n", cleaned_text)     # 연속된 개행을 1개로
cleaned_text = re.sub(r"[ \t]+", " ", cleaned_text)   # 연속된 공백/탭을 1개로

# 7) 앞뒤 공백 정리
final_text = cleaned_text.strip()

print(final_text)
````
