````python
import requests
from bs4 import BeautifulSoup
import json
import time

def collect_relic_links():
    """
    국립중앙박물관 웹사이트에서 문화재 정보 링크를 수집하는 함수
    
    Returns:
        list: 페이지 번호와 링크 주소를 포함하는 딕셔너리 리스트
    """
    base_url = "https://www.museum.go.kr/site/main/relic/treasure/list"
    all_links = []
    
    # 37페이지 순회
    for page in range(37):
        # 페이지 파라미터 설정 (startCount=0, 12, 24, ...)
        start_count = page * 12
        url = f"{base_url}?startCount={start_count}"
        
        try:
            # 웹페이지 요청
            response = requests.get(url)
            response.raise_for_status()  # 오류 발생 시 예외 발생
            
            # BeautifulSoup으로 HTML 파싱
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 카드 형태의 문화재 정보 찾기
            cards = soup.find_all('li', class_='card')
            
            # 각 카드에서 링크 추출
            for card in cards:
                link_tag = card.find('a', href=True)
                if link_tag and 'relicId=' in link_tag['href']:
                    link = link_tag['href']
                    all_links.append({
                        "page": page + 1,  # 페이지 번호는 1부터 시작
                        "link": link
                    })
            
            print(f"페이지 {page + 1}/37 처리 완료: {len(cards)}개 항목 발견")
            
            # 서버 부하 방지를 위한 1초 대기
            time.sleep(1)
            
        except Exception as e:
            print(f"페이지 {page + 1} 처리 중 오류 발생: {e}")
    
    return all_links

print("문화재 정보 링크 수집을 시작합니다...")
relic_links = collect_relic_links() 
print(f"총 {len(relic_links)}개의 링크를 수집하여 'relics_links.json' 파일에 저장했습니다.")
````

````python
# relic_links 데이터를 JSON 파일로 저장
import json

# 파일 경로 설정
file_path = 'relics_links.json'

# JSON 형식으로 파일에 저장
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(relic_links, f, ensure_ascii=False, indent=2)

print(f"총 {len(relic_links)}개의 링크를 '{file_path}' 파일에 저장했습니다.")
````

````python
import os
import json
import requests
from bs4 import BeautifulSoup
import re
import time
from urllib.parse import urljoin, urlparse

def extract_relic_data(url):
    """
    URL에서 유물 데이터를 추출하는 함수
    """
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 이미지 URL 추출
        thumbs = []
        swiper_slides = soup.select('.swiper-slide')
        for slide in swiper_slides:
            style = slide.get('style', '')
            img_match = re.search(r"background:url\('([^']+)'\)", style)
            if img_match:
                img_url = img_match.group(1)
                thumbs.append(img_url)
        
        # 메인 이미지는 첫 번째 이미지
        main_img = thumbs[0] if thumbs else ""
        
        # 라벨 정보 추출
        label_data = {}
        outview_list = soup.select('.outview-list li')
        
        # 명칭 추출
        title_elem = soup.select_one('.outveiw-tit')
        if title_elem:
            label_data['명칭'] = title_elem.text.strip()
        
        # 다른 라벨 정보 추출
        for item in outview_list:
            strong = item.select_one('strong')
            p = item.select_one('p')
            if strong and p:
                key = strong.text.strip()
                value = p.text.strip()
                label_data[key] = value
        
        # 내용 추출
        content_elem = soup.select_one('.view-info-cont p')
        content = content_elem.text.strip() if content_elem else ""
        
        # 저작권 이미지 URL 추출
        copyright_img_elem = soup.select_one('.codeCopyright img')
        copyright_img = copyright_img_elem.get('src') if copyright_img_elem else ""
        
        # relicId 추출
        relic_id = None
        if url:
            match = re.search(r'relicId=(\d+)', url)
            if match:
                relic_id = match.group(1)
        
        # 결과 데이터 구성
        relic_data = {
            'url': url,
            'img': main_img,
            'thumbs': thumbs,
            'label': label_data,
            'content': content,
            'copyright_img': copyright_img
        }
        
        return relic_data, relic_id
    
    except Exception as e:
        print(f"Error extracting data from {url}: {e}")
        return None, None

def save_relic_data(url, base_dir="relics"):
    """
    유물 데이터를 저장하는 함수
    """
    relic_data, relic_id = extract_relic_data(url)
    
    if not relic_data or not relic_id:
        print(f"Failed to extract data from {url}")
        return False
    
    # 폴더 생성
    folder_path = os.path.join(base_dir, relic_id)
    os.makedirs(folder_path, exist_ok=True)
    
    # JSON 파일 저장
    json_path = os.path.join(folder_path, "relic_data.json")
    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(relic_data, f, ensure_ascii=False, indent=2)
    
    # 이미지 다운로드 함수
    def download_image(img_url, folder):
        try:
            if not img_url.startswith(('http://', 'https://')):
                img_url = urljoin("https://www.museum.go.kr", img_url)
            
            response = requests.get(img_url, stream=True)
            response.raise_for_status()
            
            # 파일명 추출
            filename = os.path.basename(urlparse(img_url).path)
            file_path = os.path.join(folder, filename)
            
            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            
            print(f"Downloaded: {filename}")
            return True
        
        except Exception as e:
            print(f"Error downloading {img_url}: {e}")
            return False
    
    # 메인 이미지 다운로드
    if relic_data['img']:
        download_image(relic_data['img'], folder_path)
    
    # 썸네일 이미지 다운로드
    for thumb_url in relic_data['thumbs']:
        download_image(thumb_url, folder_path)
        # 서버 부하를 줄이기 위한 짧은 대기
        time.sleep(0.5)
    
    # 저작권 이미지 다운로드
    if relic_data['copyright_img']:
        download_image(relic_data['copyright_img'], folder_path)
    
    print(f"Successfully saved data for relic ID: {relic_id}")
    return True

def process_url_list(url_list, base_dir="relics"):
    """
    URL 목록을 처리하는 함수
    """
    os.makedirs(base_dir, exist_ok=True)
    
    success_count = 0
    fail_count = 0
    
    for i, url in enumerate(url_list, 1):
        print(f"Processing {i}/{len(url_list)}: {url}")
        
        if save_relic_data(url, base_dir):
            success_count += 1
        else:
            fail_count += 1
        
        # 서버 부하를 줄이기 위한 대기
        if i < len(url_list):
            time.sleep(2)
    
    print(f"\nProcessing complete. Success: {success_count}, Failed: {fail_count}")

# 사용 예시
if __name__ == "__main__":

    # relic_links에서 link만 추출하고 기본 URL을 붙여서 완전한 URL 목록 생성
    url_list = ["https://www.museum.go.kr" + item["link"] for item in relic_links]

    # 예시 URL 목록
    # url_list = [
    #     "https://www.museum.go.kr/site/main/relic/treasure/view?relicId=348",
    #     "https://www.museum.go.kr/site/main/relic/treasure/view?relicId=349",
    #     # 추가 URL을 여기에 넣을 수 있습니다
    # ]    
    
    process_url_list(url_list)
````

````python
# relics 폴더 내의 모든 유물 정보를 하나의 딕셔너리로 합치기
import os
import json

def combine_all_relics_info():
    combined_data = {}
    base_dir = "relics"
    copyright_img = "new_img_opentype01.png"
    relics_dir_list = sorted([relic_id for relic_id in os.listdir(base_dir)], key=int)   
    
    # relics 폴더 내의 모든 하위 폴더 확인
    for relic_id in relics_dir_list:
        relic_folder = os.path.join(base_dir, relic_id)

        # copyright_img 파일이 존재하는지 확인
        copyright_img_path = os.path.join(relic_folder, copyright_img)
        if not os.path.exists(copyright_img_path):
            print(f"{relic_id} 폴더에 new_img_opentype01.png 파일이 존재하지 않습니다.")
            continue
        
        # 디렉토리인지 확인
        if os.path.isdir(relic_folder):
            json_path = os.path.join(relic_folder, "relic_data.json")
            
            # relic_data.json 파일이 존재하는지 확인
            if os.path.exists(json_path):
                try:
                    with open(json_path, "r", encoding="utf-8") as f:
                        relic_data = json.load(f)
                    
                    # relic_id를 숫자로 변환하여 키로 사용
                    numeric_id = int(relic_id)
                    combined_data[numeric_id] = relic_data
                    #print(f"Added relic {relic_id} to combined data")
                except ValueError as ve:
                    print(f"Error converting {relic_id} to integer: {ve}")
                except Exception as e:
                    print(f"Error processing {relic_id}: {e}")
    
    # 결과 저장
    output_path = "combined_relics.json"
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(combined_data, f, ensure_ascii=False, indent=4)
    
    print(f"\n모든 유물 정보가 {output_path}에 저장되었습니다.")
    print(f"총 {len(combined_data)}개의 유물 정보가 결합되었습니다.")
    
    return combined_data

# 모든 유물 정보 합치기 실행
combined_relics = combine_all_relics_info()
````
